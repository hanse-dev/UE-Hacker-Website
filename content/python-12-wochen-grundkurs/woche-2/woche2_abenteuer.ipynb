## ğŸ“¦ Zauberformel 2: Die vier Elementtypen

**Was es ist:** Jede Information in Pythonia hat einen Elementtyp - wie die vier Elemente der Natur.

**Warum nÃ¼tzlich:**
- ğŸ”¥ **String (str):** FÃ¼r Texte, Namen und ZaubersprÃ¼che

- ğŸŒ **Float (float):** FÃ¼r BrÃ¼che, Prozente und Schadenswerte
- ğŸ’¨ **Boolean (bool):** FÃ¼r Wahrheitswerte, Ja/Nein-Entscheidungen

```python
# Beispiel 1: Strings - Textelemente
zauber = "Feuerball"
beschreibung = "Ein mÃ¤chtiger Feuerzauber"
print(f"Zauber: {zauber}")
print(f"Typ: {type(zauber)}")

# Beispiel 2: Integers - Ganze Zahlen
level = 12
lebenspunkte = 150
print(f"Level: {level}, LP: {lebenspunkte}")
print(f"Typ: {type(level)}")

# Beispiel 3: Floats - Kommazahlen
schaden = 23.5
prozent = 87.25
print(f"Schaden: {schaden}, Trefferchance: {prozent}%")
print(f"Typ: {type(schaden)}")

# Beispiel 4: Booleans - Wahrheitswerte
magie_aktiv = True
vergiftet = False
print(f"Magie aktiv: {magie_aktiv}")
print(f"Vergiftet: {vergiftet}")
print(f"Typ: {type(magie_aktiv)}")
```
