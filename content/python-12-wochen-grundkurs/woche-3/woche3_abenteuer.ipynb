### ğŸ¯ Skill-ErklÃ¤rung: and vs or

**and** = Beide mÃ¼ssen wahr sein:
- `True and True` â†’ True
- `True and False` â†’ False
- `False and False` â†’ False

**or** = Mindestens eines muss wahr sein:
- `True or True` â†’ True
- `True or False` â†’ True
- `False or False` â†’ False

**EselsbrÃ¼cke:** 
- **and** = streng (BEIDE Bedingungen)
- **or** = groÃŸzÃ¼gig (EINE Bedingung reicht)

````python
# ğŸ° Level 7: not-Operator

# Der not-Operator kehrt eine Bedingung um
vergiftet = False
verflucht = False
held_name = "Legolas"

print(f"ğŸ§™ Status von {held_name}:")
print(f"â˜ ï¸ Vergiftet: {vergiftet}")
print(f"ğŸ‘» Verflucht: {verflucht}")
print()

# not kehrt True zu False um (und umgekehrt)
if not vergiftet:
    print("âœ… Du bist nicht vergiftet!")

if not verflucht:
    print("âœ… Du bist nicht verflucht!")

# Kombiniere not mit and
if not vergiftet and not verflucht:
    print("ğŸ’ª Du bist in Topform!")
    print("âš”ï¸ Bereit fÃ¼r den Kampf!")

print()

# Noch ein Beispiel
quest_abgeschlossen = False

if not quest_abgeschlossen:
    print("ğŸ“œ Die Quest ist noch nicht abgeschlossen.")
    print("ğŸ¯ Mach weiter!")

print()
print("ğŸ‰ +250 XP: not-Operator verstanden!")
````

**Praktisch fÃ¼r:**
