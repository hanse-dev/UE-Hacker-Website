## üìö Daten-Struktur 1: Das Dictionary - Der digitale Schrank

**Was es ist:** Ein Dictionary ist eine Sammlung von Schl√ºssel-Wert-Paaren, die du mit geschweiften Klammern erstellst.

**Die Anatomie eines Dictionary:**
```python
# Leeres Dictionary erstellen
leeres_dict = {}

# Dictionary mit Elementen erstellen
schiff = {"name": "Nebula-Explorer", "typ": "Forschung", "crew": 150}
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`{}`** - Geschweifte Klammern definieren ein Dictionary
2. **`:`** - Doppelpunkt trennt Schl√ºssel vom Wert
3. **`,`** - Kommas trennen die Paare
4. **`[key]`** - Mit Schl√ºssel greifst du auf Werte zu

**Warum so n√ºtzlich:**
- üéØ Daten mit Namen statt Index organisieren
- üîç Schneller Zugriff √ºber Schl√ºssel
- ‚ö° Flexibel erweiterbar und √§nderbar
- üé≤ Verschiedene Datentypen mischbar

````python
# Beispiel 1: Einfache Dictionaries erstellen
# Leeres Dictionary
leeres_dict = {}
print(f"Leeres Dictionary: {leeres_dict}")

# Schiff-Dictionary
schiff = {
    "name": "Nebula-Explorer",
    "typ": "Forschung",
    "crew": 150,
    "geschwindigkeit": 0.8
}
print(f"Schiff-Dictionary: {schiff}")

# Planeten-Dictionary
planet = {
    "name": "Kepler-452b",
    "distanz": 1400,
    "typ": "Gesteinsplanet",
    "bewohnbar": True
}
print(f"Planet-Dictionary: {planet}")
````

````python
# Beispiel 2: Auf Dictionary-Elemente zugreifen
schiff = {
    "name": "Nebula-Explorer",
    "typ": "Forschung",
    "crew": 150,
    "geschwindigkeit": 0.8
}

print("=== Schiff-Zugriff ===")
print(f"Schiff-Name: {schiff['name']}")
print(f"Schiff-Typ: {schiff['typ']}")
print(f"Crew-Gr√∂√üe: {schiff['crew']}")
print(f"Geschwindigkeit: {schiff['geschwindigkeit']}c")

# Mit get() Methode (sicherer)
print(f"\nMit get(): {schiff.get('name')}")
print(f"Nicht existierend: {schiff.get('raumschiff', 'Unbekannt')}")
````

````python
# Beispiel 3: Dictionaries √§ndern und erweitern
schiff = {
    "name": "Nebula-Explorer",
    "typ": "Forschung",
    "crew": 150
}

print(f"Original: {schiff}")

# Wert √§ndern
schiff["crew"] = 200
print(f"Nach Crew-√Ñnderung: {schiff}")

# Neues Element hinzuf√ºgen
schiff["kapitaen"] = "Captain Alex"
print(f"Nach Kapit√§n hinzuf√ºgen: {schiff}")

# Element entfernen
entfernt = schiff.pop("typ")
print(f"Entfernt: {entfernt}")
print(f"Nach entfernen: {schiff}")
````

## üì¶ Daten-Struktur 2: Das Tupel - Die Datenkapsel

**Die wichtigsten Tupel-Eigenschaften:**

**Erstellung:**
- `tupel = (1, 2, 3)` - Mit runden Klammern
- `tupel = 1, 2, 3` - Ohne Klammern
- `tupel = (1,)` - Ein Element mit Komma

**Unver√§nderlichkeit:**
- Einmal erstellt, k√∂nnen Elemente nicht ge√§ndert werden
- Perfekt f√ºr konstante Daten
- Speichereffizienter als Listen

````python
# üîç √úbungen mit Tupeln

# Beispiel 1: Tupel erstellen
koordinaten = (100, 200, 300)
print(f"Koordinaten: {koordinaten}")

# Ohne Klammern
position = 10, 20, 30
print(f"Position: {position}")

# Ein Element
einzel = (42,)
print(f"Einzelnes Element: {einzel}")

# Zugriff auf Elemente
print(f"\nX-Koordinate: {koordinaten[0]}")
print(f"Y-Koordinate: {koordinaten[1]}")
print(f"Z-Koordinate: {koordinaten[2]}")
````

````python
# Beispiel 2: Tupel sind unver√§nderlich
system_status = ("Online", "Stabil", "Aktiv")
print(f"System-Status: {system_status}")

# Zugriff funktioniert
print(f"Status: {system_status[0]}")

# √Ñndern funktioniert NICHT!
try:
    system_status[0] = "Offline"
except TypeError as e:
    print(f"Fehler: {e}")

# Aber Tupel in Listen k√∂nnen ge√§ndert werden
systeme = [
    ("Navigation", "Online"),
    ("Antrieb", "Stabil"),
    ("Kommunikation", "Aktiv")
]
systeme[1] = ("Antrieb", "Offline")
print(f"\nSysteme nach √Ñnderung: {systeme}")
````

````python
# Beispiel 3: Tupel unpacking und Methoden
schiff_daten = ("Nebula-Explorer", "Forschung", 150, 0.8)

# Tupel unpacking
name, typ, crew, geschwindigkeit = schiff_daten
print("=== Tupel Unpacking ===")
print(f"Name: {name}")
print(f"Typ: {typ}")
print(f"Crew: {crew}")
print(f"Geschwindigkeit: {geschwindigkeit}")

# Tupel-Methoden
print(f"\nAnzahl Elemente: {len(schiff_daten)}")
print(f"Index von 'Forschung': {schiff_daten.index('Forschung')}")
print(f"Anzahl 'Forschung': {schiff_daten.count('Forschung')}")
````

## ‚ú® Daten-Struktur 3: Verschachtelte Strukturen - Das Daten-Netzwerk

**Die m√§chtigsten Kombinationen:**

**Dictionary in Dictionary:**
- `person = {"name": "Alex", "adresse": {"strasse": "X", "stadt": "Y"}}`

**Liste von Dictionaries:**
- `crew = [{"name": "A", "rolle": "P"}, {"name": "B", "rolle": "K"}]`

**Tupel als Dictionary-Schl√ºssel:**
- `koordinaten = {(10, 20): "Punkt A", (30, 40): "Punkt B"}`

````python
# Beispiel 1: Dictionary in Dictionary
raumstation = {
    "name": "Nebula-7",
    "position": {
        "sektor": "Alpha",
        "koordinaten": (100, 200, 300),
        "system": "Sol"
    },
    "systeme": {
        "antrieb": "Online",
        "lebenserhaltung": "Stabil",
        "kommunikation": "Aktiv"
    }
}

print("=== Verschachteltes Dictionary ===")
print(f"Station: {raumstation['name']}")
print(f"Sektor: {raumstation['position']['sektor']}")
print(f"Koordinaten: {raumstation['position']['koordinaten']}")
print(f"Antrieb: {raumstation['systeme']['antrieb']}")
````

````python
# Beispiel 2: Liste von Dictionaries
crew = [
    {
        "name": "Captain Alex",
        "rolle": "Kommandant",
        "alter": 35,
        "erfahrung": 15
    },
    {
        "name": "Dr. Zara",
        "rolle": "Wissenschaftlerin",
        "alter": 28,
        "erfahrung": 8
    },
    {
        "name": "Lt. Nova",
        "rolle": "Pilotin",
        "alter": 26,
        "erfahrung": 6
    }
]

print("=== Crew-Liste ===")
for mitglied in crew:
    print(f"{mitglied['name']} - {mitglied['rolle']} ({mitglied['alter']} Jahre)")

# Nach Erfahrung filtern
erfahrene = [m for m in crew if m['erfahrung'] > 10]
print(f"\nErfahrene Crew: {[m['name'] for m in erfahrene]}")
````

````python
# Beispiel 3: Komplexe Datenstrukturen
# Tupel als Dictionary-Schl√ºssel
galaxie_karte = {
    (100, 200): "Sonne A",
    (300, 400): "Sonne B",
    (500, 600): "Sonne C"
}

print("=== Galaxie-Karte ===")
for koordinate, stern in galaxie_karte.items():
    print(f"{koordinate}: {stern}")

# Komplexes Missionssystem
missionen = {
    "aktive": [
        {
            "id": 1,
            "ziel": (100, 200, 300),
            "crew": ["Alex", "Zara"],
            "status": "In Bearbeitung"
        }
    ],
    "abgeschlossen": [
        {
            "id": 2,
            "ziel": (400, 500, 600),
            "crew": ["Nova"],
            "status": "Erfolg"
        }
    ]
}

print(f"\nAktive Missionen: {len(missionen['aktive'])}")
print(f"Abgeschlossene Missionen: {len(missionen['abgeschlossen'])}")
````
