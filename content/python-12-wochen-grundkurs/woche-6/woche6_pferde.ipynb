## ğŸ“š Sammlungs-Technik 1: Die Listen-Erstellung - Der erste Plan

**Was es ist:** Eine Liste ist eine geordnete Sammlung von Elementen, die du mit eckigen Klammern erstellst.

**Die Anatomie einer Liste:**
```python
# Leere Liste erstellen
leere_liste = []

# Liste mit Elementen erstellen
pferde_liste = ["Thunder", "Luna", "Storm"]
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`[]`** - Eckige Klammern definieren eine Liste
2. **`,`** - Kommas trennen die Elemente
3. **`""`** - Strings werden in AnfÃ¼hrungszeichen gesetzt
4. **`[]`** - Mit Index greifst du auf Elemente zu (beginnt bei 0)

**Warum so nÃ¼tzlich:**
- ğŸ¯ Mehrere Werte in einer Variable speichern
- ğŸ” Geordnete Struktur fÃ¼r Daten
- âš¡ Flexibel erweiterbar und verÃ¤nderbar
- ğŸ² Verschiedene Datentypen mischbar

````python
# Beispiel 1: Einfache Listen erstellen
# Leere Liste
leere_liste = []
print(f"Leere Liste: {leere_liste}")

# Liste mit Pferden
pferde = ["Thunder", "Luna", "Storm", "Shadow"]
print(f"Pferde-Liste: {pferde}")

# Liste mit Zahlen
trainings_level = [1, 2, 3, 4, 5]
print(f"Trainings-Level: {trainings_level}")

# Gemischte Liste
mixed = ["Sattel", 25, True, "Zaumzeug"]
print(f"Gemischte Liste: {mixed}")
````

````python
# Beispiel 2: Auf Listenelemente zugreifen
ausruestung = ["Sattel", "Zaumzeug", "Gerte", "Satteldecke", "Hufpflege"]

print("=== AusrÃ¼stungs-Zugriff ===")
print(f"Gesamte AusrÃ¼stung: {ausruestung}")
print(f"Erstes Item: {ausruestung[0]}")
print(f"Zweites Item: {ausruestung[1]}")
print(f"Letztes Item: {ausruestung[-1]}")

# LÃ¤nge der Liste
print(f"Anzahl Items: {len(ausruestung)}")

# Slicing (Ausschnitte)
print(f"Erste 3 Items: {ausruestung[0:3]}")
print(f"Letzte 2 Items: {ausruestung[-2:]}")
````

## ğŸ“¦ Sammlungs-Technik 2: Elemente verwalten - Der Stallmeister

**Die wichtigsten Listen-Methoden:**

**Elemente hinzufÃ¼gen:**
- `append(element)` - FÃ¼gt am Ende hinzu
- `insert(index, element)` - FÃ¼gt an bestimmter Position ein

**Elemente entfernen:**
- `remove(element)` - Entfernt erstes Vorkommen
- `pop(index)` - Entfernt und gibt Element zurÃ¼ck
- `clear()` - Leert die gesamte Liste

````python
# ğŸ” Ãœbungen mit append() und insert()

# Beispiel 1: Elemente hinzufÃ¼gen
trainings_plan = []
print(f"Leerer Trainingsplan: {trainings_plan}")

# Mit append() am Ende hinzufÃ¼gen
trainings_plan.append("Longieren")
print(f"Nach append: {trainings_plan}")

trainings_plan.append("Dressur")
trainings_plan.append("Springen")
print(f"Voll gefÃ¼llt: {trainings_plan}")

# Mit insert() an bestimmter Position einfÃ¼gen
trainings_plan.insert(0, "AufwÃ¤rmen")  # Am Anfang
print(f"Nach insert(0): {trainings_plan}")

trainings_plan.insert(2, "Cavaletti")  # An Position 2
print(f"Nach insert(2): {trainings_plan}")
````

````python
# Beispiel 2: Elemente entfernen
futter_plan = ["Heu", "Hafer", "MÃ¼sli", "Karotten", "Ã„pfel", "MÃ¼sli"]
print(f"Original-Futterplan: {futter_plan}")

# Mit remove() erstes Vorkommen entfernen
futter_plan.remove("MÃ¼sli")
print(f"Nach remove('MÃ¼sli'): {futter_plan}")

# Mit pop() letztes Element entfernen und zurÃ¼ckgeben
letztes_futter = futter_plan.pop()
print(f"Nach pop(): {futter_plan}")
print(f"Entferntes Futter: {letztes_futter}")

# Mit pop(index) bestimmtes Element entfernen
zweites_futter = futter_plan.pop(1)
print(f"Nach pop(1): {futter_plan}")
print(f"Entferntes Futter: {zweites_futter}")

# Mit clear() alles lÃ¶schen
futter_plan.clear()
print(f"Nach clear(): {futter_plan}")
````

````python
# Beispiel 3: Listen kombinieren
pferde = ["Thunder", "Luna", "Storm"]
reiter = ["Anna", "Max", "Sarah"]
trainings = ["Dressur", "Springen", "Western"]

print("=== Listen kombinieren ===")
print(f"Pferde: {pferde}")
print(f"Reiter: {reiter}")
print(f"Trainings: {trainings}")

# Mit + Operator kombinieren
alle = pferde + reiter + trainings
print(f"Alles kombiniert: {alle}")

# Mit extend() eine Liste zur anderen hinzufÃ¼gen
pferde.extend(reiter)
print(f"Pferde erweitert: {pferde}")
````

## âœ¨ Sammlungs-Technik 3: Listen durchsuchen und sortieren - Der Ordnungs-Zauber

**Die mÃ¤chtigsten Such-Methoden:**

**Suchen und prÃ¼fen:**
- `index(element)` - Gibt Index des Elements zurÃ¼ck
- `element in liste` - PrÃ¼ft ob Element vorhanden
- `count(element)` - ZÃ¤hlt Vorkommen

**Sortieren:**
- `sort()` - Sortiert die Liste selbst
- `sorted(liste)` - Gibt sortierte Kopie zurÃ¼ck
- `reverse()` - Kehrt Reihenfolge um

````python
# Beispiel 1: In Listen suchen
pferde = ["Thunder", "Luna", "Storm", "Shadow", "Thunder", "Luna"]

print("=== In Listen suchen ===")
print(f"Pferde-Liste: {pferde}")

# Mit index() Position finden
position_luna = pferde.index("Luna")
print(f"Luna an Position: {position_luna}")

# Mit in() prÃ¼fen ob vorhanden
hat_storm = "Storm" in pferde
hat_bobby = "Bobby" in pferde
print(f"Storm vorhanden: {hat_storm}")
print(f"Bobby vorhanden: {hat_bobby}")

# Mit count() zÃ¤hlen
anzahl_thunder = pferde.count("Thunder")
print(f"Anzahl Thunder: {anzahl_thunder}")
````

````python
# Beispiel 2: Listen sortieren
level = [5, 2, 8, 1, 9, 3]
namen = ["Zara", "Anna", "Max", "Berta", "Klaus"]

print("=== Listen sortieren ===")
print(f"Original-Level: {level}")
print(f"Original-Namen: {namen}")

# Mit sort() direkt sortieren
level.sort()
print(f"Level nach sort(): {level}")

# Mit sorted() Kopie sortieren
sortierte_namen = sorted(namen)
print(f"Namen Original: {namen}")
print(f"Namen sortiert: {sortierte_namen}")

# Umgekehrt sortieren
level.sort(reverse=True)
print(f"Level absteigend: {level}")

# Mit reverse() Reihenfolge umkehren
namen.reverse()
print(f"Namen umgekehrt: {namen}")
````

````python
# Beispiel 3: Fortgeschrittene Listen-Operationen
disziplinen = ["Dressur", "Springen", "Western", "Dressur", "Western"]

print("=== Fortgeschrittene Operationen ===")
print(f"Disziplinen-Liste: {disziplinen}")

# Einzigartige Elemente (mit set)
einzigartige_disziplinen = list(set(disziplinen))
print(f"Einzigartige Disziplinen: {einzigartige_disziplinen}")

# Listen mit for-Schleife durchlaufen
print("\nAlle Disziplinen:")
for i, d in enumerate(disziplinen):
    print(f"  Position {i}: {d}")

# List Comprehension (fortgeschritten)
lange_namen = [d for d in disziplinen if len(d) > 6]
print(f"\nDisziplinen mit langen Namen: {lange_namen}")
````
