### ğŸ¯ Skill-ErklÃ¤rung: random.choices() vs random.choice()

- **`random.choice(liste)`** â€“ WÃ¤hlt EIN Element zufÃ¤llig
- **`random.choices(liste, weights, k)`** â€“ WÃ¤hlt k Elemente mit Gewichtung (mit ZurÃ¼cklegen)

````python
# ğŸ”¥ Level 7: Kombination von math und random

import math
import random

# Simuliere einen Zauberangriff mit Streuung
ziel_x, ziel_y = 10, 10
streuung = 2

# FÃ¼ge zufÃ¤llige Abweichung hinzu
treffer_x = ziel_x + random.uniform(-streuung, streuung)
treffer_y = ziel_y + random.uniform(-streuung, streuung)

# Berechne Abweichung vom Ziel
abweichung = math.sqrt((treffer_x - ziel_x)**2 + (treffer_y - ziel_y)**2)

print(f"ğŸ¯ Ziel-Position: ({ziel_x}, {ziel_y})")
print(f"ğŸ’¥ Treffer-Position: ({treffer_x:.2f}, {treffer_y:.2f})")
print(f"ğŸ“ Abweichung: {abweichung:.2f} Meter")

# Berechne Schadens-Reduktion basierend auf Abweichung
max_schaden = 100
schaden = max_schaden * math.exp(-abweichung / 3)  # Exponentieller Abfall

print(f"âš¡ Verursachter Schaden: {schaden:.1f}")

if abweichung < 0.5:
    print(f"ğŸ¯ VOLLTREFFER!")
elif abweichung < 1.5:
    print(f"âœ¨ Guter Treffer!")
else:
    print(f"ğŸ’¨ Knapp daneben...")
print()
print(f"ğŸ‰ +250 XP: Module kombiniert wie ein Meister!")
````

````python
# ğŸ† Level 8: Fortgeschrittene Anwendung â€“ Kampfsimulator

import math
import random

print(f"âš”ï¸ KAMPF-SIMULATOR âš”ï¸")
print(f"="*40)

# Spieler-Stats
spieler_hp = 100
spieler_angriff = 25
spieler_verteidigung = 10

# Gegner-Stats
gegner_hp = 80
gegner_angriff = 20
gegner_verteidigung = 8

runde = 1

while spieler_hp > 0 and gegner_hp > 0:
    print(f"\nğŸ”· Runde {runde}")
    print(f"â¤ï¸ Deine HP: {spieler_hp:.1f} | ğŸ’€ Gegner HP: {gegner_hp:.1f}")
    
    # Spieler greift an
    trefferchance = random.random()
    if trefferchance > 0.1:  # 90% Trefferchance
        # ZufÃ¤lliger Schadensfaktor
        schadensfaktor = random.uniform(0.8, 1.2)
        schaden = (spieler_angriff * schadensfaktor) - gegner_verteidigung
        schaden = max(1, schaden)  # Mindestens 1 Schaden
        
        # Kritischer Treffer?
        if random.random() < 0.15:  # 15% Krit-Chance
            schaden *= 2
            print(f"ğŸ’¥ KRITISCHER TREFFER! Du verursachst {schaden:.1f} Schaden!")
        else:
            print(f"âš”ï¸ Du triffst fÃ¼r {schaden:.1f} Schaden!")
        
        gegner_hp -= schaden
    else:
        print(f"ğŸ’¨ Dein Angriff verfehlt!")
    
    # PrÃ¼fe ob Gegner besiegt
    if gegner_hp <= 0:
        print(f"\nğŸ† SIEG! Du hast den Gegner besiegt!")
        break
    
    # Gegner greift an
    trefferchance = random.random()
    if trefferchance > 0.15:  # 85% Trefferchance
        schadensfaktor = random.uniform(0.8, 1.2)
        schaden = (gegner_angriff * schadensfaktor) - spieler_verteidigung
        schaden = max(1, schaden)
        print(f"ğŸ—¡ï¸ Gegner trifft dich fÃ¼r {schaden:.1f} Schaden!")
        spieler_hp -= schaden
    else:
        print(f"ğŸ›¡ï¸ Du blockst den Angriff!")
    
    # PrÃ¼fe ob Spieler besiegt
    if spieler_hp <= 0:
        print(f"\nğŸ’€ NIEDERLAGE! Du wurdest besiegt...")
        break
    
    runde += 1

print(f"\n" + "="*40)
print(f"ğŸ“Š Kampf endete nach {runde} Runden")
print()
print(f"ğŸ‰ +400 XP: Kampfsimulator gemeistert!")
````
