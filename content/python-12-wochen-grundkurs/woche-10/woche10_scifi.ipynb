## ğŸ›¸ Systemprotokoll 3: Der Konstruktor (__init__)

**Was es ist:** Der Konstruktor ist eine spezielle Methode, die automatisch ausgefÃ¼hrt wird, wenn ein Objekt initialisiert wird.

**Warum nÃ¼tzlich:**
- Objekte werden direkt mit richtigen Werten erstellt
- Du kann Startparameter Ã¼bergeben
- Initialisierungslogik wird zentralisiert

**Wichtig:** __init__ hat immer `self` als ersten Parameter! (Warum `self` nÃ¶tig ist, erfÃ¤hrst du in der nÃ¤chsten Sektion.)

````python
# Beispiel 1: Einfacher Konstruktor
class Android:
    def __init__(self, name):
        self.name = name

# Beispiel 2: Konstruktor mit mehreren Werten
class Planet:
    def __init__(self, name, typ, durchmesser):
        self.name = name
        self.typ = typ
        self.durchmesser = durchmesser

# Beispiel 3: Konstruktor mit Standardwerten
class Alien:
    def __init__(self, species="Greys", hp=50):
        self.species = species
        self.hp = hp
````

## ğŸª Warum `self`? â€“ Der SchlÃ¼ssel zu deinen Objekten

**Was ist `self`?** `self` ist die Referenz auf **das konkrete Objekt**, das gerade angesprochen wird. Wenn du `enterprise.starten()` aufrufst, ist `self` in der Methode genau das Objekt `enterprise`.

**Warum brauchen wir es Ã¼berhaupt?** Eine Klasse kann viele Objekte erzeugen. Python muss wissen: *Welches* dieser Objekte ist gemeint? Bei 10 Raumschiffen muss die Methode wissen: â€Ich arbeite fÃ¼r DIESES Schiff hier.â€œ

**Im Konstruktor `__init__`:**
- `self.name = name` bedeutet: **Speichere** den Ã¼bergebenen `name` auf DIESEM Objekt.
- Ohne `self` wÃ¼rde Python nicht wissen, wo die Variable hingehÃ¶rt â€“ sie wÃ¼rde nur lokal existieren und nach dem Aufruf verschwinden.

**In Methoden:**
- `self.name` bedeutet: **Lese den Namen** von DIESEM Objekt.
- Ohne `self` kÃ¶nnte die Methode nicht auf die Attribute des Objekts zugreifen â€“ sie wÃ¼rde nicht wissen, ob es Enterprise, Millennium oder ein anderes Schiff ist.

**So funktioniert der Aufruf im Hintergrund:**
- `enterprise.starten()` wird von Python zu `Raumschiff.starten(enterprise)` Ã¼bersetzt.
- Das erste Argument ist immer das Objekt selbst â€“ deshalb heiÃŸt der Parameter `self`.
