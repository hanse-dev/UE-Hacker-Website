## ğŸƒ Lektion 3: Der Konstruktor (__init__)

**Was es ist:** Der Konstruktor ist eine spezielle Methode, die automatisch ausgefÃ¼hrt wird, wenn ein Pferd geboren wird.

**Warum nÃ¼tzlich:**
- Pferde werden direkt mit richtigen Werten erstellt
- Du kannst Startwerte Ã¼bergeben
- Geburtsinformationen werden zentralisiert

**Wichtig:** __init__ hat immer `self` als ersten Parameter! (Warum `self` nÃ¶tig ist, erfÃ¤hrst du in der nÃ¤chsten Sektion.)

````python
# Beispiel 1: Einfacher Konstruktor
class Pferd:
    def __init__(self, name):
        self.name = name

# Beispiel 2: Konstruktor mit mehreren Werten
class Fohlen:
    def __init__(self, name, alter, geschlecht):
        self.name = name
        self.alter = alter
        self.geschlecht = geschlecht

# Beispiel 3: Konstruktor mit Standardwerten
class Pferd:
    def __init__(self, name="Stormy", rasse="Pony"):
        self.name = name
        self.rasse = rasse
````

## ğŸª Warum `self`? â€“ Der SchlÃ¼ssel zu deinen Pferden

**Was ist `self`?** `self` ist die Referenz auf **das konkrete Pferd**, das gerade angesprochen wird. Wenn du `lucky.wiehern()` aufrufst, ist `self` in der Methode genau das Objekt `lucky`.

**Warum brauchen wir es Ã¼berhaupt?** Eine Klasse kann viele Pferde erzeugen. Python muss wissen: *Welches* dieser Pferde ist gemeint? Bei 10 Pferden muss die Methode wissen: â€Ich arbeite fÃ¼r DIESES Pferd hier.â€œ

**Im Konstruktor `__init__`:**
- `self.name = name` bedeutet: **Speichere** den Ã¼bergebenen `name` auf DIESEM Pferd.
- Ohne `self` wÃ¼rde Python nicht wissen, wo die Variable hingehÃ¶rt â€“ sie wÃ¼rde nur lokal existieren und nach dem Aufruf verschwinden.

**In Methoden:**
- `self.name` bedeutet: **Lese den Namen** von DIESEM Pferd.
- Ohne `self` kÃ¶nnte die Methode nicht auf die Attribute des Objekts zugreifen â€“ sie wÃ¼rde nicht wissen, ob es Lucky, Spirit oder ein anderes Pferd ist.

**So funktioniert der Aufruf im Hintergrund:**
- `lucky.wiehern()` wird von Python zu `Pony.wiehern(lucky)` Ã¼bersetzt.
- Das erste Argument ist immer das Objekt selbst â€“ deshalb heiÃŸt der Parameter `self`.
