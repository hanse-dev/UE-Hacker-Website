## üìö Zauberformel 1: Der einfache Weg (if)

**Was es ist:** `if` pr√ºft eine Bedingung und f√ºhrt Code nur aus, wenn die Bedingung wahr ist.

**Warum n√ºtzlich:**
- üéØ Treffe Entscheidungen basierend auf Bedingungen
- üîç Pr√ºfe, ob etwas wahr oder falsch ist
- ‚öîÔ∏è Aktiviere F√§higkeiten nur bei Bedarf
- üé≤ Reagiere auf zuf√§llige Ereignisse

**Wichtig:** Der Code nach `if` muss einger√ºckt werden!

````python
# Beispiel 1: Einfache if-Bedingung
level = 5
if level >= 5:
    print("üéâ Du bist bereit f√ºr die Quest!")

# Beispiel 2: Pr√ºfen auf True/False
hat_schluessel = True
if hat_schluessel:
    print("üîë Die T√ºr √∂ffnet sich!")

# Beispiel 3: Text pr√ºfen
zauber = "Feuerball"
if zauber == "Feuerball":
    print("üî• Du wirkest einen Feuerball!")
````

## üîç Die Runen des Vergleichs: Wichtige Operatoren

### Zuweisung vs Vergleich: = vs ==

**Das ist ein entscheidender Unterschied!**

- **`=` (ein Gleichheitszeichen)**: Weist einer Variable einen Wert zu
  ```python
  level = 10  # Weise level den Wert 10 zu
  ```

- **`==` (zwei Gleichheitszeichen)**: Vergleicht zwei Werte auf Gleichheit
  ```python
  if level == 10:  # Pr√ºfe OB level 10 ist
      print("Level 10 erreicht!")
  ```

### Die vier Vergleichsrunen

- **`<` (kleiner als)**: Pr√ºft, ob der linke Wert kleiner ist
  ```python
  if level < 10:
      print("Noch nicht Level 10")
  ```

- **`<=` (kleiner oder gleich)**: Pr√ºft, ob der linke Wert kleiner oder gleich ist
  ```python
  if level <= 10:
      print("Level 10 oder niedriger")
  ```

- **`>` (gr√∂√üer als)**: Pr√ºft, ob der linke Wert gr√∂√üer ist
  ```python
  if level > 10:
      print("√úber Level 10!")
  ```

- **`>=` (gr√∂√üer oder gleich)**: Pr√ºft, ob der linke Wert gr√∂√üer oder gleich ist
  ```python
  if level >= 10:
      print("Level 10 oder h√∂her")
  ```

**Merke:** In Bedingungen immer `==` f√ºr Vergleiche verwenden, nie `=`!

````python
# üîç √úbungen zu Vergleichsoperatoren

# Beispiel 1: = vs == im Vergleich
level = 15  # Zuweisung mit =
if level == 15:  # Vergleich mit ==
    print("‚úÖ Richtig: level ist 15")

# Beispiel 2: Alle vier Vergleichsoperatoren
wert = 10
print(f"Wert: {wert}")
print(f"{wert} < 15: {wert < 15}")
print(f"{wert} <= 10: {wert <= 10}")
print(f"{wert} > 5: {wert > 5}")
print(f"{wert} >= 10: {wert >= 10}")

# Beispiel 3: Praktische Anwendung
gold = 100
kosten = 80
if gold > kosten:
    print(f"üõçÔ∏è Kaufen m√∂glich! √úbrig: {gold - kosten}")
elif gold == kosten:
    print("üí∞ Genug f√ºr genau einen Kauf!")
else:
    print(f"‚ùå Zu teuer! Fehlen: {kosten - gold}")
````

## üì¶ Zauberformel 2: Der zweigeteilte Weg (if-else)

**Was es ist:** `if-else` bietet zwei Pfade - einen f√ºr wahre und einen f√ºr falsche Bedingungen.

**Warum n√ºtzlich:**
- ‚úÖ Handle unterschiedlich bei Erfolg/Misserfolg
- üîÑ Gib alternative Antworten
- üéÆ Erstelle Win/Lose-Szenarien
- ‚öñÔ∏è Balance gute und schlechte Ergebnisse

**Wichtig:** `else` braucht keine Bedingung und wird bei allem anderen ausgef√ºhrt!

````python
# Beispiel 1: Erfolg oder Misserfolg
wurf = 18
if wurf >= 15:
    print("‚öîÔ∏è Kritischer Treffer!")
else:
    print("üí® Normaler Schlag")

# Beispiel 2: Alterspr√ºfung
alter = 17
if alter >= 18:
    print("üç∫ Du darfst betreten.")
else:
    print("üö´ Zu jung f√ºr diese Taverne.")

# Beispiel 3: Inventar-Check
gold = 45
kosten = 50
if gold >= kosten:
    print(f"üí∞ Kauf erfolgreich! Rest: {gold - kosten}")
else:
    print(f"‚ùå Nicht genug Gold! Fehlt: {kosten - gold}")
````

## ‚ú® Zauberformel 3: Der multiple Pfad (if-elif-else)

**Was es ist:** `if-elif-else` pr√ºft mehrere Bedingungen nacheinander.

**Warum n√ºtzlich:**
- üé≠ Handle unterschiedlich bei verschiedenen Situationen
- üåü Erstelle Stufen- und Level-Systeme
- üé® Gib personalisierte Antworten
- üèÜ Bewerte Leistungen mit Noten

**Wichtig:** `elif` braucht immer eine Bedingung! Nur der erste wahre Pfad wird ausgef√ºhrt.

````python
# Beispiel 1: Level-Bewertung
xp = 750
if xp >= 1000:
    print("üèÜ Meister!")
elif xp >= 500:
    print("‚≠ê Fortgeschritten!")
elif xp >= 100:
    print("üåü Anf√§nger!")
else:
    print("üìö Noch viel zu lernen...")

# Beispiel 2: Wetter-Zauber
element = "Wasser"
if element == "Feuer":
    print("üî• Feuerschlag!")
elif element == "Wasser":
    print("üíß Wasserhose!")
elif element == "Luft":
    print("üå™Ô∏è Windsto√ü!")
elif element == "Erde":
    print("üóø Erdbeben!")
else:
    print("‚ùì Unbekanntes Element!")

# Beispiel 3: Heldenklasse
staerke = 8
intelligenz = 12
if staerke > 10 and intelligenz > 10:
    print("‚öîÔ∏èüß† Paladin!")
elif staerke > 10:
    print("üí™ Krieger!")
elif intelligenz > 10:
    print("üßô Magier!")
else:
    print("üë§ Abenteurer!")
````
