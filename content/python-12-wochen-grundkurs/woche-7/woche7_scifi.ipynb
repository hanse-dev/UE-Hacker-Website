### ğŸ”® Skill-ErklÃ¤rung: random-Funktionen

Die wichtigsten Zufallsfunktionen:

- **`random.randint(a, b)`** â€“ Ganze Zufallszahl zwischen a und b (inklusive)

````python
# ğŸ§® Level 3: Orbital-Berechnungen

import math

# Berechne Orbital-Geschwindigkeit
planet_masse = 5.972e24  # kg (Erde)
gravitationskonstante = 6.674e-11
orbit_radius = 400000  # Meter (ISS-HÃ¶he + Erdradius)

# v = sqrt(G * M / r)
geschwindigkeit = math.sqrt(gravitationskonstante * planet_masse / orbit_radius)
geschwindigkeit_kmh = geschwindigkeit * 3.6

print(f"ğŸŒ Orbit-Radius: {orbit_radius/1000:.0f} km")
print(f"ğŸš€ Orbital-Geschwindigkeit: {geschwindigkeit:.2f} m/s")
print(f"ğŸ“Š Das entspricht: {geschwindigkeit_kmh:.2f} km/h")

# Berechne Umlaufzeit
umfang = 2 * math.pi * orbit_radius
umlaufzeit_sekunden = umfang / geschwindigkeit
umlaufzeit_minuten = umlaufzeit_sekunden / 60

print(f"â±ï¸ Umlaufzeit: {umlaufzeit_minuten:.2f} Minuten")

# Runde auf ganze Minuten
minuten_auf = math.ceil(umlaufzeit_minuten)
minuten_ab = math.floor(umlaufzeit_minuten)

print(f"â¬†ï¸ Aufgerundet: {minuten_auf} Minuten")
print(f"â¬‡ï¸ Abgerundet: {minuten_ab} Minuten")
print()
print(f"ğŸ‰ +200 Cyber Credits: Orbital-Mechanik gemeistert!")
````

````python
# ğŸ¯ Level 4: Asteroiden-Generator

import random

# Generiere ein zufÃ¤lliges Asteroidenfeld
anzahl_asteroiden = random.randint(10, 25)
print(f"â˜„ï¸ Asteroidenfeld mit {anzahl_asteroiden} Objekten")
print(f"="*50)

gefaehrliche = 0
gesamtmasse = 0

for i in range(anzahl_asteroiden):
    # ZufÃ¤llige Eigenschaften
    durchmesser = random.uniform(10, 500)  # Meter
    geschwindigkeit = random.uniform(5, 30)  # km/s
    distanz = random.uniform(100, 5000)  # km
    
    # SchÃ¤tze Masse (vereinfacht)
    volumen = (4/3) * 3.14159 * (durchmesser/2)**3
    masse = volumen * 2.5  # Dichte ca. 2.5 g/cmÂ³
    gesamtmasse += masse
    
    # GefÃ¤hrlichkeit bewerten
    if distanz < 1000 and geschwindigkeit > 20:
        status = "ğŸ”´ GEFÃ„HRLICH"
        gefaehrliche += 1
    elif distanz < 2000:
        status = "ğŸŸ¡ BEOBACHTEN"
    else:
        status = "ğŸŸ¢ SICHER"
    
    print(f"Asteroid {i+1}: {durchmesser:.1f}m, {geschwindigkeit:.1f}km/s, {distanz:.0f}km - {status}")

print(f"\nğŸ“Š Statistik:")
print(f"âš ï¸ GefÃ¤hrliche Asteroiden: {gefaehrliche}")
print(f"âš–ï¸ Gesamtmasse: {gesamtmasse/1e12:.2f} Millionen Tonnen")
print()
print(f"ğŸ‰ +250 Cyber Credits: Asteroidenfeld analysiert!")
````

````python
# ğŸŒŸ Level 5: Flugbahn-Berechnungen

import math

# Berechne die Flugbahn eines Projektils im Weltraum
start_x, start_y = 0, 0
ziel_x, ziel_y = 1000, 500  # km

# Berechne Distanz und Winkel
distanz = math.sqrt((ziel_x - start_x)**2 + (ziel_y - start_y)**2)
winkel_rad = math.atan2(ziel_y - start_y, ziel_x - start_x)
winkel_grad = math.degrees(winkel_rad)

print(f"ğŸ¯ Start: ({start_x}, {start_y})")
print(f"ğŸ¯ Ziel: ({ziel_x}, {ziel_y})")
print(f"ğŸ“ Distanz: {distanz:.2f} km")
print(f"ğŸ“ Winkel: {winkel_grad:.2f}Â°")

# Berechne benÃ¶tigte Geschwindigkeit (vereinfacht)
flugzeit = 60  # Sekunden
geschwindigkeit = distanz / flugzeit

print(f"â±ï¸ Flugzeit: {flugzeit} Sekunden")
print(f"ğŸš€ BenÃ¶tigte Geschwindigkeit: {geschwindigkeit:.2f} km/s")

# Berechne Treibstoffbedarf (exponentiell)
masse_schiff = 1000  # kg
delta_v = geschwindigkeit * 1000  # m/s
spezifischer_impuls = 300  # Sekunden

masse_verhaeltnis = math.exp(delta_v / (spezifischer_impuls * 9.81))
treibstoff = masse_schiff * (masse_verhaeltnis - 1)

print(f"â›½ Treibstoffbedarf: {treibstoff:.2f} kg")
print()
print(f"ğŸ‰ +200 Cyber Credits: Flugbahn berechnet!")
````

- 360Â° = 2Ï€ Radiant

### ğŸ¯ Skill-ErklÃ¤rung: random.choices() vs random.choice()

- **`random.choice(liste)`** â€“ WÃ¤hlt EIN Element zufÃ¤llig
- **`random.choices(liste, weights, k)`** â€“ WÃ¤hlt k Elemente mit Gewichtung (mit ZurÃ¼cklegen)
- **`random.sample(liste, k)`** â€“ WÃ¤hlt k Elemente ohne ZurÃ¼cklegen (keine Duplikate)

````python
# ğŸ”¥ Level 7: Kombination von math und random

import math
import random

# Simuliere einen Laser-Angriff mit Streuung
print(f"ğŸ”« LASER-ANGRIFFS-SIMULATION")
print(f"="*50)

ziel_x, ziel_y = 500, 300  # km
streuung = 50  # km

treffer = 0
schuesse = 10

for i in range(schuesse):
    # FÃ¼ge zufÃ¤llige Abweichung hinzu
    treffer_x = ziel_x + random.uniform(-streuung, streuung)
    treffer_y = ziel_y + random.uniform(-streuung, streuung)
    
    # Berechne Abweichung
    abweichung = math.sqrt((treffer_x - ziel_x)**2 + (treffer_y - ziel_y)**2)
    
    # Berechne Schadens-Reduktion (exponentieller Abfall)
    max_schaden = 100
    schaden = max_schaden * math.exp(-abweichung / 30)
    
    print(f"\nSchuss {i+1}:")
    print(f"  ğŸ“ Position: ({treffer_x:.1f}, {treffer_y:.1f})")
    print(f"  ğŸ“ Abweichung: {abweichung:.2f} km")
    print(f"  ğŸ’¥ Schaden: {schaden:.1f}%")
    
    if abweichung < 20:
        print(f"  âœ… VOLLTREFFER!")
        treffer += 1
    elif abweichung < 40:
        print(f"  âš ï¸ Getroffen")
        treffer += 1
    else:
        print(f"  âŒ Verfehlt")

trefferquote = (treffer / schuesse) * 100
print(f"\nğŸ“Š Trefferquote: {treffer}/{schuesse} ({trefferquote:.1f}%)")
print()
print(f"ğŸ‰ +250 Cyber Credits: Waffensystem kalibriert!")
````

### ğŸŒŸ Wichtiges Konzept: Exponentialfunktion

`math.exp(x)` berechnet e^x (Eulersche Zahl hoch x). Das ist nÃ¼tzlich fÃ¼r:
- Schadens-Abfall Ã¼ber Distanz
- Radioaktiver Zerfall
- SignalstÃ¤rke-Abnahme
- Wachstumsprozesse

````python
# ğŸ† Level 8: Fortgeschrittene Anwendung â€“ Raumschlacht-Simulator

import math
import random

print(f"âš”ï¸ RAUMSCHLACHT-SIMULATOR âš”ï¸")
print(f"="*50)

# Schiffs-Stats
spieler_schilde = 100
spieler_hull = 100
spieler_waffen = 25

gegner_schilde = 80
gegner_hull = 80
gegner_waffen = 20

runde = 1

while (spieler_schilde > 0 or spieler_hull > 0) and (gegner_schilde > 0 or gegner_hull > 0):
    print(f"\nâš¡ Runde {runde}")
    print(f"ğŸ›¡ï¸ Deine Schilde: {spieler_schilde:.1f} | ğŸš€ Hull: {spieler_hull:.1f}")
    print(f"ğŸ›¡ï¸ Gegner Schilde: {gegner_schilde:.1f} | ğŸš€ Hull: {gegner_hull:.1f}")
    
    # Spieler greift an
    trefferchance = random.random()
    if trefferchance > 0.15:  # 85% Trefferchance
        schadensfaktor = random.uniform(0.8, 1.3)
        schaden = spieler_waffen * schadensfaktor
        
        # Kritischer Treffer?
        if random.random() < 0.2:  # 20% Krit-Chance
            schaden *= 1.5
            print(f"ğŸ’¥ KRITISCHER TREFFER! {schaden:.1f} Schaden!")
        else:
            print(f"ğŸ”« Treffer fÃ¼r {schaden:.1f} Schaden!")
        
        # Schaden auf Schilde, dann Hull
        if gegner_schilde > 0:
            if schaden > gegner_schilde:
                rest_schaden = schaden - gegner_schilde
                gegner_schilde = 0
                gegner_hull -= rest_schaden
                print(f"  âš ï¸ Schilde durchbrochen! {rest_schaden:.1f} Hull-Schaden!")
            else:
                gegner_schilde -= schaden
        else:
            gegner_hull -= schaden
    else:
        print(f"ğŸ’¨ Schuss verfehlt!")
    
    # PrÃ¼fe ob Gegner zerstÃ¶rt
    if gegner_hull <= 0:
        print(f"\nğŸ† SIEG! Gegnerisches Schiff zerstÃ¶rt!")
        break
    
    # Gegner greift an
    trefferchance = random.random()
    if trefferchance > 0.20:  # 80% Trefferchance
        schadensfaktor = random.uniform(0.8, 1.3)
        schaden = gegner_waffen * schadensfaktor
        print(f"ğŸ”´ Gegner trifft fÃ¼r {schaden:.1f} Schaden!")
        
        if spieler_schilde > 0:
            if schaden > spieler_schilde:
                rest_schaden = schaden - spieler_schilde
                spieler_schilde = 0
                spieler_hull -= rest_schaden
                print(f"  âš ï¸ Deine Schilde sind down! {rest_schaden:.1f} Hull-Schaden!")
            else:
                spieler_schilde -= schaden
        else:
            spieler_hull -= schaden
    else:
        print(f"âœ… Gegner verfehlt!")
    
    # PrÃ¼fe ob Spieler zerstÃ¶rt
    if spieler_hull <= 0:
        print(f"\nğŸ’€ NIEDERLAGE! Dein Schiff wurde zerstÃ¶rt...")
        break
    
    # Schild-Regeneration (10% pro Runde)
    if spieler_schilde > 0 and spieler_schilde < 100:
        spieler_schilde = min(100, spieler_schilde + 10)
    if gegner_schilde > 0 and gegner_schilde < 80:
        gegner_schilde = min(80, gegner_schilde + 8)
    
    runde += 1
    
    # Maximale Rundenzahl
    if runde > 20:
        print(f"\nâ±ï¸ ZEITLIMIT! Schlacht unentschieden.")
        break

print(f"\n" + "="*50)
print(f"ğŸ“Š Schlacht endete nach {runde} Runden")
print()
print(f"ğŸ‰ +400 Cyber Credits: Kampfsystem gemeistert!")
````
