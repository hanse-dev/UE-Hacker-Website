## üìö Systemprotokoll 1: Die for-Schleife - Der Zeit-Loop

**Was es ist:** Eine for-Schleife wiederholt Code f√ºr jedes Element in einer Sequenz.

**Die Anatomie einer for-Schleife:**
```python
for variable in sequenz:
    # Code, der f√ºr jedes Element ausgef√ºhrt wird
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`for`** - Das Kommando, das den Zeit-Loop startet
2. **`variable`** - Ein Platzhalter, der bei jedem Durchlauf das aktuelle Element enth√§lt
3. **`in`** - Verbindet die Variable mit der Sequenz
4. **`sequenz`** - Die Sammlung, durch die wir iterieren (z.B. Liste, String, range())
5. **`:`** - Beendet die Schleifen-Definition
6. **Einger√ºckter Code** - Wird bei jedem Durchlauf ausgef√ºhrt

**Der Iterator:**
- Der Iterator ist wie ein unsichtbarer Zeit-Messer, der sich merkt, wo wir in der Sequenz sind
- Er startet beim ersten Element
- Nach jedem Durchlauf springt er automatisch zum n√§chsten Element
- Die Schleife endet, wenn kein Element mehr √ºbrig ist

**Warum so n√ºtzlich:**
- üéØ Automatisiere wiederholende Aufgaben
- üîç Verarbeite jedes Element einzeln
- ‚ö° Mache Code effizienter und k√ºrzer
- üé≤ Erstelle Muster und Sequenzen

````python
# Beispiel 1: Einfache for-Schleife mit range()
print("=== Beispiel 1: 5 Zeit-Zyklen ===")
# range(5) erzeugt die Zahlen 0, 1, 2, 3, 4
for i in range(5):
    print(f"Zeit-Zyklus {i+1}: Systeme synchronisiert")

print("\n=== Was passiert hier? ===")
print("1. range(5) erstellt: [0, 1, 2, 3, 4]")
print("2. i wird bei jedem Durchlauf auf das n√§chste Element gesetzt")
print("3. Wir geben i+1 aus, um von 1-5 zu z√§hlen")
print("4. Die Schleife endet nach dem letzten Element")
````

````python
# Beispiel 2: For-Schleife mit Start und Ende
print("=== Beispiel 2: Sektoren 3 bis 7 scannen ===")
# range(3, 8) erzeugt die Zahlen 3, 4, 5, 6, 7 (8 ist exklusiv!)
for sektor in range(3, 8):
    print(f"Sektor {sektor}: Scan abgeschlossen")

print("\n=== Wichtige Beobachtung ===")
print("range(start, ende) enth√§lt start, aber nicht ende!")
print("Wie bei Koordinaten: von 3 bis VOR 8")
````

````python
# Beispiel 3: For-Schleife mit Schrittweite
print("=== Beispiel 3: Jeder zweite Sektor ===")
# range(0, 11, 2) erzeugt: 0, 2, 4, 6, 8, 10
for sektor in range(0, 11, 2):
    print(f"Sektor {sektor}: Verst√§rkung aktiviert")

print("\n=== Die Schrittweite ===")
print("range(start, ende, schritt) √ºberspringt Sektoren")
print("Hier: jeder zweite Sektor √ºberspringen")
````

````python
# Beispiel 4: For-Schleife √ºber einen String
print("=== Beispiel 4: Buchstaben eines Codes ===")
code = "ALPHA"
print(f"Der Code ist: {code}")
print("Buchstaben f√ºr Buchstabe:")

for buchstabe in code:
    print(f"  - {buchstabe}")

print("\n=== Was passiert hier? ===")
print("String wird als Sequenz von Buchstaben behandelt")
print("Jeder Buchstabe wird einzeln durchlaufen")
````

````python
# Beispiel 5: For-Schleife √ºber eine Liste
print("=== Beispiel 5: Planeten-System scannen ===")
planeten = ["Merkur", "Venus", "Erde", "Mars", "Jupiter"]
print("Planeten im System:")

for planet in planeten:
    print(f"  - {planet}")

print("\n=== Mit Index und Name ===")
# enumerate gibt uns sowohl Index als auch Wert
for index, planet in enumerate(planeten, start=1):
    print(f"  Planet {index}: {planet}")
````

## üîç Das Geheimnis von range() - Der Zeit-Generator

**range() ist dein m√§chtigstes Werkzeug f√ºr Zeit-Sequenzen!**

### Die drei Formen von range():

1. **`range(n)`** - Zahlen von 0 bis n-1
   ```python
   range(5) ‚Üí 0, 1, 2, 3, 4
   ```

2. **`range(start, ende)`** - Zahlen von start bis ende-1
   ```python
   range(2, 7) ‚Üí 2, 3, 4, 5, 6
   ```

3. **`range(start, ende, schritt)`** - Zahlen mit Abst√§nden
   ```python
   range(1, 10, 2) ‚Üí 1, 3, 5, 7, 9
   ```

**Wichtige Regeln:**
- Die **obere Grenze ist immer exklusiv** (wird nicht mitgez√§hlt)
- **Negative Schritte** sind m√∂glich f√ºr r√ºckw√§rts z√§hlen
- range() ist **sehr speichereffizient** (erzeugt Zahlen erst bei Bedarf)

````python
# üîç √úbungen mit range()

# √úbung 1: Countdown zum Sprung
print("=== Countdown zum Hypersprung ===")
for i in range(5, 0, -1):
    print(f"{i}...")
print("üöÄ HYPERSPRUNG!")

# √úbung 2: Gerade Zeitstempel
print("\n=== Gerade Zeitstempel ===")
for zeit in range(0, 21, 2):
    print(f"T{zeit:02d}", end=" ")
print()

# √úbung 3: Energie-Zyklen
print("\n=== Energie-Zyklen ===")
for zyklus in range(1, 6):
    print(f"Zyklus {zyklus}: Energie-Wellen aktiv")
````

## üì¶ Systemprotokoll 2: Die while-Schleife - Der Dauer-Scanner

**Was es ist:** Eine while-Schleife wiederholt Code, solange eine Bedingung wahr ist.

**Die Anatomie einer while-Schleife:**
```python
while bedingung:
    # Code, der wiederholt wird, solange bedingung wahr ist
    # WICHTIG: Die Bedingung muss irgendwann falsch werden!
```

**Warum n√ºtzlich:**
- ‚è≥ Wiederhole, bis etwas eintritt
- üéÆ Daueraufgaben im Hintergrund
- üîç Sensoren √ºberwachen
- ‚ö° Unbekannte Anzahl von Wiederholungen

**‚ö†Ô∏è Vorsicht:** Endlosschleifen entstehen, wenn die Bedingung niemals falsch wird!

````python
# Beispiel 1: Einfache while-Schleife
print("=== Beispiel 1: Bis 10 scannen ===")
scan = 0
while scan <= 10:
    print(f"Scan {scan}: Daten empfangen")
    scan += 1  # WICHTIG: Hochz√§hlen, sonst Endlosschleife!

print("\n=== Was passiert hier? ===")
print("1. scan startet bei 0")
print("2. Solange scan <= 10, wird der Code ausgef√ºhrt")
print("3. scan += 1 erh√∂ht den Wert bei jedem Durchlauf")
print("4. Bei scan = 11 wird die Bedingung falsch ‚Üí Schleife endet")
````

````python
# Beispiel 2: Schild-Generator
print("=== Beispiel 2: Schild-Generator ===")
energie = 100
minute = 1

while energie > 20:
    verbrauch = 15
    energie -= verbrauch
    print(f"Minute {minute}: -{verbrauch} Energie, √ºbrig: {energie}%")
    minute += 1

print(f"\n‚ö†Ô∏è Schild-Kritisch nach {minute-1} Minuten!")
print("Energie nachladen!")
````

````python
# Beispiel 3: Zugangscodes
print("=== Beispiel 3: Zugangscodes ===")
code = ""
versuche = 0
max_versuche = 3

while code != "NEXUS" and versuche < max_versuche:
    code = input("Gib den Zugangscode ein: ")
    versuche += 1
    
    if code == "NEXUS":
        print("‚úÖ Zugang gew√§hrt!")
    elif versuche >= max_versuche:
        print("‚ùå Zu viele Versuche! System gesperrt.")
    else:
        print(f"Falsch! Noch {max_versuche - versuche} Versuche")
````

## ‚ú® Systemprotokoll 3: break und continue - Die Zeit-Manipulatoren

**break** - Beendet den Zeit-Loop sofort
**continue** - Springt zum n√§chsten Zyklus

**Warum n√ºtzlich:**
- üõë Fr√ºhzeitiges Beenden bei Gefahr
- ‚è≠Ô∏è Bestimmte Zyklen √ºberspringen
- üéØ Effiziente System-Optimierung
- ‚ö° Energie sparen

````python
# Beispiel 1: break bei Alarm
print("=== Beispiel 1: System-Scan bei Alarm beenden ===")
systeme = ["Kern", "Schild", "Waffen", "ANOMALIE", "Antrieb"]

for system in systeme:
    print(f"Scanne: {system}")
    if system == "ANOMALIE":
        print("üö® ANOMALIE ENTDECKT! Scan abgebrochen.")
        break  # Schleife sofort beenden

print("\n=== Ohne break w√ºrden alle Systeme gescannt ===")
````

````python
# Beispiel 2: continue f√ºr Filter
print("=== Beispiel 2: Nur Kritische Systeme ===")
systeme = ["Licht", "üö® REAKTOR", "Klima", "‚ö†Ô∏è SCHILD", "Musik"]

print("Kritische Systeme:")
for system in systeme:
    # √úberspringe, wenn kein Warn-Symbol
    if "üö®" not in system and "‚ö†Ô∏è" not in system:
        continue  # Zum n√§chsten Durchlauf springen
    print(f"  - {system}")
````

````python
# Beispiel 3: Komplexe Zeit-Manipulation
print("=== Beispiel 3: Missions-Protokoll ===")
stunde = 1
max_stunden = 10

while stunde <= max_stunden:
    print(f"\nStunde {stunde}:")
    
    # 3 System-Checks pro Stunde
    for check in range(3):
        print(f"  Check {check+1}: Systeme OK")
        
        # Spezial-Event bei Stunde 5
        if stunde == 5 and check == 1:
            print("  üåü Solar-Flare: Zus√§tzliche Scans!")
            continue
    
    stunde += 1
    
    # Mission abbrechen bei St√∂rung
    if stunde > 7:
        print("\n‚ö° Energie-St√∂rung! Mission abgebrochen.")
        break
````
