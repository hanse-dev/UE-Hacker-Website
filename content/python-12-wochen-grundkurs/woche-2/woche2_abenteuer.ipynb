### ğŸ” Skill: Datentypen erkennen

Jede Variable hat einen **Typ**. Mit der magischen Funktion `type()` kannst du herausfinden, welchen Typ eine Variable hat:

- `str` â†’ String (Text)
- `int` â†’ Integer (Ganze Zahl)
- `float` â†’ Float (Kommazahl)
- `bool` â†’ Boolean (True/False)

**Warum ist das wichtig?** Verschiedene Typen haben verschiedene FÃ¤higkeiten! Du kannst Zahlen addieren, aber nicht Texte multiplizieren (naja, fast...).

````python
# ğŸ® Level 2: Datentypen prÃ¼fen

name = "Gandalf"
alter = 2000
groesse = 1.85
ist_magier = True

print("ğŸ” Datentypen-Analyse:")
print(f"Name ist vom Typ: {type(name).__name__}")
print(f"Alter ist vom Typ: {type(alter).__name__}")
print(f"GrÃ¶ÃŸe ist vom Typ: {type(groesse).__name__}")
print(f"Ist Magier ist vom Typ: {type(ist_magier).__name__}")

print(f"\nğŸ‰ +150 XP: Du kannst jetzt Datentypen erkennen!")
````

### âš¡ Power-Moves: Mit Zahlen rechnen

Variablen mit Zahlen kÃ¶nnen wie normale Zahlen verwendet werden:

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `%` Modulo (Rest bei Division)
- `**` Potenz

````python
# ğŸ® Level 3: Mathematische Operationen

# Grundrechenarten
schaden = 25
verteidigung = 10
echter_schaden = schaden - verteidigung
print(f"âš”ï¸ Schaden: {schaden}")
print(f"ğŸ›¡ï¸ Verteidigung: {verteidigung}")
print(f"ğŸ’¥ Echter Schaden: {echter_schaden}")

# Multiplikation und Division
basis_xp = 100
multiplikator = 1.5
bonus_xp = basis_xp * multiplikator
print(f"\nğŸ“Š Basis XP: {basis_xp}")
print(f"âœ¨ Multiplikator: {multiplikator}")
print(f"ğŸ Bonus XP: {bonus_xp}")

# Modulo (Rest bei Division) - nÃ¼tzlich fÃ¼r Zyklen!
tag = 15
wochentag = tag % 7
print(f"\nğŸ“… Tag: {tag}")
print(f"ğŸ—“ï¸ Wochentag (0-6): {wochentag}")

print(f"\nğŸ‰ +200 XP: Du beherrschst mathematische Operationen!")
````

### ğŸ“ String-Magie: Text kombinieren

Mit Strings (Text) kannst du auch rechnen â€“ aber anders!

- `+` verbindet Strings (Konkatenation)
- `*` wiederholt Strings
- `len()` gibt die LÃ¤nge zurÃ¼ck

````python
# ğŸ® Level 4: String-Operationen

vorname = "Frodo"
nachname = "Beutlin"
voller_name = vorname + " " + nachname
print(f"ğŸ‘¤ Voller Name: {voller_name}")

# Strings wiederholen
kampfruf = "FÃ¼r Gondor! " * 3
print(f"âš”ï¸ Kampfruf: {kampfruf}")

# String-LÃ¤nge
zauberspruch = "Abracadabra"
laenge = len(zauberspruch)
print(f"\nâœ¨ Zauberspruch: {zauberspruch}")
print(f"ğŸ“ LÃ¤nge: {laenge} Zeichen")

print(f"\nğŸ‰ +250 XP: Du kannst mit Strings jonglieren!")
````

### ğŸ”„ Die Kunst der Typumwandlung

Manchmal musst du einen Typ in einen anderen verwandeln:

- `int()` â†’ Wandelt in Integer um
- `float()` â†’ Wandelt in Float um
- `str()` â†’ Wandelt in String um
- `bool()` â†’ Wandelt in Boolean um

**Wichtig:** Nicht alle Umwandlungen funktionieren! `int("Hallo")` gibt einen Fehler.

````python
# ğŸ® Level 5: Typumwandlung meistern

# String zu Integer
level_text = "42"
level_zahl = int(level_text)
print(f"ğŸ“ Als Text: {level_text}")
print(f"ğŸ”¢ Als Zahl: {level_zahl}")
print(f"â• Plus 1: {level_zahl + 1}")

# Integer zu String (fÃ¼r Ausgaben)
gold = 500
nachricht = "Du hast " + str(gold) + " Gold!"
print(f"\nğŸ’° {nachricht}")

# Float zu Integer (Rundung)
leben = 87.9
leben_gerundet = int(leben)
print(f"\nâ¤ï¸ Leben (genau): {leben}")
print(f"â¤ï¸ Leben (gerundet): {leben_gerundet}")

# String zu Float
schaden_text = "25.5"
schaden_zahl = float(schaden_text)
print(f"\nâš”ï¸ Schaden: {schaden_zahl}")

print(f"\nğŸ‰ +200 XP: Du beherrschst Typumwandlungen!")
````

### ğŸ§™ Magische Variablen-Tricks

Fortgeschrittene Techniken fÃ¼r echte Magier:

````python
# ğŸ® Level 6: Erweiterte Variablen-Techniken

# Mehrere Variablen gleichzeitig zuweisen
x, y, z = 10, 20, 30
print(f"ğŸ“ Koordinaten: x={x}, y={y}, z={z}")

# Variablen tauschen (ohne Hilfsvariable!)
a = 5
b = 10
print(f"\nğŸ”„ Vorher: a={a}, b={b}")
a, b = b, a
print(f"ğŸ”„ Nachher: a={a}, b={b}")

# Verkettete Zuweisungen
leben = mana = energie = 100
print(f"\nâš¡ Leben: {leben}, Mana: {mana}, Energie: {energie}")

# Augmented Assignment (Kurzschreibweise)
gold = 100
print(f"\nğŸ’° Start-Gold: {gold}")
gold += 50  # Gleich wie: gold = gold + 50
print(f"ğŸ’° Nach Quest: {gold}")
gold *= 2   # Gleich wie: gold = gold * 2
print(f"ğŸ’° Nach Bonus: {gold}")

print(f"\nğŸ‰ +300 XP: Du bist ein Variablen-Meister!")
````
