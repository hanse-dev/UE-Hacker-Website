````python
# ğŸ§® Level 3: Geschwindigkeiten berechnen

import math

# Berechne die Durchschnittsgeschwindigkeit
distanz_meter = 1200
zeit_sekunden = 180
geschwindigkeit_ms = distanz_meter / zeit_sekunden
geschwindigkeit_kmh = geschwindigkeit_ms * 3.6

print(f"ğŸ‡ Gerittene Distanz: {distanz_meter} Meter")
print(f"â±ï¸ Zeit: {zeit_sekunden} Sekunden ({zeit_sekunden/60:.1f} Minuten)")
print(f"ğŸ“Š Geschwindigkeit: {geschwindigkeit_ms:.2f} m/s")
print(f"ğŸš€ Das entspricht: {geschwindigkeit_kmh:.2f} km/h")

# Runde Werte auf/ab
punkte = 87.6
punkte_ab = math.floor(punkte)
punkte_auf = math.ceil(punkte)

print(f"\nğŸ† Turnierpunkte: {punkte}")
print(f"â¬‡ï¸ Abgerundet: {punkte_ab}")
print(f"â¬†ï¸ Aufgerundet: {punkte_auf}")
print()
print(f"ğŸ‰ +200 Huf-Punkte: Berechnungen gemeistert!")
````

### ğŸ“ Power-Move: math-Konstanten

Das math-Modul enthÃ¤lt wichtige mathematische Konstanten:

- **`math.pi`** â€“ Die Kreiszahl Ï€ (3.14159...)
- **`math.e`** â€“ Eulersche Zahl e (2.71828...)
- **`math.inf`** â€“ Unendlich
- **`math.nan`** â€“ "Not a Number" (ungÃ¼ltige Zahl)

````python
# ğŸ¯ Level 4: Parcours-Generator

import random

# Generiere einen zufÃ¤lligen Parcours
anzahl_hindernisse = random.randint(8, 12)
print(f"ğŸ‡ Parcours mit {anzahl_hindernisse} Hindernissen")
print(f"="*40)

hindernistypen = ["Oxer", "Steilsprung", "Wassergraben", "Doppel", "Mauer"]
gesamthoehe = 0

for i in range(anzahl_hindernisse):
    typ = random.choice(hindernistypen)
    hoehe = random.uniform(0.8, 1.4)  # HÃ¶he in Metern
    gesamthoehe += hoehe
    
    print(f"Hindernis {i+1}: {typ} ({hoehe:.2f}m)")

durchschnittshoehe = gesamthoehe / anzahl_hindernisse
print(f"\nğŸ“Š DurchschnittshÃ¶he: {durchschnittshoehe:.2f}m")

# Schwierigkeitsgrad bestimmen
if durchschnittshoehe < 1.0:
    print(f"â­ Schwierigkeit: Leicht")
elif durchschnittshoehe < 1.2:
    print(f"â­â­ Schwierigkeit: Mittel")
else:
    print(f"â­â­â­ Schwierigkeit: Schwer")
print()
print(f"ğŸ‰ +250 Huf-Punkte: Parcours erstellt!")
````

````python
# ğŸŒŸ Level 5: Winkelberechnungen fÃ¼r SprÃ¼nge

import math

# Berechne den optimalen Absprungwinkel
hindernis_hoehe = 1.2  # Meter
absprung_distanz = 2.5  # Meter vor dem Hindernis

# Berechne den Winkel (Tangens)
winkel_rad = math.atan(hindernis_hoehe / absprung_distanz)
winkel_grad = math.degrees(winkel_rad)

print(f"ğŸ‡ Hindernis-HÃ¶he: {hindernis_hoehe}m")
print(f"ğŸ“ Absprung-Distanz: {absprung_distanz}m")
print(f"ğŸ“ Optimaler Winkel: {winkel_grad:.1f}Â°")

# Berechne die Flugbahn-LÃ¤nge
flugbahn = math.sqrt(absprung_distanz**2 + hindernis_hoehe**2)
print(f"âœˆï¸ Flugbahn-LÃ¤nge: {flugbahn:.2f}m")

# Berechne benÃ¶tigte Geschwindigkeit
geschwindigkeit_ms = math.sqrt(2 * 9.81 * hindernis_hoehe)
geschwindigkeit_kmh = geschwindigkeit_ms * 3.6

print(f"ğŸš€ Mindest-Geschwindigkeit: {geschwindigkeit_kmh:.1f} km/h")
print()
print(f"ğŸ‰ +200 Huf-Punkte: Physik des Springreitens verstanden!")
````

### ğŸ“Š Power-Move: Grad vs. Radiant

Trigonometrische Funktionen in Python arbeiten mit **Radiant**, nicht mit Grad!

- **`math.radians(grad)`** â€“ Konvertiert Grad zu Radiant
- **`math.degrees(rad)`** â€“ Konvertiert Radiant zu Grad
- 360Â° = 2Ï€ Radiant
- 180Â° = Ï€ Radiant

````python
# ğŸ° Level 6: Fortgeschrittene Trainingsplanung

import random

# Erstelle einen Wochen-Trainingsplan
tage = ["Montag", "Dienstag", "Mittwoch", "Donnerstag", "Freitag", "Samstag", "Sonntag"]
aktivitaeten = ["Dressur", "Springen", "Ausritt", "Bodenarbeit", "Longieren", "Ruhetag"]
gewichte = [20, 20, 15, 15, 15, 15]  # Wahrscheinlichkeiten in %

print(f"ğŸ“… Trainingsplan fÃ¼r diese Woche:")
print(f"="*40)

for tag in tage:
    aktivitaet = random.choices(aktivitaeten, weights=gewichte, k=1)[0]
    
    if aktivitaet == "Ruhetag":
        print(f"{tag}: ğŸ˜´ {aktivitaet}")
    elif aktivitaet == "Ausritt":
        dauer = random.randint(60, 120)
        print(f"{tag}: ğŸŒ² {aktivitaet} ({dauer} Min.)")
    else:
        dauer = random.randint(30, 60)
        print(f"{tag}: ğŸ´ {aktivitaet} ({dauer} Min.)")
print()
print(f"ğŸ‰ +300 Huf-Punkte: Perfekter Trainingsplan!")
````

### ğŸ¯ Skill-ErklÃ¤rung: random.choices() vs random.choice()

- **`random.choice(liste)`** â€“ WÃ¤hlt EIN Element zufÃ¤llig
- **`random.choices(liste, weights, k)`** â€“ WÃ¤hlt k Elemente mit Gewichtung (mit ZurÃ¼cklegen)
- **`random.sample(liste, k)`** â€“ WÃ¤hlt k Elemente ohne ZurÃ¼cklegen (keine Duplikate)

````python
# ğŸ”¥ Level 7: Kombination von math und random

import math
import random

# Simuliere einen Springparcours mit Zeitfehler
print(f"ğŸ‡ SPRINGPARCOURS-SIMULATION")
print(f"="*40)

sollzeit = 60.0  # Sekunden
hindernisse = 10
fehler = 0
zeitfehler = 0

# Simuliere jeden Sprung
for i in range(hindernisse):
    # ZufÃ¤llige Erfolgsrate
    erfolg = random.random()
    
    if erfolg < 0.05:  # 5% Chance fÃ¼r Abwurf
        print(f"Hindernis {i+1}: âŒ Abwurf! (+4 Strafpunkte)")
        fehler += 4
    elif erfolg < 0.10:  # 5% Chance fÃ¼r Verweigerung
        print(f"Hindernis {i+1}: ğŸ›‘ Verweigerung! (+4 Strafpunkte)")
        fehler += 4
    else:
        print(f"Hindernis {i+1}: âœ… Sauber!")

# Berechne die tatsÃ¤chliche Zeit mit Zufallsabweichung
zeit_abweichung = random.uniform(-3, 5)
tatsaechliche_zeit = sollzeit + zeit_abweichung

# Berechne Zeitfehler
if tatsaechliche_zeit > sollzeit:
    zeitfehler = math.ceil(tatsaechliche_zeit - sollzeit)

print(f"\nâ±ï¸ Sollzeit: {sollzeit:.2f}s")
print(f"â±ï¸ TatsÃ¤chliche Zeit: {tatsaechliche_zeit:.2f}s")
print(f"ğŸ“Š Hindernisfehler: {fehler} Punkte")
print(f"ğŸ“Š Zeitfehler: {zeitfehler} Punkte")
print(f"ğŸ† Gesamtfehler: {fehler + zeitfehler} Punkte")
print()
print(f"ğŸ‰ +250 Huf-Punkte: Realistische Simulation erstellt!")
````

### ğŸŒŸ Wichtiges Konzept: Strafpunkte im Springreiten

Im Springreiten gibt es verschiedene Fehler:
- **Abwurf:** 4 Strafpunkte
- **Verweigerung:** 4 Strafpunkte (beim 3. Mal Ausschluss)
- **ZeitÃ¼berschreitung:** 1 Punkt pro angefangene Sekunde
- **Sturz:** Ausschluss

````python
# ğŸ† Level 8: Fortgeschrittene Anwendung â€“ Turnier-Rangliste

import math
import random

print(f"ğŸ† TURNIER-RANGLISTE GENERATOR")
print(f"="*40)

# Erstelle Teilnehmer
teilnehmer = [
    "Luna & Stella",
    "Max & Thunder",
    "Emma & Moonlight",
    "Felix & Shadow",
    "Sophie & Blitz"
]

ergebnisse = []

for reiter in teilnehmer:
    # Simuliere den Ritt
    fehler = 0
    
    # ZufÃ¤llige Anzahl AbwÃ¼rfe (0-2)
    abwuerfe = 0
    for _ in range(10):  # 10 Hindernisse
        if random.random() < 0.08:  # 8% Abwurf-Chance
            abwuerfe += 1
    
    fehler += abwuerfe * 4
    
    # ZufÃ¤llige Zeit
    sollzeit = 65.0
    zeit = sollzeit + random.uniform(-2, 8)
    
    # Zeitfehler berechnen
    if zeit > sollzeit:
        zeitfehler = math.ceil(zeit - sollzeit)
        fehler += zeitfehler
    else:
        zeitfehler = 0
    
    ergebnisse.append({
        "name": reiter,
        "fehler": fehler,
        "zeit": zeit,
        "abwuerfe": abwuerfe,
        "zeitfehler": zeitfehler
    })

# Sortiere nach Fehlern (weniger ist besser), dann nach Zeit
ergebnisse.sort(key=lambda x: (x["fehler"], x["zeit"]))

# Zeige Rangliste
print(f"\nğŸ… ENDERGEBNIS:")
print(f"="*40)

for i, ergebnis in enumerate(ergebnisse, 1):
    if i == 1:
        medaille = "ğŸ¥‡"
    elif i == 2:
        medaille = "ğŸ¥ˆ"
    elif i == 3:
        medaille = "ğŸ¥‰"
    else:
        medaille = f"{i}."
    
    print(f"{medaille} {ergebnis['name']}")
    print(f"   Fehler: {ergebnis['fehler']} ({ergebnis['abwuerfe']} AbwÃ¼rfe, {ergebnis['zeitfehler']}s Zeit)")
    print(f"   Zeit: {ergebnis['zeit']:.2f}s")
    print()

print(f"ğŸ‰ +400 Huf-Punkte: Turnier-System gemeistert!")
````
