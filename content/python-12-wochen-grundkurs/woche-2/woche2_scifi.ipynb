## üìö Systemprotokoll 1: Der f-String-Befehl

**Was es ist:** Der f-String ist die moderne Methode, Text und Variablen zu pr√§zisen Befehlen zu fusionieren.

**Warum n√ºtzlich:**
- ‚ú® Verbindet Befehle m√ºhelos mit Systemdaten

## üì¶ Systemprotokoll 2: Die vier Quanten-Typen

**Was es ist:** Jede Information auf der Nebula-7 hat einen Quanten-Typ - wie die vier Aggregatzust√§nde der Materie.

**Warum n√ºtzlich:**
- üî§ **String (str):** F√ºr Bezeichnungen, Codes und Nachrichten
- üî¢ **Integer (int):** F√ºr ganze Zahlen, IDs und Z√§hler
- üíß **Float (float):** F√ºr pr√§zise Messwerte und Koordinaten
- ‚ö° **Boolean (bool):** f√ºr Systemzust√§nde und Entscheidungen

**Wichtig:** `type()` zeigt dir den Quanten-Typ jeder Information!

````python
# Beispiel 1: Strings - Textdaten
schiff_name = "Starlight"
klasse = "Kreuzer"
print(f"Schiff: {schiff_name}")
print(f"Klasse: {klasse}")
print(f"Typ: {type(schiff_name)}")

# Beispiel 2: Integers - Ganze Zahlen
crew_anzahl = 250
deck_anzahl = 15
print(f"Crew: {crew_anzahl}, Decks: {deck_anzahl}")
print(f"Typ: {type(crew_anzahl)}")

# Beispiel 3: Floats - Pr√§zise Werte
geschwindigkeit = 2.5
koordinate_x = 127.35
print(f"Geschwindigkeit: {geschwindigkeit} Lichtjahre/h")
print(f"Position X: {koordinate_x}")
print(f"Typ: {type(geschwindigkeit)}")

# Beispiel 4: Booleans - Systemzust√§nde
system_online = True
alarm_aktiv = False
print(f"System online: {system_online}")
print(f"Alarm aktiv: {alarm_aktiv}")
print(f"Typ: {type(system_online)}")
````

## ‚ú® Systemprotokoll 3: Daten-Operationen und Transformationen

**Was es ist:** Wie du Daten kombinierst, ver√§nderst und transformierst.

**Warum n√ºtzlich:**
- üî¨ Wissenschaftliche Berechnungen
- üî§ Textoperationen f√ºr Protokolle
- üîÑ Datentransformation f√ºr Flexibilit√§t
- üìä Sensordaten verarbeiten und nutzen

**Wichtig:** Nicht alle Datentypen k√∂nnen direkt kombiniert werden!

````python
# Beispiel 1: Mathematische Operationen
planeten_a = 8
planeten_b = 4
print(f"Gesamtplaneten: {planeten_a + planeten_b}")
print(f"Differenz: {planeten_a - planeten_b}")
print(f"Verdopplung: {planeten_a * 2}")
print(f"Pro Sektor: {planeten_a / 2}")

# Beispiel 2: String-Operationen
prefix = "Neo"
suffix = "tron"
system_name = prefix + suffix
print(f"Systemname: {system_name}")
print(f"Signal: {prefix * 3}")

# Beispiel 3: Datentransformation
temperatur_str = "273"
temperatur_int = int(temperatur_str)
print(f"Temperatur als Text: {temperatur_str}K (Typ: {type(temperatur_str)})")
print(f"Temperatur als Zahl: {temperatur_int}K (Typ: {type(temperatur_int)})")

# Beispiel 4: Sensordaten verarbeiten
# Sensor-Eingabe simuliert (sp√§ter mit input())
sensor_wert = "42.5"
wert_float = float(sensor_wert)
print(f"Sensorwert {wert_float} verdoppelt: {wert_float * 2}")
````
