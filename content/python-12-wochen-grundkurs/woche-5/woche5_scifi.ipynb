### ğŸ” Skill: Funktionen-Arbeitsmittel (Parameter)
Manchmal braucht eine Funktion **Arbeitsmittel**, um zu funktionieren! Eine FÃ¼tterungs-Funktion braucht vielleicht den Raumschiffnamen, eine Pflege-Funktion braucht die Art der Pflege.
Diese "Arbeitsmittel" nennen wir **Parameter**:
- ğŸ“ **Definition**: `def begruesse(name):` â€“ `name` ist ein Parameter (Platzhalter)
- âš¡ **Aufruf**: `begruesse("Enterprise")` â€“ `"Enterprise"` ist das Argument (das echte Arbeitsmittel)
**Merke:** Parameter = Platzhalter im Rezept, Argument = echtes Arbeitsmittel bei der AusfÃ¼hrung!

````python
# ğŸš€ Level 3: Funktionen mit Arbeitsmitteln!
# Definition mit Parameter 'name' (Platzhalter fÃ¼r ein Arbeitsmittel)
def persoenliche_begruessung(name):
    print(f"ğŸŒŸ Willkommen auf der Raumfahrt-Station, {name}!")
    print(f"âœ¨ MÃ¶ge deine Zeit mit den Raumschiffen erfolgreich sein, {name}!")
# Aufrufe mit verschiedenen Argumenten (echte Arbeitsmittel)
print("ğŸ§™ Die Missionsleiterin begrÃ¼ÃŸt die neuen Raumfahrt-Kadetten:\n")
persoenliche_begruessung("Alex")
print()
persoenliche_begruessung("Maria")
print("\nğŸ‰ +150 Cyber Credits: Du kannst jetzt Funktionen mit Arbeitsmitteln erstellen!")
````

### âš¡ Power-Moves: `return` â€“ Praktische Ergebnisse zurÃ¼ckbekommen
Stell dir vor, du fragst eine Raumfahrt-Expertin: "Wie viel Treibstoff braucht ein Raumschiff?" 
Du willst nicht, dass sie das Ergebnis nur **ansagt** (`print`), sondern dass sie es **dir gibt** (`return`), damit du damit weiterarbeiten kannst!
**Das macht `return`:**
- ğŸ Gibt ein Ergebnis an die Stelle zurÃ¼ck, wo die Funktion durchgefÃ¼hrt wurde
- ğŸ›‘ Beendet die Funktion sofort
- ğŸ’¾ ErmÃ¶glicht es, das Ergebnis zu speichern und weiterzuverwenden
**Wann brauchst du `return`?**
- âœ… Wenn du das Ergebnis speichern willst
- âœ… Wenn du damit weiterrechnen willst
- âœ… Wenn du es an eine andere Funktion Ã¼bergeben willst
**Wann reicht `print`?**
- âœ… Wenn du nur etwas anzeigen willst (wie eine Nachricht)

````python
# ğŸš€ Level 4: Funktionen mit praktischen Ergebnissen!
# Diese Funktion GIBT ein Ergebnis zurÃ¼ck (wie ein Geschenk!)
def addiere(a, b):
    ergebnis = a + b
    return ergebnis  # ğŸ Das Ergebnis wird zurÃ¼ckgegeben
# Das Ergebnis wird in 'summe' gespeichert
print("ğŸ§® Die Treibstoffberechnung fÃ¼r die Raumschiffe wird durchgefÃ¼hrt...")
summe = addiere(10, 5)
print(f"ğŸ’ Das praktische Ergebnis: {summe}")
print(f"âœ¨ Wir kÃ¶nnen damit weiterarbeiten: {summe * 2}")
print(f"ğŸ”® Oder es in neuen Berechnungen nutzen: {summe + 100}")
print("\nğŸ‰ +200 Cyber Credits: Du beherrschst jetzt return!")
````
