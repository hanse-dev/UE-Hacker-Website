### âš¡ Power-Moves: Stalltafel verÃ¤ndern

Deine Stalltafel ist **lebendig** â€“ du kannst sie jederzeit anpassen! Das nennt man in der Programmierer:innen-Sprache "mutable" (verÃ¤nderbar).

#### ğŸ“¥ Neue Pferde aufnehmen
- `stalltafel.append(eintrag)` â†’ FÃ¼gt den Eintrag in die nÃ¤chste freie Box
- `stalltafel.insert(box, eintrag)` â†’ Setzt den Eintrag an eine bestimmte Position

````python
# ğŸ´ Level 3: Neue Pferde aufnehmen!

# Du startest deine Stall-TÃ¤tigkeit mit Grundausstattung
ausruestung = ["Sattel", "Zaumzeug"]
print(f"âš”ï¸ Start-AusrÃ¼stung: {ausruestung}")

# Du kaufst eine neue Pferdedecke!
print("\nğŸ’¥ Neue AusrÃ¼stung gekauft!")
ausruestung.append("Pferdedecke")
print(f"ğŸ“¦ Neue AusrÃ¼stung! Liste: {ausruestung}")

# Ein HÃ¤ndler verkauft dir einen Trensen - den willst du an Position 1
print("\nğŸ›’ Du kaufst eine Trense beim AusrÃ¼ster...")
ausruestung.insert(1, "Trense")
print(f"âœ¨ AusrÃ¼stung neu sortiert: {ausruestung}")

print("\nğŸ‰ +200 XP: Du kannst jetzt neue EintrÃ¤ge aufnehmen!")
````

#### ğŸ—‘ï¸ EintrÃ¤ge entfernen
- `stalltafel.remove(eintrag)` â†’ Entfernt einen bestimmten Eintrag (nur das erste Vorkommen!)
- `stalltafel.pop(box)` â†’ Nimmt Eintrag aus Box und gibt ihn dir zurÃ¼ck
- `del stalltafel[box]` â†’ LÃ¶scht Eintrag aus Box (ohne RÃ¼ckgabe)

````python
# ğŸ´ Level 4: Stalltafel-Management

# Deine Aufgaben-Liste
aufgaben = ["Heu besorgen", "Stall ausmisten", "Pferde reiten", "Futter kontrollieren", "Pferde reiten"]
print(f"ğŸ“‹ Deine Aufgaben: {aufgaben}")

# Du hast die erste Aufgabe erledigt!
print("\nâœ… Aufgabe erledigt: 'Pferde reiten'")
aufgaben.remove("Pferde reiten")  # Entfernt nur das ERSTE Vorkommen
print(f"ğŸ“‹ Verbleibende Aufgaben: {aufgaben}")

# Die letzte Aufgabe ist die wichtigste - du nimmst sie raus um sie zu prÃ¼fen
print("\nğŸ” Welche Aufgabe ist am wichtigsten?")
wichtigste_aufgabe = aufgaben.pop()  # Nimmt die letzte Aufgabe
print(f"â­ Wichtigste Aufgabe: {wichtigste_aufgabe}")
print(f"ğŸ“‹ Ãœbrige Aufgaben: {aufgaben}")

# Die erste Aufgabe ist zu einfach - weg damit!
print("\nğŸ—‘ï¸ Die erste Aufgabe ist zu einfach...")
del aufgaben[0]
print(f"ğŸ“‹ Finale Aufgaben-Liste: {aufgaben}")

print("\nğŸ‰ +250 XP: Du bist ein Stalltafel-Management-Profi!")
````
