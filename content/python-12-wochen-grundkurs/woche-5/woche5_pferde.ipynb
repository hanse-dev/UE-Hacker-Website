## ğŸ“š Trainings-Methode 1: Die Funktions-Definition - Der Trainingsplan

**Was es ist:** Eine Funktion ist ein wiederverwendbarer Trainings-Block, den du einmal definierst und immer wieder aufrufen kannst.

**Die Anatomie einer Funktion:**
```python
def trainings_name(parameter1, parameter2):
    """Beschreibung des Trainings"""
    # Trainings-Code
    return ergebnis  # Optional: Ergebnis zurÃ¼ckgeben
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`def`** - Das Kommando, das die Trainings-Definition startet
2. **`trainings_name`** - Ein eindeutiger Name fÃ¼r dein Training
3. **`(parameter)`** - Trainings-Variablen (optional)
4. **`:`** - Beendet die Trainings-Definition
5. **`"""Docstring"""`** - Trainingshandbuch (sehr wichtig!)
6. **EingerÃ¼ckter Code** - Der eigentliche Trainings-Code
7. **`return`** - Gibt das Ergebnis zurÃ¼ck (optional)

**Warum so nÃ¼tzlich:**
- ğŸ¯ Trainings wiederverwenden statt neu schreiben
- ğŸ” Training organisieren und strukturieren
- âš¡ Fehler an einer Stelle beheben
- ğŸ² Komplexe Trainings in kleine Teile zerlegen

````python
# Beispiel 1: Einfache Funktion ohne Parameter
def begruessung_reiterhof():
    """BegrÃ¼ÃŸt Besucher auf dem Reiterhof"""
    print("Willkommen auf dem Reiterhof Sonnental!")
    print("Dein Training beginnt jetzt!")

# Funktion aufrufen
print("=== Erster Besucher ===")
begruessung_reiterhof()

print("\n=== Zweiter Besucher ===")
begruessung_reiterhof()

print("\n=== Was passiert hier? ===")
print("1. def begruessung_reiterhof(): definiert die Funktion")
print("2. begruessung_reiterhof() ruft die Funktion auf")
print("3. Der Code in der Funktion wird jedes Mal ausgefÃ¼hrt")
````

````python
# Beispiel 2: Funktion mit einem Parameter
def begruesse_reiter(name):
    """BegrÃ¼ÃŸt einen Reiter mit seinem Namen"""
    print(f"Hallo {name}!")
    print(f"Viel Erfolg beim Training, {name}!")

# Funktion mit verschiedenen Namen aufrufen
print("=== Verschiedene Reiter ===")
begruesse_reiter("Anna")
begruesse_reiter("Max")
begruesse_reiter("Lena")

print("\n=== Was passiert hier? ===")
print("1. name ist ein Platzhalter fÃ¼r den Ã¼bergebenen Wert")
print("2. Jeder Aufruf Ã¼bergibt einen anderen Namen")
print("3. Die Funktion arbeitet mit dem Ã¼bergebenen Wert")
````

````python
# Beispiel 3: Funktion mit mehreren Parametern
def erstelle_pferd_profil(name, rasse, alter):
    """Erstellt ein Pferd-Profil mit Name, Rasse und Alter"""
    print(f"Pferd-Profil:")
    print(f"  Name: {name}")
    print(f"  Rasse: {rasse}")
    print(f"  Alter: {alter} Jahre")
    print(f"  Status: Bereit fÃ¼r das Training!")

# Verschiedene Pferde-Profile erstellen
print("=== Pferd-Profile ===")
erstelle_pferd_profil("Thunder", "Hannoveraner", 8)
print()
erstelle_pferd_profil("Luna", "IslÃ¤nder", 5)
print()
erstelle_pferd_profil("Storm", "Quarter Horse", 6)
````

## ğŸ” Das Geheimnis von return - Das Trainingsergebnis

**return** ist der mÃ¤chtigste Teil einer Funktion!

**Was return tut:**
- Beendet die Funktion sofort
- Gibt einen Wert zurÃ¼ck an den Aufrufer
- Dieser Wert kann in einer Variable gespeichert werden

**Ohne return:** Funktion macht etwas (Side Effect)
**Mit return:** Funktion berechnet und gibt Ergebnis zurÃ¼ck

````python
# ğŸ” Ãœbungen mit return

# Beispiel 1: Einfache Berechnung mit return
def berechne_trainingszeit(einheiten, dauer_pro_einheit):
    """Berechnet die gesamte Trainingszeit"""
    gesamtzeit = einheiten * dauer_pro_einheit
    return gesamtzeit

# Ergebnis in Variable speichern
print("=== Trainingszeit-Berechnung ===")
zeit1 = berechne_trainingszeit(5, 15)
print(f"Training 1: {zeit1} Minuten")

zeit2 = berechne_trainingszeit(3, 20)
print(f"Training 2: {zeit2} Minuten")

# Ergebnis direkt verwenden
print(f"Intensivtraining: {berechne_trainingszeit(10, 30} Minuten")
````

````python
# Beispiel 2: Funktion mit bedingtem return
def pruefe_training_level(erfahrung):
    """PrÃ¼ft, ob ein Pferd fÃ¼r fortgeschrittenes Training bereit ist"""
    if erfahrung >= 50:
        return "Bereit fÃ¼r fortgeschrittenes Training!"
    else:
        return f"Noch {50 - erfahrung} Trainingseinheiten needed"

print("=== Training-Level-PrÃ¼fung ===")
status1 = pruefe_training_level(25)
print(f"Erfahrung 25: {status1}")

status2 = pruefe_training_level(75)
print(f"Erfahrung 75: {status2}")

status3 = pruefe_training_level(40)
print(f"Erfahrung 40: {status3}")
````

````python
# Beispiel 3: Funktion ohne expliziten return (None)
def zeige_ausruestung():
    """Zeigt die Standard-AusrÃ¼stung (gibt nichts zurÃ¼ck)"""
    print("Deine AusrÃ¼stung:")
    print("  - Sattel")
    print("  - Zaumzeug")
    print("  - Gerte")
    # Kein return - gibt automatisch None zurÃ¼ck

print("=== AusrÃ¼stung anzeigen ===")
ergebnis = zeige_ausruestung()
print(f"RÃ¼ckgabewert: {ergebnis}")
print("Funktionen ohne return geben immer None zurÃ¼ck!")
````

## ğŸ“¦ Trainings-Methode 2: Docstrings - Die TrainingshandbÃ¼cher

**Was es ist:** Docstrings sind Dokumentations-Strings, die erklÃ¤ren, was eine Funktion tut.

**Warum extrem wichtig:**
- ğŸ“– Andere (und du spÃ¤ter) verstehen deine Funktion
- ğŸ” IDEs zeigen dir Hilfe beim Aufruf
- ğŸ¤– Tools kÃ¶nnen automatisch Dokumentation erstellen
- â­ Zeichen von professionellem Code

**Die drei Arten von Docstrings:**
1. **Einzeilig:** `"""Kurze Beschreibung"""`
2. **Mehrzeilig:** `"""Kurze Beschreibung\n\nDetails..."""`
3. **VollstÃ¤ndig:** `"""Beschreibung\n\nArgs:\nReturns:\n"""`

````python
# Beispiel 1: Verschiedene Docstring-Stile
def einfache_funktion():
    """Dies ist ein einzeiliger Docstring."""
    return "Einfach"

def detaillierte_funktion(pferd_name, training_level):
    """Erstellt eine Trainings-Beschreibung.
    
    Args:
        pferd_name (str): Der Name des Pferdes
        training_level (int): Das Trainings-Level des Pferdes
    
    Returns:
        str: Eine formatierte Beschreibung
    """
    return f"{pferd_name} hat Trainings-Level {training_level}."

# Docstring aufrufen (so funktioniert es!)
print("=== Docstring-Beispiele ===")
print(f"Einfacher Docstring: {einfache_funktion.__doc__}")
print()
print(f"Detaillierter Docstring: {detaillierte_funktion.__doc__}")
````

````python
# Beispiel 2: Hilfe-Funktion nutzen
print("=== Hilfe fÃ¼r Funktionen ===")
help(detaillierte_funktion)
````

## âœ¨ Trainings-Methode 3: Namenskonventionen - Die Regeln des Reitstalls

**Gute Funktionsnamen sind entscheidend!**

**Die Regeln:**
- **Kleinbuchstaben** mit Unterstrichen: `berechne_trainingszeit`
- **Beschreibend**: Was die Funktion **tut** (Verb)
- **Kurz aber klar**: `get_name()` nicht `retrieve_horse_data_from_stable()`
- **Keine Verben im Namen**: `name_get()` falsch, `get_name()` richtig

**Beispiele:**
- âœ… `berechne_trainingszeit()`
- âœ… `erstelle_pferd_profil()`
- âœ… `pruefe_training_level()`
- âŒ `BerechneTrainingszeit()` (CamelCase)
- âŒ `berechne_das_training()` (zu lang)
- âŒ `training_berechnen()` (Substantiv zuerst)

````python
# âœ¨ Beispiele fÃ¼r gute und schlechte Namen

# âœ… Gute Namen (klar und beschreibend)
def berechne_futter_menge(gewicht, aktivitaet):
    """Berechnet die tÃ¤gliche Futtermenge."""
    return gewicht * aktivitaet * 0.02

def finde_bestes_training(pferde_liste):
    """Findet das beste Training fÃ¼r eine Pferdeliste."""
    return max(pferde_liste)

def ist_pferd_verfuegbar(pferd_id, zeitpunkt):
    """PrÃ¼ft, ob ein Pferd verfÃ¼gbar ist."""
    return pferd_id not in belegte_pferde[zeitpunkt]

# âŒ Schlechte Namen (unklar oder falsche Konvention)
# def berechneFutterMenge():  # CamelCase
# def fm(g, a):  # Zu kurz, nicht beschreibend
# def feeding_calculation():  # Substantiv zuerst

print("=== Gute Funktionsnamen in Aktion ===")
futter = berechne_futter_menge(500, 1.5)
print(f"Futtermenge: {futter} kg")

trainings = ["Dressur", "Springen", "Western"]
bestes_training = finde_bestes_training(trainings)
print(f"Bestes Training: {bestes_training}")

# Beispiel-Daten fÃ¼r VerfÃ¼gbarkeitsprÃ¼fung
belegte_pferde = {"morgen": ["Thunder", "Luna"], "mittag": []}
verfuegbar = ist_pferd_verfuegbar("Storm", "morgen")
print(f"Pferd verfÃ¼gbar: {verfuegbar}")
````
