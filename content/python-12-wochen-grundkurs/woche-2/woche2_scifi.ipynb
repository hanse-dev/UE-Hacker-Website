### ğŸ›¸ Tutorial: Dein erster Speicherslot

### ğŸ” Skill: Datentypen scannen

Jeder Datenwert hat einen **Typ**. Mit dem Scanner `type()` kannst du den Typ analysieren:

- `str` â†’ String (Text)
- `int` â†’ Integer (Ganze Zahl)
- `float` â†’ Float (Kommazahl)
- `bool` â†’ Boolean (True/False)

**Warum ist das wichtig?** Verschiedene Typen haben verschiedene Eigenschaften! Du kannst Zahlen addieren, aber Text wird anders verarbeitet.

````python
# ğŸš€ Level 2: Datentypen-Scan durchfÃ¼hren

station_name = "Nebula-7"
sektor = 42
gravitation = 9.81
ist_bewohnbar = True

print("ğŸ” Datentypen-Analyse:")
print(f"Station-Name ist vom Typ: {type(station_name).__name__}")
print(f"Sektor ist vom Typ: {type(sektor).__name__}")
print(f"Gravitation ist vom Typ: {type(gravitation).__name__}")
print(f"Ist bewohnbar ist vom Typ: {type(ist_bewohnbar).__name__}")

print(f"\nğŸ‰ +150 Cyber Credits: Du kannst Datentypen scannen!")
````

### âš¡ Power-Moves: Berechnungen durchfÃ¼hren

Variablen mit Zahlen kÃ¶nnen wie normale Zahlen verwendet werden:

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `%` Modulo (Rest bei Division)
- `**` Potenz

````python
# ğŸš€ Level 3: Treibstoff-Berechnungen

# Grundrechenarten
treibstoff_start = 1000  # Liter
treibstoff_verbraucht = 250
treibstoff_aktuell = treibstoff_start - treibstoff_verbraucht
print(f"â›½ Start-Treibstoff: {treibstoff_start} L")
print(f"ğŸ”¥ Verbraucht: {treibstoff_verbraucht} L")
print(f"ğŸ“Š Aktuell: {treibstoff_aktuell} L")

# Multiplikation und Division
energie_pro_stunde = 50  # kWh
stunden = 24
tagesverbrauch = energie_pro_stunde * stunden
print(f"\nâš¡ Energie/Stunde: {energie_pro_stunde} kWh")
print(f"â° Stunden: {stunden}")
print(f"ğŸ“ˆ Tagesverbrauch: {tagesverbrauch} kWh")

# Modulo - fÃ¼r Zyklen!
mission_tag = 47
wochentag = mission_tag % 7
print(f"\nğŸ“… Missions-Tag: {mission_tag}")
print(f"ğŸ—“ï¸ Wochentag (0-6): {wochentag}")

print(f"\nğŸ‰ +200 Cyber Credits: Du beherrschst Berechnungen!")
````

### ğŸ“ String-Operationen: Nachrichten erstellen

Mit Strings (Text) kannst du auch rechnen â€“ aber anders!

- `+` verbindet Strings (Konkatenation)

### ğŸ”„ Die Kunst der Daten-Konvertierung

Manchmal musst du einen Typ in einen anderen konvertieren:

- `int()` â†’ Wandelt in Integer um
- `float()` â†’ Wandelt in Float um
- `str()` â†’ Wandelt in String um
- `bool()` â†’ Wandelt in Boolean um

**Wichtig:** Nicht alle Konvertierungen funktionieren! `int("Nebula")` gibt einen Fehler.

### ğŸŒŸ Erweiterte Speicher-Techniken

Fortgeschrittene Techniken fÃ¼r erfahrene Kommandanten:

````python
# ğŸš€ Level 6: Profi-Techniken

# Mehrere Variablen gleichzeitig zuweisen
x, y, z = 100, 200, 300
print(f"ğŸ“ Koordinaten: x={x}, y={y}, z={z}")

# Variablen tauschen
station_a = "Alpha"
station_b = "Beta"
print(f"\nğŸ”„ Vorher: A={station_a}, B={station_b}")
station_a, station_b = station_b, station_a
print(f"ğŸ”„ Nachher: A={station_a}, B={station_b}")

# Verkettete Zuweisungen
energie = schilde = waffen = 100
print(f"\nâš¡ Energie: {energie}%, Schilde: {schilde}%, Waffen: {waffen}%")

# Augmented Assignment (Kurzschreibweise)
treibstoff = 1000
print(f"\nâ›½ Start-Treibstoff: {treibstoff} L")
treibstoff -= 100  # Gleich wie: treibstoff = treibstoff - 100
print(f"â›½ Nach Flug: {treibstoff} L")
treibstoff += 500  # Aufgetankt!
print(f"â›½ Nach Betankung: {treibstoff} L")

print(f"\nğŸ‰ +300 Cyber Credits: Du bist ein Speicher-Meister!")
````
