## ğŸ“š Magie-Sammlung 1: Der Zauber-Import - Das erste Modul

**Was es ist:** Ein Modul ist eine Sammlung von vordefinierten Zaubern, die du in dein Programm laden kannst.

**Die Anatomie eines Imports:**
```python
# Ganzes Modul importieren
import math

# Mit Alias importieren
import math as m

# Nur bestimmte Zauber importieren
from math import pi, sqrt
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`import`** - Zauberwort zum Laden
2. **`math`** - Name der Bibliothek
3. **`as`** - Optionaler Spitzname
4. **`from`** - Importiert spezifische Zauber

**Warum so nÃ¼tzlich:**
- ğŸ¯ Zauber nicht neu erfinden mÃ¼ssen
- ğŸ” Professionelle, optimierte SprÃ¼che
- âš¡ Dein Code bleibt Ã¼bersichtlich
- ğŸ² Zugriff auf spezialisierte Magie

````python
# Beispiel 1: Module importieren
# Ganzes Modul importieren
import math
print(f"Pi aus math: {math.pi}")

# Mit Alias importieren
import random as r
print(f"Zufallszahl: {r.randint(1, 10)}")

# Nur bestimmte Zauber importieren
from math import sqrt, sin, cos
print(f"Wurzel aus 16: {sqrt(16)}")
print(f"Sinus 90Â°: {sin(1.57)}")
print(f"Kosinus 0Â°: {cos(0)}")
````

````python
# Beispiel 2: Verschiedene Import-Methoden vergleichen
# Methode 1: import math
import math
print("Methode 1 - import math:")
print(f"  math.pi: {math.pi}")
print(f"  math.sqrt(25): {math.sqrt(25)}")

# Methode 2: import math as m
import math as m
print("\nMethode 2 - import math as m:")
print(f"  m.pi: {m.pi}")
print(f"  m.sqrt(25): {m.sqrt(25)}")

# Methode 3: from math import pi, sqrt
from math import pi, sqrt
print("\nMethode 3 - from math import pi, sqrt:")
print(f"  pi: {pi}")
print(f"  sqrt(25): {sqrt(25)}")
````

````python
# Beispiel 3: Mehrere Module importieren
import math
import random
import time

# Berechne eine zufÃ¤llige Flugdauer
distanz = 1000  # Meilen
geschwindigkeit = random.uniform(0.1, 0.9)  # Bruchteil der Lichtgeschwindigkeit
flugzeit = distanz / geschwindigkeit

print("=== Magische Reise-Simulation ===")
print(f"Distanz: {distanz} Meilen")
print(f"Geschwindigkeit: {geschwindigkeit:.2f}c")
print(f"Reisezeit: {flugzeit:.1f} Stunden")
print(f"Warte 2 Sekunden...")
time.sleep(2)
print(f"Ankunft am Ziel!")
````

## ğŸ“¦ Magie-Sammlung 2: Das math-Modul - Der Berechnungs-Zauber

````python
# ğŸ” Ãœbungen mit math-Konstanten
import math

print("=== Magische Konstanten ===")
print(f"Pi (Ï€): {math.pi}")
print(f"Eulersche Zahl (e): {math.e}")
print(f"Unendlich: {math.inf}")
print(f"Tau (2Ï€): {math.tau}")

# Praktische Anwendung: Kreisberechnungen
radius = 10
umfang = 2 * math.pi * radius
flaeche = math.pi * radius ** 2
print(f"\nKreis mit r={radius}:")
print(f"  Umfang: {umfang:.2f}")
print(f"  FlÃ¤che: {flaeche:.2f}")
````

## âœ¨ Magie-Sammlung 3: Das random-Modul - Der Zufalls-Zauber

**Die mÃ¤chtigsten Zufalls-Zauber:**

**Zahlen erzeugen:**
- `random.randint(a, b)` - ZufÃ¤llige Ganzzahl von a bis b
- `random.uniform(a, b)` - ZufÃ¤llige Kommazahl von a bis b
- `random.random()` - Zufallszahl zwischen 0 und 1

**Aus Listen wÃ¤hlen:**
- `random.choice(liste)` - Ein zufÃ¤lliges Element
- `random.choices(liste, k=n)` - n zufÃ¤llige Elemente
- `random.shuffle(liste)` - Liste mischen

**Verteilungen:**
- `random.gauss(mu, sigma)` - Normalverteilung
- `random.expovariate(lambda)` - Exponentialverteilung

````python
# Beispiel 1: Zufallszahlen erzeugen
import random

print("=== Zufallszahlen ===")

# Ganzzahlen
print(f"ZufÃ¤llige Zahl 1-6: {random.randint(1, 6)}")
print(f"ZufÃ¤llige Zahl 0-100: {random.randint(0, 100)}")

# Kommazahlen
print(f"Zufallszahl 0-1: {random.random():.3f}")
print(f"Zufallszahl 10-20: {random.uniform(10, 20):.2f}")

# Mit Seeds reproduzierbar
random.seed(42)
print(f"Mit Seed 42: {random.randint(1, 100)}")
random.seed(42)
print(f"Mit Seed 42: {random.randint(1, 100)} (gleiches Ergebnis!)")
````

````python
# Beispiel 2: Aus Listen wÃ¤hlen
import random

print("=== Zufallsauswahl aus Listen ===")

# Reiche im KÃ¶nigreich
reiche = ["Pyralia", "Aquaria", "Terrania", "Aeris", "Ignis", "Natura"]
print(f"Alle Reiche: {reiche}")

# Ein zufÃ¤lliges Reich
ziel = random.choice(reiche)
print(f"Zielreich: {ziel}")

# Drei zufÃ¤llige Reiche
flotte = random.choices(reiche, k=3)
print(f"Reiseziele: {flotte}")

# Liste mischen
reihenfolge = reiche.copy()
random.shuffle(reihenfolge)
print(f"Gemischte Reihenfolge: {reihenfolge}")
````

````python
# Beispiel 3: Fortgeschrittene Zufallszauber
import random
import math

print("=== Fortgeschrittene Zufallszauber ===")

# Normalverteilung fÃ¼r Energiewerte
print("Magische Energie (Normalverteilung):")
for i in range(5):
    energie = random.gauss(100, 20)  # Mittelwert 100, Stdabw 20
    print(f"  Magier {i+1}: {max(0, energie):.1f} Mana")

# Exponentialverteilung fÃ¼r Wartezeiten
print("\nWartezeiten (Exponentialverteilung):")
for i in range(5):
    warten = random.expovariate(0.1)  # Lambda = 0.1
    print(f"  Wartezeit {i+1}: {warten:.1f} Minuten")

# ZufÃ¤lliger Punkt im Kreis
def zufalls_punkt_kreis(radius):
    winkel = random.uniform(0, 2 * math.pi)
    r = radius * math.sqrt(random.random())
    x = r * math.cos(winkel)
    y = r * math.sin(winkel)
    return x, y

x, y = zufalls_punkt_kreis(10)
print(f"\nZufÃ¤lliger Punkt im Kreis r=10: ({x:.2f}, {y:.2f})")
````
