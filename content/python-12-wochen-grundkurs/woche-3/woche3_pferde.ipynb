### ğŸ¯ Skill-ErklÃ¤rung: and vs or

**and** = Beide mÃ¼ssen wahr sein:
- `True and True` â†’ True
- `True and False` â†’ False
- `False and False` â†’ False

**or** = Mindestens eines muss wahr sein:
- `True or True` â†’ True
- `True or False` â†’ True
- `False or False` â†’ False

**EselsbrÃ¼cke:** 
- **and** = streng (BEIDE Bedingungen)
- **or** = groÃŸzÃ¼gig (EINE Bedingung reicht)

````python
# ğŸ° Level 7: not-Operator

# Der not-Operator kehrt eine Bedingung um
pferd_name = "Shadow"
pferd_krank = False
pferd_verletzt = False
pferd_mÃ¼de = False

print(f"ğŸ´ Status von {pferd_name}:")
print(f"ğŸ¤’ Krank: {pferd_krank}")
print(f"ğŸ¤• Verletzt: {pferd_verletzt}")
print(f"ğŸ˜´ MÃ¼de: {pferd_mÃ¼de}")
print()

# not kehrt True zu False um (und umgekehrt)
if not pferd_krank:
    print("âœ… Nicht krank!")

if not pferd_verletzt:
    print("âœ… Nicht verletzt!")

if not pferd_mÃ¼de:
    print("âœ… Nicht mÃ¼de!")

# Kombiniere not mit and
if not pferd_krank and not pferd_verletzt and not pferd_mÃ¼de:
    print()
    print(f"ğŸ’ª {pferd_name} ist in Topform!")
    print("ğŸ‡ Perfekt fÃ¼r einen langen Ausritt!")

print()

# Noch ein Beispiel
box_sauber = False

if not box_sauber:
    print("ğŸ§¹ Die Box muss noch gemistet werden.")
    print("ğŸ’ª Zeit fÃ¼r die Stallarbeit!")

print()
print("ğŸ‰ +250 Huf-Punkte: not-Operator verstanden!")
````
