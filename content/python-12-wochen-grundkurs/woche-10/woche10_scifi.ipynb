## 🛸 Systemprotokoll 3: Der Konstruktor (__init__)

**Was es ist:** Der Konstruktor ist eine spezielle Methode, die automatisch ausgeführt wird, wenn ein Objekt initialisiert wird.

**Warum nützlich:**
- Objekte werden direkt mit richtigen Werten erstellt
- Du kann Startparameter übergeben
- Initialisierungslogik wird zentralisiert

**Wichtig:** __init__ hat immer `self` als ersten Parameter! (Warum `self` nötig ist, erfährst du in der nächsten Sektion.)

````python
# Beispiel 1: Einfacher Konstruktor
class Android:
    def __init__(self, name):
        self.name = name

# Beispiel 2: Konstruktor mit mehreren Werten
class Planet:
    def __init__(self, name, typ, durchmesser):
        self.name = name
        self.typ = typ
        self.durchmesser = durchmesser

# Beispiel 3: Konstruktor mit Standardwerten
class Alien:
    def __init__(self, species="Greys", hp=50):
        self.species = species
        self.hp = hp
````

## 🪞 Warum `self`? – Der Schlüssel zu deinen Objekten

**Was ist `self`?** `self` ist die Referenz auf **das konkrete Objekt**, das gerade angesprochen wird. Wenn du `enterprise.starten()` aufrufst, ist `self` in der Methode genau das Objekt `enterprise`.

**Warum brauchen wir es überhaupt?** Eine Klasse kann viele Objekte erzeugen. Python muss wissen: *Welches* dieser Objekte ist gemeint? Bei 10 Raumschiffen muss die Methode wissen: „Ich arbeite für DIESES Schiff hier.“

**Im Konstruktor `__init__`:**
- `self.name = name` bedeutet: **Speichere** den übergebenen `name` auf DIESEM Objekt.
- Ohne `self` würde Python nicht wissen, wo die Variable hingehört – sie würde nur lokal existieren und nach dem Aufruf verschwinden.

**In Methoden:**
- `self.name` bedeutet: **Lese den Namen** von DIESEM Objekt.
- Ohne `self` könnte die Methode nicht auf die Attribute des Objekts zugreifen – sie würde nicht wissen, ob es Enterprise, Millennium oder ein anderes Schiff ist.

**So funktioniert der Aufruf im Hintergrund:**
- `enterprise.starten()` wird von Python zu `Raumschiff.starten(enterprise)` übersetzt.
- Das erste Argument ist immer das Objekt selbst – deshalb heißt der Parameter `self`.

**Hast du die Fehler gefunden?** Hier sind die Lösungen:

````python
# ✅ Lösung Bug #1: Klasse ohne Großbuchstaben

# Problem: Klassen müssen mit einem Großbuchstaben beginnen
class Raumschiff:
    name = "Enterprise"

schiff = Raumschiff()
print(schiff.name)
````

````python
# ✅ Lösung Bug #2: Falscher Konstruktor

# Problem: __init__ braucht self als ersten Parameter – sonst weiß Python nicht,
# auf welchem Objekt die Attribute (name, power_level) gespeichert werden sollen.
class Cyborg:
    def __init__(self, name, power_level):
        self.name = name
        self.power_level = power_level

terminator = Cyborg("T-800", 9000)
print(terminator.name)
````

````python
# ✅ Lösung Bug #3: Methode ohne self

# Problem: Methoden brauchen immer self als ersten Parameter – ohne self kann
# die Methode nicht auf self.bezeichnung zugreifen (sie weiß nicht, welcher Roboter gemeint ist).
class Roboter:
    def __init__(self, bezeichnung):
        self.bezeichnung = bezeichnung
    
    def bewegen(self):
        print(f"{self.bezeichnung} bewegt sich! 🤖")

roboter = Roboter("R2-D2")
roboter.bewegen()
````
