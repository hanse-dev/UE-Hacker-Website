## üìö Systemprotokoll 1: Die Funktions-Definition - Der Befehlscode

**Was es ist:** Eine Funktion ist ein wiederverwendbarer Protokoll-Block, den du einmal definierst und immer wieder aufrufen kannst.

**Die Anatomie einer Funktion:**
```python
def protokoll_name(parameter1, parameter2):
    """Beschreibung des Protokolls"""
    # Protokoll-Code
    return ergebnis  # Optional: Ergebnis zur√ºckgeben
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`def`** - Das Kommando, das die Protokoll-Definition startet
2. **`protokoll_name`** - Ein eindeutiger Name f√ºr dein Protokoll
3. **`(parameter)`** - System-Variablen (optional)
4. **`:`** - Beendet die Protokoll-Definition
5. **`"""Docstring"""`** - Systemhandbuch (sehr wichtig!)
6. **Einger√ºckter Code** - Der eigentliche Protokoll-Code
7. **`return`** - Gibt das Ergebnis zur√ºck (optional)

**Warum so n√ºtzlich:**
- üéØ Protokolle wiederverwenden statt neu programmieren
- üîç System organisieren und strukturieren
- ‚ö° Fehler an einer Stelle beheben
- üé≤ Komplexe Systeme in kleine Teile zerlegen

````python
# Beispiel 1: Einfache Funktion ohne Parameter
def system_begruessung():
    """Begr√º√üt Besitzer auf der Raumstation"""
    print("Willkommen an Bord der Nebula-7!")
    print("Deine Mission beginnt jetzt!")

# Funktion aufrufen
print("=== Erster Besucher ===")
system_begruessung()

print("\n=== Zweiter Besucher ===")
system_begruessung()

print("\n=== Was passiert hier? ===")
print("1. def system_begruessung(): definiert die Funktion")
print("2. system_begruessung() ruft die Funktion auf")
print("3. Der Code in der Funktion wird jedes Mal ausgef√ºhrt")
````

````python
# Beispiel 2: Funktion mit einem Parameter
def begruesse_piloten(name):
    """Begr√º√üt einen Piloten mit seinem Namen"""
    print(f"Hallo {name}!")
    print(f"Viel Erfolg bei deiner Mission, {name}!")

# Funktion mit verschiedenen Namen aufrufen
print("=== Verschiedene Piloten ===")
begruesse_piloten("Alex")
begruesse_piloten("Zara")
begruesse_piloten("Nova")

print("\n=== Was passiert hier? ===")
print("1. name ist ein Platzhalter f√ºr den √ºbergebenen Wert")
print("2. Jeder Aufruf √ºbergibt einen anderen Namen")
print("3. Die Funktion arbeitet mit dem √ºbergebenen Wert")
````

````python
# Beispiel 3: Funktion mit mehreren Parametern
def erstelle_schiff_profil(name, klasse, besatzung):
    """Erstellt ein Schiff-Profil mit Name, Klasse und Besatzung"""
    print(f"Schiff-Profil:")
    print(f"  Name: {name}")
    print(f"  Klasse: {klasse}")
    print(f"  Besatzung: {besatzung} Personen")
    print(f"  Status: Bereit f√ºr den Einsatz!")

# Verschiedene Schiff-Profile erstellen
print("=== Schiff-Profile ===")
erstelle_schiff_profil("Nebula-Explorer", "Forschungsschiff", 150)
print()
erstelle_schiff_profil("Star-Fighter", "J√§ger", 2)
print()
erstelle_schiff_profil("Cargo-Hauler", "Frachter", 25)
````

## üîç Das Geheimnis von return - Das Systemergebnis

**return** ist der m√§chtigste Teil einer Funktion!

**Was return tut:**
- Beendet die Funktion sofort
- Gibt einen Wert zur√ºck an den Aufrufer
- Dieser Wert kann in einer Variable gespeichert werden

**Ohne return:** Funktion macht etwas (Side Effect)
**Mit return:** Funktion berechnet und gibt Ergebnis zur√ºck

````python
# üîç √úbungen mit return

# Beispiel 1: Einfache Berechnung mit return
def berechne_warp_distanz(warp_faktor, zeit):
    """Berechnet die zur√ºckgelegte Distanz im Warp"""
    distanz = warp_faktor * zeit * 1000
    return distanz

# Ergebnis in Variable speichern
print("=== Warp-Distanz-Berechnung ===")
distanz1 = berechne_warp_distanz(5, 10)
print(f"Sprung 1: {distanz1} Lichtjahre")

distanz2 = berechne_warp_distanz(8, 15)
print(f"Sprung 2: {distanz2} Lichtjahre")

# Ergebnis direkt verwenden
print(f"Notfall-Sprung: {berechne_warp_distanz(10, 5)} Lichtjahre")
````

````python
# Beispiel 2: Funktion mit bedingtem return
def pruefe_system_status(energie_level):
    """Pr√ºft, ob das System f√ºr eine Mission bereit ist"""
    if energie_level >= 75:
        return "System bereit f√ºr Mission!"
    else:
        return f"Noch {75 - energie_level}% Energie needed"

print("=== System-Status-Pr√ºfung ===")
status1 = pruefe_system_status(60)
print(f"Energie 60%: {status1}")

status2 = pruefe_system_status(90)
print(f"Energie 90%: {status2}")

status3 = pruefe_system_status(70)
print(f"Energie 70%: {status3}")
````

````python
# Beispiel 3: Funktion ohne expliziten return (None)
def zeige_ausruestung():
    """Zeigt die Standard-Ausr√ºstung (gibt nichts zur√ºck)"""
    print("Deine Ausr√ºstung:")
    print("  - Phaser")
    print("  - Tricorder")
    print("  - Kommunikator")
    # Kein return - gibt automatisch None zur√ºck

print("=== Ausr√ºstung anzeigen ===")
ergebnis = zeige_ausruestung()
print(f"R√ºckgabewert: {ergebnis}")
print("Funktionen ohne return geben immer None zur√ºck!")
````

## üì¶ Systemprotokoll 2: Docstrings - Die Systemhandb√ºcher

**Was es ist:** Docstrings sind Dokumentations-Strings, die erkl√§ren, was eine Funktion tut.

**Warum extrem wichtig:**
- üìñ Andere (und du sp√§ter) verstehen deine Funktion
- üîç IDEs zeigen dir Hilfe beim Aufruf
- ü§ñ Tools k√∂nnen automatisch Dokumentation erstellen
- ‚≠ê Zeichen von professionellem Code

**Die drei Arten von Docstrings:**
1. **Einzeilig:** `"""Kurze Beschreibung"""`
2. **Mehrzeilig:** `"""Kurze Beschreibung\n\nDetails..."""`
3. **Vollst√§ndig:** `"""Beschreibung\n\nArgs:\nReturns:\n"""`

````python
# Beispiel 1: Verschiedene Docstring-Stile
def einfache_funktion():
    """Dies ist ein einzeiliger Docstring."""
    return "Einfach"

def detaillierte_funktion(schiff_name, flotten_id):
    """Erstellt eine Missions-Beschreibung.
    
    Args:
        schiff_name (str): Der Name des Schiffs
        flotten_id (int): Die ID der Flotte
    
    Returns:
        str: Eine formatierte Beschreibung
    """
    return f"{schiff_name} geh√∂rt zur Flotte {flotten_id}."

# Docstring aufrufen (so funktioniert es!)
print("=== Docstring-Beispiele ===")
print(f"Einfacher Docstring: {einfache_funktion.__doc__}")
print()
print(f"Detaillierter Docstring: {detaillierte_funktion.__doc__}")
````

````python
# Beispiel 2: Hilfe-Funktion nutzen
print("=== Hilfe f√ºr Funktionen ===")
help(detaillierte_funktion)
````

## ‚ú® Systemprotokoll 3: Namenskonventionen - Die Regeln des Codes

**Gute Funktionsnamen sind entscheidend!**

**Die Regeln:**
- **Kleinbuchstaben** mit Unterstrichen: `berechne_warp_distanz`
- **Beschreibend**: Was die Funktion **tut** (Verb)
- **Kurz aber klar**: `get_name()` nicht `retrieve_starship_data_from_database()`
- **Keine Verben im Namen**: `name_get()` falsch, `get_name()` richtig

**Beispiele:**
- ‚úÖ `berechne_warp_distanz()`
- ‚úÖ `erstelle_schiff_profil()`
- ‚úÖ `pruefe_system_status()`
- ‚ùå `BerechneWarpDistanz()` (CamelCase)
- ‚ùå `berechne_das_warp()` (zu lang)
- ‚ùå `warp_berechnen()` (Substantiv zuerst)

````python
# ‚ú® Beispiele f√ºr gute und schlechte Namen

# ‚úÖ Gute Namen (klar und beschreibend)
def berechne_schild_effizienz(energie, flaeche):
    """Berechnet die Effizienz der Schilde."""
    return (energie / flaeche) * 100

def finde_naechsten_hafen(schiff_position, hafen_liste):
    """Findet den n√§chsten Hafen zur aktuellen Position."""
    return min(hafen_liste, key=lambda h: abs(h - schiff_position))

def ist_kommunikation_moeglich(reichweite, distanz):
    """Pr√ºft, ob Kommunikation m√∂glich ist."""
    return distanz <= reichweite

# ‚ùå Schlechte Namen (unklar oder falsche Konvention)
# def berechneSchildEffizienz():  # CamelCase
# def se(e, f):  # Zu kurz, nicht beschreibend
# def shield_calculation():  # Substantiv zuerst

print("=== Gute Funktionsnamen in Aktion ===")
effizienz = berechne_schild_effizienz(1000, 50)
print(f"Schild-Effizienz: {effizienz}%")

position = 100
haefen = [50, 200, 350, 500]
naechster_hafen = finde_naechsten_hafen(position, haefen)
print(f"N√§chster Hafen: {naechster_hafen}")

komm_ok = ist_kommunikation_moeglich(1000, 750)
print(f"Kommunikation m√∂glich: {komm_ok}")
````
