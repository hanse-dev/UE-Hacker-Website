## üìö Modul-Sammlung 1: Der Import-Protokoll - Das erste Modul

**Was es ist:** Ein Modul ist eine Sammlung von vordefinierten Funktionen, die du in dein Programm laden kannst.

**Die Anatomie eines Imports:**
```python
# Ganzes Modul importieren
import math

# Mit Alias importieren
import math as m

# Nur bestimmte Funktionen importieren
from math import pi, sqrt
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`import`** - Schl√ºsselwort zum Laden
2. **`math`** - Name des Moduls
3. **`as`** - Optionaler Alias/Spitzname
4. **`from`** - Importiert spezifische Funktionen

**Warum so n√ºtzlich:**
- üéØ Code nicht neu erfinden m√ºssen
- üîç Professionelle, optimierte Funktionen
- ‚ö° Dein Code bleibt √ºbersichtlich
- üé≤ Zugriff auf spezialisierte Werkzeuge

````python
# Beispiel 1: Module importieren
# Ganzes Modul importieren
import math
print(f"Pi aus math: {math.pi}")

# Mit Alias importieren
import random as r
print(f"Zufallszahl: {r.randint(1, 10)}")

# Nur bestimmte Funktionen importieren
from math import sqrt, sin, cos
print(f"Wurzel aus 16: {sqrt(16)}")
print(f"Sinus 90¬∞: {sin(1.57)}")
print(f"Kosinus 0¬∞: {cos(0)}")
````

````python
# Beispiel 2: Verschiedene Import-Methoden vergleichen
# Methode 1: import math
import math
print("Methode 1 - import math:")
print(f"  math.pi: {math.pi}")
print(f"  math.sqrt(25): {math.sqrt(25)}")

# Methode 2: import math as m
import math as m
print("\nMethode 2 - import math as m:")
print(f"  m.pi: {m.pi}")
print(f"  m.sqrt(25): {m.sqrt(25)}")

# Methode 3: from math import pi, sqrt
from math import pi, sqrt
print("\nMethode 3 - from math import pi, sqrt:")
print(f"  pi: {pi}")
print(f"  sqrt(25): {sqrt(25)}")
````

````python
# Beispiel 3: Mehrere Module importieren
import math
import random
import time

# Berechne eine zuf√§llige Flugdauer
distanz = 1000  # Lichtjahre
geschwindigkeit = random.uniform(0.1, 0.9)  # Bruchteil der Lichtgeschwindigkeit
flugzeit = distanz / geschwindigkeit

print("=== Raumflug-Simulation ===")
print(f"Distanz: {distanz} Lichtjahre")
print(f"Geschwindigkeit: {geschwindigkeit:.2f}c")
print(f"Flugzeit: {flugzeit:.1f} Jahre")
print(f"Warte 2 Sekunden...")
time.sleep(2)
print(f"Ankunft im Zielsystem!")
````

## ‚ú® Modul-Sammlung 3: Das random-Modul - Der Zufalls-Generator

**Die m√§chtigsten Zufalls-Funktionen:**

**Zahlen erzeugen:**
- `random.randint(a, b)` - Zuf√§llige Ganzzahl von a bis b
- `random.uniform(a, b)` - Zuf√§llige Kommazahl von a bis b
- `random.random()` - Zufallszahl zwischen 0 und 1

**Aus Listen w√§hlen:**
- `random.choice(liste)` - Ein zuf√§lliges Element
- `random.choices(liste, k=n)` - n zuf√§llige Elemente
- `random.shuffle(liste)` - Liste mischen

**Verteilungen:**
- `random.gauss(mu, sigma)` - Normalverteilung
- `random.expovariate(lambda)` - Exponentialverteilung

````python
# Beispiel 1: Zufallszahlen erzeugen
import random

print("=== Zufallszahlen ===")

# Ganzzahlen
print(f"Zuf√§llige Zahl 1-6: {random.randint(1, 6)}")
print(f"Zuf√§llige Zahl 0-100: {random.randint(0, 100)}")

# Kommazahlen
print(f"Zufallszahl 0-1: {random.random():.3f}")
print(f"Zufallszahl 10-20: {random.uniform(10, 20):.2f}")

# Mit Seeds reproduzierbar
random.seed(42)
print(f"Mit Seed 42: {random.randint(1, 100)}")
random.seed(42)
print(f"Mit Seed 42: {random.randint(1, 100)} (gleiches Ergebnis!)")
````

````python
# Beispiel 2: Aus Listen w√§hlen
import random

print("=== Zufallsauswahl aus Listen ===")

# Planeten im System
planeten = ["Merkur", "Venus", "Erde", "Mars", "Jupiter", "Saturn"]
print(f"Alle Planeten: {planeten}")

# Ein zuf√§lliger Planet
ziel = random.choice(planeten)
print(f"Zielplanet: {ziel}")

# Drei zuf√§llige Planeten
flotte = random.choices(planeten, k=3)
print(f"Flottenziele: {flotte}")

# Liste mischen
reihenfolge = planeten.copy()
random.shuffle(reihenfolge)
print(f"Gemischte Reihenfolge: {reihenfolge}")
````

````python
# Beispiel 3: Fortgeschrittene Zufallsfunktionen
import random
import math

print("=== Fortgeschrittene Zufallsfunktionen ===")

# Normalverteilung f√ºr Energiewerte
print("Energieverbrauch (Normalverteilung):")
for i in range(5):
    energie = random.gauss(100, 20)  # Mittelwert 100, Stdabw 20
    print(f"  System {i+1}: {max(0, energie):.1f} MW")

# Exponentialverteilung f√ºr Wartezeiten
print("\nWartezeiten (Exponentialverteilung):")
for i in range(5):
    warten = random.expovariate(0.1)  # Lambda = 0.1
    print(f"  Wartezeit {i+1}: {warten:.1f} Minuten")

# Zuf√§lliger Punkt im Kreis
def zufalls_punkt_kreis(radius):
    winkel = random.uniform(0, 2 * math.pi)
    r = radius * math.sqrt(random.random())
    x = r * math.cos(winkel)
    y = r * math.sin(winkel)
    return x, y

x, y = zufalls_punkt_kreis(10)
print(f"\nZuf√§lliger Punkt im Kreis r=10: ({x:.2f}, {y:.2f})")
````
