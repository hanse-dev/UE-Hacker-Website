## ğŸ” Die Algorithmen des Vergleichs: Wichtige Operatoren

### Zuweisung vs Vergleich: = vs ==

**Das ist ein entscheidender Unterschied!**

- **`=` (ein Gleichheitszeichen)**: Weist einer Variable einen Wert zu
  ```python
  level = 10  # Weise level den Wert 10 zu
  ```

- **`==` (zwei Gleichheitszeichen)**: Vergleicht zwei Werte auf Gleichheit
  ```python
  if level == 10:  # PrÃ¼fe OB level 10 ist
      print("Zugriffslevel 10 erreicht!")
  ```

### Die vier Vergleichsalgorithmen

- **`<` (kleiner als)**: PrÃ¼ft, ob der linke Wert kleiner ist
  ```python
  if energie < 20:
      print("âš ï¸ Niedrige Energie!")
  ```

- **`<=` (kleiner oder gleich)**: PrÃ¼ft, ob der linke Wert kleiner oder gleich ist
  ```python
  if temperatur <= 100:
      print("ğŸŒ¡ï¸ Normale Temperatur!")
  ```

- **`>` (grÃ¶ÃŸer als)**: PrÃ¼ft, ob der linke Wert grÃ¶ÃŸer ist
  ```python
  if geschwindigkeit > 1000:
      print("ğŸ’¨ Ãœberlichtgeschwindigkeit!")
  ```

- **`>=` (grÃ¶ÃŸer oder gleich)**: PrÃ¼ft, ob der linke Wert grÃ¶ÃŸer oder gleich ist
  ```python
  if schildstaerke >= 50:
      print("ğŸ›¡ï¸ Schilde stabil!")
  ```

**Merke:** In Bedingungen immer `==` fÃ¼r Vergleiche verwenden, nie `=`!

````python
# ğŸ” Ãœbungen zu Vergleichsoperatoren

# Beispiel 1: = vs == im Vergleich
rang = 15  # Zuweisung mit =
if rang == 15:  # Vergleich mit ==
    print("âœ… Richtig: Rang 15 erreicht")

# Beispiel 2: Alle vier Vergleichsoperatoren
wert = 42
print(f"Wert: {wert}")
print(f"{wert} < 50: {wert < 50}")
print(f"{wert} <= 42: {wert <= 42}")
print(f"{wert} > 30: {wert > 30}")
print(f"{wert} >= 42: {wert >= 42}")

# Beispiel 3: Praktische Anwendung
energie = 85
verbrauch = 60
if energie > verbrauch:
    print(f"âš¡ Sprung mÃ¶glich! Ãœbrig: {energie - verbrauch}%")
elif energie == verbrauch:
    print("âš–ï¸ Genau fÃ¼r einen Sprung!")
else:
    print(f"âŒ Sprung unmÃ¶glich! Fehlen: {verbrauch - energie}%")
````
