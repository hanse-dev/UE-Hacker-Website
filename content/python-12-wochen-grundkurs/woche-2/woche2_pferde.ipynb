## ğŸ“¦ Lektion 2: Die vier Hufschlag-Typen

**Was es ist:** Jede Information im Reiterhof hat einen Typ - wie die verschiedenen Gangarten eines Pferdes.

**Warum nÃ¼tzlich:**
- ğŸ´ **String (str):** FÃ¼r Namen, Rassen und Bezeichnungen

- ğŸŒ¾ **Float (float):** FÃ¼r Gewichte, MaÃŸe und Prozente
