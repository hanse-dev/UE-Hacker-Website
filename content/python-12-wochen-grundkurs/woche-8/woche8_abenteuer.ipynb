## üìö Datenstruktur 1: Der Steckbrief ‚Äì Eigenschaften mit Namen

**Was es ist:** Ein Dictionary ist wie ein **Steckbrief**: Zu jeder **Eigenschaft** (Schl√ºssel) geh√∂rt ein **Wert**. Du erstellst ihn mit geschweiften Klammern.

**Die Anatomie eines Steckbriefs:**
```python
# Leeren Steckbrief anlegen
leerer_steckbrief = {}

# Steckbrief mit Eigenschaften f√ºllen
held = {"name": "Aria", "klasse": "Magierin", "level": 15}
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`{}`** ‚Äì Geschweifte Klammern definieren einen Steckbrief
2. **`:`** ‚Äì Doppelpunkt trennt Eigenschaftsname (Schl√ºssel) vom Wert
3. **`,`** ‚Äì Kommas trennen die Eintr√§ge
4. **`[key]`** ‚Äì Mit dem Schl√ºssel greifst du auf den Wert zu

**Warum so n√ºtzlich:**
- üéØ Eigenschaften mit Namen statt Position (wie bei Listen)
- üîç Schneller Zugriff √ºber den Schl√ºssel
- ‚ö° Flexibel erweiterbar und √§nderbar
- üé≤ Verschiedene Werttypen pro Eigenschaft m√∂glich

````python
# Beispiel 1: Einfache Steckbriefe erstellen
# Leerer Steckbrief
leerer_steckbrief = {}
print(f"Leerer Steckbrief: {leerer_steckbrief}")

# Helden-Steckbrief
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15,
    "lebenspunkte": 120
}
print(f"Helden-Steckbrief: {held}")

# Monster-Steckbrief
monster = {
    "name": "Drache",
    "typ": "Feuer",
    "staerke": 50,
    "beute": True
}
print(f"Monster-Steckbrief: {monster}")
````

````python
# Beispiel 2: Auf Steckbriefe zugreifen
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15,
    "lebenspunkte": 120
}

print("=== Helden-Steckbrief auslesen ===")
print(f"Helden-Name: {held['name']}")
print(f"Klasse: {held['klasse']}")
print(f"Level: {held['level']}")
print(f"Lebenspunkte: {held['lebenspunkte']}")

# Mit get() Methode (sicherer)
print(f"\nMit get(): {held.get('name')}")
print(f"Nicht existierend: {held.get('mana', 'Unbekannt')}")
````

````python
# Beispiel 3: Steckbriefe √§ndern und erweitern
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15
}

print(f"Original: {held}")

# Wert √§ndern
held["level"] = 16
print(f"Nach Level-Up: {held}")

# Neue Eigenschaft hinzuf√ºgen
held["mana"] = 80
print(f"Nach Mana hinzuf√ºgen: {held}")

# Eintrag entfernen
entfernt = held.pop("klasse")
print(f"Entfernt: {entfernt}")
print(f"Nach entfernen: {held}")
````

## üì¶ Datenstruktur 2: Das Artefakt ‚Äì Die magische Kapsel

**Die wichtigsten Artefakt-Eigenschaften:**

**Erstellung:**
- `artefakt = (1, 2, 3)` - Mit runden Klammern
- `artefakt = 1, 2, 3` - Ohne Klammern
- `artefakt = (1,)` - Ein Element mit Komma

**Unver√§nderlichkeit:**
- Einmal erstellt, k√∂nnen Artefakte nicht ge√§ndert werden
- Perfekt f√ºr ewige Sch√§tze
- Magisch gesch√ºtzt vor Ver√§nderung

````python
# üîç √úbungen mit Artefakten

# Beispiel 1: Artefakte erstellen
koordinaten = (100, 200)
print(f"Koordinaten: {koordinaten}")

# Ohne Klammern
position = 10, 20
print(f"Position: {position}")

# Ein Element
einzel = (42,)
print(f"Einzelnes Artefakt: {einzel}")

# Zugriff auf Elemente
print(f"\nX-Koordinate: {koordinaten[0]}")
print(f"Y-Koordinate: {koordinaten[1]}")
````

````python
# Beispiel 2: Artefakte sind unver√§nderlich
zauber_status = ("Aktiv", "Stabil", "M√§chtig")
print(f"Zauber-Status: {zauber_status}")

# Zugriff funktioniert
print(f"Status: {zauber_status[0]}")

# √Ñndern funktioniert NICHT!
try:
    zauber_status[0] = "Inaktiv"
except TypeError as e:
    print(f"Fehler: {e}")

# Aber Artefakte in Listen k√∂nnen ge√§ndert werden
zauber = [
    ("Feuerball", "Aktiv"),
    ("Heilung", "Stabil"),
    ("Schild", "M√§chtig")
]
zauber[1] = ("Heilung", "Inaktiv")
print(f"\nZauber nach √Ñnderung: {zauber}")
````

````python
# Beispiel 3: Artefakt unpacking und Methoden
held_daten = ("Aria", "Magierin", 15, 120)

# Artefakt unpacking
name, klasse, level, lebenspunkte = held_daten
print("=== Artefakt Unpacking ===")
print(f"Name: {name}")
print(f"Klasse: {klasse}")
print(f"Level: {level}")
print(f"Lebenspunkte: {lebenspunkte}")

# Artefakt-Methoden
print(f"\nAnzahl Elemente: {len(held_daten)}")
print(f"Index von 'Magierin': {held_daten.index('Magierin')}")
print(f"Anzahl 'Magierin': {held_daten.count('Magierin')}")
````

import math

def generiere_schatz():
    typen = ["Gold", "Artefakt", "Juwel"]
    seltenheiten = ["normal", "selten", "legend√§r"]

        "name": f"Schatz-{random.randint(1, 1000)}",
        "typ": random.choice(typen),
        "wert": random.randint(10, 500),
        "seltenheit": random.choice(seltenheiten)

def berechne_entfernung(koord1, koord2):
    if len(koord1) != 2 or len(koord2) != 2:

    return math.sqrt((koord1[0]-koord2[0])**2 + (koord1[1]-koord2[1])**2)
