### ğŸ” Skill: GegenstÃ¤nde aus dem Inventar holen (Indexing)

Jeder Gegenstand in deinem Inventar hat eine **Slot-Nummer** (Index). 

**âš ï¸ Wichtige Magier-Regel:** In Pythonia fangen wir bei **0** an zu zÃ¤hlen!

- `inventar[0]` â†’ Slot 1 (erster Gegenstand)
- `inventar[1]` â†’ Slot 2 (zweiter Gegenstand)  
- `inventar[-1]` â†’ Letzter Slot (negative Zahlen zÃ¤hlen von hinten!)

````python
# ğŸ® Level 2: GegenstÃ¤nde aus dem Inventar nehmen

# Dein Trank-Inventar
traenke = ["Heiltrank", "Manatrank", "Gifttrank", "Unsichtbarkeitstrank", "StÃ¤rketrank"]

print("ğŸ§ª Dein Trank-Inventar:")
print(f"Slot 0 (Erster): {traenke[0]}")
print(f"Slot 1 (Zweiter): {traenke[1]}")
print(f"Slot -1 (Letzter): {traenke[-1]}")
print(f"Slot -2 (Vorletzter): {traenke[-2]}")

print("\nğŸ’¬ Der HÃ¤ndler fragt: 'Was ist in deinem ersten Slot?'")
print(f"Du antwortest: '{traenke[0]}!'")

print("\nğŸ‰ +150 XP: Du kannst jetzt gezielt auf GegenstÃ¤nde zugreifen!")
````

#### ğŸ“¥ Neue Beute aufsammeln
- `inventar.append(gegenstand)` â†’ Legt den Gegenstand in den nÃ¤chsten freien Slot
- `inventar.insert(slot, gegenstand)` â†’ Quetscht den Gegenstand an eine bestimmte Position

````python
# ğŸ® Level 3: Beute sammeln!

# Du startest deine Quest mit GrundausrÃ¼stung
ausruestung = ["Holzschwert", "Lederjacke"]
print(f"âš”ï¸ Start-AusrÃ¼stung: {ausruestung}")

# Du besiegst einen Goblin und findest einen Helm!
print("\nğŸ’¥ Du besiegst einen Goblin!")
ausruestung.append("Eisenhelm")
print(f"ğŸ“¦ Beute aufgesammelt! Inventar: {ausruestung}")

# Ein HÃ¤ndler verkauft dir einen Schild - den willst du an Position 1
print("\nğŸ›’ Du kaufst einen Schild beim HÃ¤ndler...")
ausruestung.insert(1, "Holzschild")
print(f"âœ¨ Inventar neu sortiert: {ausruestung}")

print("\nğŸ‰ +200 XP: Du kannst jetzt Beute aufsammeln!")
````

#### ğŸ—‘ï¸ GegenstÃ¤nde loswerden
- `inventar.remove(gegenstand)` â†’ Wirft einen bestimmten Gegenstand weg (nur das erste Vorkommen!)
- `inventar.pop(slot)` â†’ Nimmt Gegenstand aus Slot und gibt ihn dir zurÃ¼ck
- `del inventar[slot]` â†’ LÃ¶scht Gegenstand aus Slot (ohne RÃ¼ckgabe)

````python
# ğŸ® Level 4: Inventar-Management

# Deine Quest-Liste
quests = ["Goblins besiegen", "HeiltrÃ¤nke brauen", "Drachen finden", "Schatz suchen", "Drachen finden"]
print(f"ğŸ“‹ Deine Quests: {quests}")

# Du hast die erste Quest erledigt!
print("\nâœ… Quest abgeschlossen: 'Drachen finden'")
quests.remove("Drachen finden")  # Entfernt nur das ERSTE Vorkommen
print(f"ğŸ“‹ Verbleibende Quests: {quests}")

# Die letzte Quest ist die wichtigste - du nimmst sie raus um sie zu prÃ¼fen
print("\nğŸ” Welche Quest ist am wichtigsten?")
wichtigste_quest = quests.pop()  # Nimmt die letzte Quest
print(f"â­ Wichtigste Quest: {wichtigste_quest}")
print(f"ğŸ“‹ Ãœbrige Quests: {quests}")

# Die erste Quest ist zu einfach - weg damit!
print("\nğŸ—‘ï¸ Die erste Quest ist zu einfach...")
del quests[0]
print(f"ğŸ“‹ Finale Quest-Liste: {quests}")

print("\nğŸ‰ +250 XP: Du bist ein Inventar-Management-Profi!")
````
