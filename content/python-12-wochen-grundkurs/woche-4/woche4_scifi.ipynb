### ğŸ” Skill: Routinen-Parameter
Manchmal braucht eine Routine **Parameter**, um zu funktionieren! Eine FÃ¼tterungs-Routine braucht vielleicht die Anzahl der Raumschiffe, eine Wartungs-Routine braucht die Art der Wartung.
Diese "Parameter" nennen wir **Schleifen-Variablen**:
- ğŸ“ **Definition**: `for i in range(5):` â€“ `i` ist eine Schleifen-Variable (ZÃ¤hler)
- âš¡ **Aufruf**: Die Schleife lÃ¤uft automatisch von 0 bis 4
**Merke:** Parameter = ZÃ¤hler im Schleifenkopf!

````python
# ğŸš€ Level 3: Routinen mit Parametern!
# Definition mit Schleifen-Variable 'i' (ZÃ¤hler fÃ¼r Wiederholungen)
def system_check(anzahl):
    print(f"ğŸŒŸ System-Check fÃ¼r {anzahl} Raumschiffe gestartet!")
    for i in range(anzahl):
        print(f"âœ… Raumschiff {i+1}: Systeme OK!")
# Aufrufe mit verschiedenen Parametern (echte Arbeitsmittel)
print("ğŸ§™ Die Missionsleiterin startet die System-Checks:\n")
system_check(3)
print()
system_check(5)
print("\nğŸ‰ +150 Cyber Credits: Du kannst jetzt Routinen mit Parametern erstellen!")
````

### âš¡ Power-Moves: `while` â€“ Endlose Wiederholungen
Stell dir vor, du fragst eine Raumfahrt-Expertin: "Wie oft soll ich das System prÃ¼fen?" 
Manchmal weiÃŸt du nicht genau, wie oft etwas wiederholt werden soll. Dann nutzt du `while`!
**Das macht `while`:**
- ğŸ”„ Wiederholt, solange eine Bedingung wahr ist
- ğŸ›‘ Beendet die Schleife, wenn die Bedingung falsch wird
- ğŸ’¾ ErmÃ¶glicht es, flexible Wiederholungen zu erstellen
**Wann brauchst du `while`?**
- âœ… Wenn du nicht genau weiÃŸt, wie oft etwas wiederholt werden soll
- âœ… Wenn du auf eine Bedingung warten musst
- âœ… Wenn du etwas abbrechen willst, wenn ein bestimmtes Ereignis eintritt

````python
# ğŸš€ Level 4: Routinen mit praktischen Ergebnissen!
# Diese Routine nutzt while fÃ¼r flexible Wiederholungen
def treibstoff_pruefung():
    treibstoff = 100
    while treibstoff > 20:
        print(f"â›½ Treibstoff-Level: {treibstoff}% - Mission lÃ¤uft!")
        treibstoff -= 10
    print("ğŸš¨ Treibstoff niedrig! ZurÃ¼ck zur Basis!")
# Teste die Routine
print("ğŸ§® Die Treibstoff-PrÃ¼fung fÃ¼r die Raumschiffe wird durchgefÃ¼hrt...")
treibstoff_pruefung()
print("\nğŸ‰ +200 Cyber Credits: Du beherrschst jetzt while-Schleifen!")
````
