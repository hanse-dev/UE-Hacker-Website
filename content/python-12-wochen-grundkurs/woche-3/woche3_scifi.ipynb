### ğŸ”— Power-Move: Logische Operatoren

Mit logischen Operatoren kannst du Bedingungen kombinieren:


### ğŸ¯ Skill-ErklÃ¤rung: and vs or

**and** = Beide mÃ¼ssen wahr sein:
- `True and True` â†’ True
- `True and False` â†’ False
- `False and False` â†’ False

**or** = Mindestens eines muss wahr sein:
- `True or True` â†’ True
- `True or False` â†’ True
- `False or False` â†’ False

**EselsbrÃ¼cke:** 
- **and** = streng (BEIDE Bedingungen)
- **or** = groÃŸzÃ¼gig (EINE Bedingung reicht)

````python
# ğŸ° Level 7: not-Operator

# Der not-Operator kehrt eine Bedingung um
raumschiff = "Nebula-Explorer"
system_fehler = False
unter_angriff = False
treibstoff_leck = False

print(f"ğŸš€ Status von {raumschiff}:")
print(f"âš ï¸ System-Fehler: {system_fehler}")
print(f"ğŸ›¡ï¸ Unter Angriff: {unter_angriff}")
print(f"â›½ Treibstoff-Leck: {treibstoff_leck}")
print()

# not kehrt True zu False um (und umgekehrt)
if not system_fehler:
    print("âœ… Keine System-Fehler!")

if not unter_angriff:
    print("âœ… Nicht unter Angriff!")

if not treibstoff_leck:
    print("âœ… Kein Treibstoff-Leck!")

# Kombiniere not mit and
if not system_fehler and not unter_angriff and not treibstoff_leck:
    print()
    print(f"ğŸ’ª {raumschiff} ist in perfektem Zustand!")
    print("ğŸš€ Bereit fÃ¼r jede Mission!")
    print("ğŸŒŒ Alle Systeme nominal.")

print()

# Noch ein Beispiel
wartung_fÃ¤llig = False

if not wartung_fÃ¤llig:
    print("âœ… Keine Wartung erforderlich.")
    print("ğŸš€ Raumschiff ist einsatzbereit!")

print()
print("ğŸ‰ +250 Cyber Credits: not-Operator verstanden!")
````
