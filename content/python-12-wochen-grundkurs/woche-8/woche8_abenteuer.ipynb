## ğŸ“š Schatz-Struktur 1: Die Schatztruhe - Der magische BehÃ¤lter

**Was es ist:** Ein Dictionary ist eine Sammlung von SchlÃ¼ssel-Wert-Paaren, die du mit geschweiften Klammern erstellst.

**Die Anatomie einer Schatztruhe:**
```python
# Leere Schatztruhe erstellen
leere_truhe = {}

# Schatztruhe mit SchÃ¤tzen fÃ¼llen
held = {"name": "Aria", "klasse": "Magierin", "level": 15}
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`{}`** - Geschweifte Klammern definieren eine Schatztruhe
2. **`:`** - Doppelpunkt trennt SchlÃ¼ssel vom Wert
3. **`,`** - Kommas trennen die Paare
4. **`[key]`** - Mit SchlÃ¼ssel greifst du auf SchÃ¤tze zu

**Warum so nÃ¼tzlich:**
- ğŸ¯ SchÃ¤tze mit Namen statt Nummer organisieren
- ğŸ” Schneller Zugriff Ã¼ber magische SchlÃ¼ssel
- âš¡ Flexibel erweiterbar und Ã¤nderbar
- ğŸ² Verschiedene Schatztypen mischbar

````python
# Beispiel 1: Einfache Schatztruhen erstellen
# Leere Schatztruhe
leere_truhe = {}
print(f"Leere Schatztruhe: {leere_truhe}")

# Helden-Schatztruhe
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15,
    "lebenspunkte": 120
}
print(f"Helden-Schatztruhe: {held}")

# Monster-Schatztruhe
monster = {
    "name": "Drache",
    "typ": "Feuer",
    "staerke": 50,
    "beute": True
}
print(f"Monster-Schatztruhe: {monster}")
````

````python
# Beispiel 2: Auf Schatztruhen zugreifen
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15,
    "lebenspunkte": 120
}

print("=== Helden-Zugriff ===")
print(f"Helden-Name: {held['name']}")
print(f"Klasse: {held['klasse']}")
print(f"Level: {held['level']}")
print(f"Lebenspunkte: {held['lebenspunkte']}")

# Mit get() Methode (sicherer)
print(f"\nMit get(): {held.get('name')}")
print(f"Nicht existierend: {held.get('mana', 'Unbekannt')}")
````

````python
# Beispiel 3: Schatztruhen Ã¤ndern und erweitern
held = {
    "name": "Aria",
    "klasse": "Magierin",
    "level": 15
}

print(f"Original: {held}")

# Wert Ã¤ndern
held["level"] = 16
print(f"Nach Level-Up: {held}")

# Neuen Schatz hinzufÃ¼gen
held["mana"] = 80
print(f"Nach Mana hinzufÃ¼gen: {held}")

# Schatz entfernen
entfernt = held.pop("klasse")
print(f"Entfernt: {entfernt}")
print(f"Nach entfernen: {held}")
````

## ğŸ“¦ Schatz-Struktur 2: Das Artefakt - Die magische Kapsel

**Die wichtigsten Artefakt-Eigenschaften:**

**Erstellung:**
- `artefakt = (1, 2, 3)` - Mit runden Klammern
- `artefakt = 1, 2, 3` - Ohne Klammern
- `artefakt = (1,)` - Ein Element mit Komma

**UnverÃ¤nderlichkeit:**
- Einmal erstellt, kÃ¶nnen Artefakte nicht geÃ¤ndert werden
- Perfekt fÃ¼r ewige SchÃ¤tze
- Magisch geschÃ¼tzt vor VerÃ¤nderung

````python
# ğŸ” Ãœbungen mit Artefakten

# Beispiel 1: Artefakte erstellen
koordinaten = (100, 200)
print(f"Koordinaten: {koordinaten}")

# Ohne Klammern
position = 10, 20
print(f"Position: {position}")

# Ein Element
einzel = (42,)
print(f"Einzelnes Artefakt: {einzel}")

# Zugriff auf Elemente
print(f"\nX-Koordinate: {koordinaten[0]}")
print(f"Y-Koordinate: {koordinaten[1]}")
````

````python
# Beispiel 2: Artefakte sind unverÃ¤nderlich
zauber_status = ("Aktiv", "Stabil", "MÃ¤chtig")
print(f"Zauber-Status: {zauber_status}")

# Zugriff funktioniert
print(f"Status: {zauber_status[0]}")

# Ã„ndern funktioniert NICHT!
try:
    zauber_status[0] = "Inaktiv"
except TypeError as e:
    print(f"Fehler: {e}")

# Aber Artefakte in Listen kÃ¶nnen geÃ¤ndert werden
zauber = [
    ("Feuerball", "Aktiv"),
    ("Heilung", "Stabil"),
    ("Schild", "MÃ¤chtig")
]
zauber[1] = ("Heilung", "Inaktiv")
print(f"\nZauber nach Ã„nderung: {zauber}")
````

````python
# Beispiel 3: Artefakt unpacking und Methoden
held_daten = ("Aria", "Magierin", 15, 120)

# Artefakt unpacking
name, klasse, level, lebenspunkte = held_daten
print("=== Artefakt Unpacking ===")
print(f"Name: {name}")
print(f"Klasse: {klasse}")
print(f"Level: {level}")
print(f"Lebenspunkte: {lebenspunkte}")

# Artefakt-Methoden
print(f"\nAnzahl Elemente: {len(held_daten)}")
print(f"Index von 'Magierin': {held_daten.index('Magierin')}")
print(f"Anzahl 'Magierin': {held_daten.count('Magierin')}")
````

## âœ¨ Schatz-Struktur 3: Verschachtelte Strukturen - Das Schatz-Netzwerk

**Die mÃ¤chtigsten Kombinationen:**

**Schatztruhe in Schatztruhe:**
- `person = {"name": "Alex", "ausruestung": {"waffe": "X", "ruestung": "Y"}}`

**Liste von Schatztruhen:**
- `helden = [{"name": "A", "klasse": "M"}, {"name": "B", "klasse": "K"}]`

**Artefakt als Schatztruhe-SchlÃ¼ssel:**
- `positionen = {(10, 20): "Punkt A", (30, 40): "Punkt B"}`

````python
# Beispiel 1: Schatztruhe in Schatztruhe
dungeon = {
    "name": "Verdammnis-Festung",
    "location": {
        "region": "Schattenlande",
        "koordinaten": (100, 200),
        "reich": "Pyralia"
    },
    "gefaehrten": {
        "boss": "Lich-KÃ¶nig",
        "monster": "Skelette",
        "fallen": "Magisch"
    }
}

print("=== Verschachtelte Schatztruhe ===")
print(f"Dungeon: {dungeon['name']}")
print(f"Region: {dungeon['location']['region']}")
print(f"Koordinaten: {dungeon['location']['koordinaten']}")
print(f"Boss: {dungeon['gefaehrten']['boss']}")
````

````python
# Beispiel 2: Liste von Schatztruhen
helden = [
    {
        "name": "Aria",
        "klasse": "Magierin",
        "level": 15,
        "erfahrung": 2500
    },
    {
        "name": "Thorin",
        "klasse": "Krieger",
        "level": 18,
        "erfahrung": 3200
    },
    {
        "name": "Luna",
        "klasse": "Schurkin",
        "level": 12,
        "erfahrung": 1800
    }
]

print("=== Helden-Liste ===")
for held in helden:
    print(f"{held['name']} - {held['klasse']} (Level {held['level']})")

# Nach Level filtern
starke = [h for h in helden if h['level'] > 14]
print(f"\nStarke Helden: {[h['name'] for h in starke]}")
````

````python
# Beispiel 3: Komplexe Schatz-Strukturen
# Artefakt als Schatztruhe-SchlÃ¼ssel
schatz_karte = {
    (100, 200): "Goldtruhe",
    (300, 400): "Magie-Truhe",
    (500, 600): "Artefakt-Truhe"
}

print("=== Schatz-Karte ===")
for position, schatz in schatz_karte.items():
    print(f"{position}: {schatz}")

# Komplexes Questsystem
quests = {
    "aktiv": [
        {
            "id": 1,
            "ziel": "Drache besiegen",
            "belohnung": ("Gold", 1000, "XP", 500),
            "status": "In Bearbeitung"
        }
    ],
    "abgeschlossen": [
        {
            "id": 2,
            "ziel": "Schatz finden",
            "belohnung": ("Silber", 500, "XP", 250),
            "status": "Erfolg"
        }
    ]
}

print(f"\nAktive Quests: {len(quests['aktiv'])}")
print(f"Abgeschlossene Quests: {len(quests['abgeschlossen'])}")
````
