## ğŸ“š Daten-Sammlung 1: Die Listen-Erstellung - Der erste Datensatz

**Was es ist:** Eine Liste ist eine geordnete Sammlung von Elementen, die du mit eckigen Klammern erstellst.

**Die Anatomie einer Liste:**
```python
# Leere Liste erstellen
leere_liste = []

# Liste mit Elementen erstellen
schiffe = ["Nebula-Explorer", "Star-Fighter", "Cargo-Hauler"]
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`[]`** - Eckige Klammern definieren eine Liste
2. **`,`** - Kommas trennen die Elemente
3. **`""`** - Strings werden in AnfÃ¼hrungszeichen gesetzt
4. **`[]`** - Mit Index greifst du auf Elemente zu (beginnt bei 0)

**Warum so nÃ¼tzlich:**
- ğŸ¯ Mehrere Werte in einer Variable speichern
- ğŸ” Geordnete Struktur fÃ¼r Daten
- âš¡ Flexibel erweiterbar und verÃ¤nderbar
- ğŸ² Verschiedene Datentypen mischbar

````python
# Beispiel 1: Einfache Listen erstellen
# Leere Liste
leere_liste = []
print(f"Leere Liste: {leere_liste}")

# Liste mit Schiffen
schiffe = ["Nebula-Explorer", "Star-Fighter", "Cargo-Hauler", "Science-Vessel"]
print(f"Schiff-Liste: {schiffe}")

# Liste mit Zahlen
koordinaten = [100, 200, 300, 400, 500]
print(f"Koordinaten: {koordinaten}")

# Gemischte Liste
mixed = ["Phaser", 25, True, "Tricorder"]
print(f"Gemischte Liste: {mixed}")
````

## ğŸ“¦ Daten-Sammlung 2: Elemente verwalten - Der Daten-Kurator

**Die wichtigsten Listen-Methoden:**

**Elemente hinzufÃ¼gen:**
- `append(element)` - FÃ¼gt am Ende hinzu
- `insert(index, element)` - FÃ¼gt an bestimmter Position ein

**Elemente entfernen:**
- `remove(element)` - Entfernt erstes Vorkommen
- `pop(index)` - Entfernt und gibt Element zurÃ¼ck
- `clear()` - Leert die gesamte Liste

````python
# ğŸ” Ãœbungen mit append() und insert()

# Beispiel 1: Elemente hinzufÃ¼gen
missions_log = []
print(f"Leeres Missions-Log: {missions_log}")

# Mit append() am Ende hinzufÃ¼gen
missions_log.append("Planetenerkundung")
print(f"Nach append: {missions_log}")

missions_log.append("Asteroidenabbau")
missions_log.append("Erste-Kontakt-Mission")
print(f"Voll gefÃ¼llt: {missions_log}")

# Mit insert() an bestimmter Position einfÃ¼gen
missions_log.insert(0, "System-Check")  # Am Anfang
print(f"Nach insert(0): {missions_log}")

missions_log.insert(2, "Wartung")  # An Position 2
print(f"Nach insert(2): {missions_log}")
````

````python
# Beispiel 2: Elemente entfernen
system_protokolle = ["Alpha", "Beta", "Gamma", "Delta", "Epsilon", "Gamma"]
print(f"Original-Protokolle: {system_protokolle}")

# Mit remove() erstes Vorkommen entfernen
system_protokolle.remove("Gamma")
print(f"Nach remove('Gamma'): {system_protokolle}")

# Mit pop() letztes Element entfernen und zurÃ¼ckgeben
letztes_protokoll = system_protokolle.pop()
print(f"Nach pop(): {system_protokolle}")
print(f"Entferntes Protokoll: {letztes_protokoll}")

# Mit pop(index) bestimmtes Element entfernen
zweites_protokoll = system_protokolle.pop(1)
print(f"Nach pop(1): {system_protokolle}")
print(f"Entferntes Protokoll: {zweites_protokoll}")

# Mit clear() alles lÃ¶schen
system_protokolle.clear()
print(f"Nach clear(): {system_protokolle}")
````

````python
# Beispiel 3: Listen kombinieren
schiffe = ["Nebula-Explorer", "Star-Fighter"]
crew = ["Captain Alex", "Dr. Zara"]
missionen = ["Forschung", "Verteidigung", "Erforschung"]

print("=== Listen kombinieren ===")
print(f"Schiffe: {schiffe}")
print(f"Crew: {crew}")
print(f"Missionen: {missionen}")

# Mit + Operator kombinieren
alle = schiffe + crew + missionen
print(f"Alles kombiniert: {alle}")

# Mit extend() eine Liste zur anderen hinzufÃ¼gen
schiffe.extend(crew)
print(f"Schiffe erweitert: {schiffe}")
````

## âœ¨ Daten-Sammlung 3: Listen durchsuchen und sortieren - Der Such-Algorithmus

**Die mÃ¤chtigsten Such-Methoden:**

**Suchen und prÃ¼fen:**
- `index(element)` - Gibt Index des Elements zurÃ¼ck
- `element in liste` - PrÃ¼ft ob Element vorhanden
- `count(element)` - ZÃ¤hlt Vorkommen

**Sortieren:**
- `sort()` - Sortiert die Liste selbst
- `sorted(liste)` - Gibt sortierte Kopie zurÃ¼ck
- `reverse()` - Kehrt Reihenfolge um

````python
# Beispiel 1: In Listen suchen
systeme = ["Waffensystem", "Schildsystem", "Antriebssystem", "Lebenserhaltung", "Waffensystem", "Antriebssystem"]

print("=== In Listen suchen ===")
print(f"System-Liste: {systeme}")

# Mit index() Position finden
position_schild = systeme.index("Schildsystem")
print(f"Schildsystem an Position: {position_schild}")

# Mit in() prÃ¼fen ob vorhanden
hat_antrieb = "Antriebssystem" in systeme
hat_teleporter = "Teleporter" in systeme
print(f"Antriebssystem vorhanden: {hat_antrieb}")
print(f"Teleporter vorhanden: {hat_teleporter}")

# Mit count() zÃ¤hlen
anzahl_waffen = systeme.count("Waffensystem")
print(f"Anzahl Waffensysteme: {anzahl_waffen}")
````

````python
# Beispiel 2: Listen sortieren
prioritaeten = [5, 2, 8, 1, 9, 3]
sektoren = ["Zeta", "Alpha", "Gamma", "Beta", "Delta"]

print("=== Listen sortieren ===")
print(f"Original-PrioritÃ¤ten: {prioritaeten}")
print(f"Original-Sektoren: {sektoren}")

# Mit sort() direkt sortieren
prioritaeten.sort()
print(f"PrioritÃ¤ten nach sort(): {prioritaeten}")

# Mit sorted() Kopie sortieren
sortierte_sektoren = sorted(sektoren)
print(f"Sektoren Original: {sektoren}")
print(f"Sektoren sortiert: {sortierte_sektoren}")

# Umgekehrt sortieren
prioritaeten.sort(reverse=True)
print(f"PrioritÃ¤ten absteigend: {prioritaeten}")

# Mit reverse() Reihenfolge umkehren
sektoren.reverse()
print(f"Sektoren umgekehrt: {sektoren}")
````

````python
# Beispiel 3: Fortgeschrittene Listen-Operationen
mission_typen = ["Forschung", "Verteidigung", "Erforschung", "Forschung", "Verteidigung"]

print("=== Fortgeschrittene Operationen ===")
print(f"Missionstypen-Liste: {mission_typen}")

# Einzigartige Elemente (mit set)
einzigartige_typen = list(set(mission_typen))
print(f"Einzigartige Missionstypen: {einzigartige_typen}")

# Listen mit for-Schleife durchlaufen
print("\nAlle Missionstypen:")
for i, m in enumerate(mission_typen):
    print(f"  Position {i}: {m}")

# List Comprehension (fortgeschritten)
lange_namen = [m for m in mission_typen if len(m) > 8]
print(f"\nMissionstypen mit langen Namen: {lange_namen}")
````
