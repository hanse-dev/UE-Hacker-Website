## ğŸ“¦ Lektion 2: Die vier Hufschlag-Typen

**Was es ist:** Jede Information im Reiterhof hat einen Typ - wie die verschiedenen Gangarten eines Pferdes.

**Warum nÃ¼tzlich:**
- ğŸ´ **String (str):** FÃ¼r Namen, Rassen und Bezeichnungen

- ğŸŒ¾ **Float (float):** FÃ¼r Gewichte, MaÃŸe und Prozente
- âœ… **Boolean (bool):** FÃ¼r Ja/Nein-Entscheidungen im Stall

**Wichtig:** `type()` zeigt dir den Typ jeder Information!

```python
# Beispiel 1: Strings - Textinformationen
pferd_name = "Windsturm"
rasse = "Hannoveraner"
print(f"Pferd: {pferd_name}")
print(f"Rasse: {rasse}")
print(f"Typ: {type(pferd_name)}")

# Beispiel 2: Integers - Ganze Zahlen
alter = 7
stockmass = 168  # cm
print(f"Alter: {alter} Jahre, StockmaÃŸ: {stockmass} cm")
print(f"Typ: {type(alter)}")

# Beispiel 3: Floats - Kommazahlen
gewicht = 550.5
futter_menge = 4.25
print(f"Gewicht: {gewicht} kg, Futter: {futter_menge} kg")
print(f"Typ: {type(gewicht)}")

# Beispiel 4: Booleans - Wahrheitswerte
hat_sattel = True
ist_besucht = False
print(f"Hat Sattel: {hat_sattel}")
print(f"Ist besucht: {ist_besucht}")
print(f"Typ: {type(hat_sattel)}")
```
