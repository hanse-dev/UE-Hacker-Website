## ğŸ“š Zauberformel 1: Die Funktions-Definition - Der magische Vertrag

**Was es ist:** Eine Funktion ist ein wiederverwendbarer Code-Block, den du einmal definierst und immer wieder aufrufen kannst.

**Die Anatomie einer Funktion:**
```python
def funktions_name(parameter1, parameter2):
    """Kurze Beschreibung des Zaubers"""
    # Code, der ausgefÃ¼hrt wird
    return ergebnis  # Optional: Magie-Ergebnis zurÃ¼ckgeben
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`def`** - Das magische Wort, das die Funktions-Definition startet
2. **`funktions_name`** - Ein eindeutiger Name fÃ¼r deinen Zauber
3. **`(parameter)`** - Zutaten, die dein Zauber braucht (optional)
4. **`:`** - Beendet die Funktions-Definition
5. **`"""Docstring"""`** - Magie-Buch mit Beschreibung (sehr wichtig!)
6. **EingerÃ¼ckter Code** - Der eigentliche Zauber-Code
7. **`return`** - Gibt das Ergebnis zurÃ¼ck (optional)

**Warum so nÃ¼tzlich:**
- ğŸ¯ Code wiederverwenden statt kopieren
- ğŸ” Logik organisieren und strukturieren
- âš¡ Fehler an einer Stelle beheben
- ğŸ² Komplexe Probleme in kleine Teile zerlegen

````python
# Beispiel 1: Einfache Funktion ohne Parameter
def gruesse_abenteurer():
    """BegrÃ¼ÃŸt einen tapferen Abenteurer"""
    print("Willkommen, tapferer Held!")
    print("Dein Abenteuer beginnt jetzt!")

# Funktion aufrufen
print("=== Erster Aufruf ===")
gruesse_abenteurer()

print("\n=== Zweiter Aufruf ===")
gruesse_abenteurer()

print("\n=== Was passiert hier? ===")
print("1. def gruesse_abenteurer(): definiert die Funktion")
print("2. gruesse_abenteurer() ruft die Funktion auf")
print("3. Der Code in der Funktion wird jedes Mal ausgefÃ¼hrt")
````

````python
# Beispiel 2: Funktion mit einem Parameter
def gruesse_nach_name(name):
    """BegrÃ¼ÃŸt einen Helden mit seinem Namen"""
    print(f"Willkommen, {name}!")
    print(f"Viel Erfolg auf deiner Reise, {name}!")

# Funktion mit verschiedenen Namen aufrufen
print("=== Verschiedene Helden ===")
gruesse_nach_name("Aria")
gruesse_nach_name("Thorin")
gruesse_nach_name("Luna")

print("\n=== Was passiert hier? ===")
print("1. name ist ein Platzhalter fÃ¼r den Ã¼bergebenen Wert")
print("2. Jeder Aufruf Ã¼bergibt einen anderen Namen")
print("3. Die Funktion arbeitet mit dem Ã¼bergebenen Wert")
````

````python
# Beispiel 3: Funktion mit mehreren Parametern
def erstelle_charakter(name, klasse, level):
    """Erstellt einen Charakter mit Name, Klasse und Level"""
    print(f"Charakter erstellt:")
    print(f"  Name: {name}")
    print(f"  Klasse: {klasse}")
    print(f"  Level: {level}")
    print(f"  Status: Bereit fÃ¼r Abenteuer!")

# Verschiedene Charaktere erstellen
print("=== Charakter-Erstellung ===")
erstelle_charakter("Aria", "Magierin", 5)
print()
erstelle_charakter("Thorin", "Krieger", 8)
print()
erstelle_charakter("Luna", "Schurkin", 3)
````

## ğŸ” Das Geheimnis von return - Die magische RÃ¼ckgabe

**return** ist der mÃ¤chtigste Teil einer Funktion!

**Was return tut:**
- Beendet die Funktion sofort
- Gibt einen Wert zurÃ¼ck an den Aufrufer
- Dieser Wert kann in einer Variable gespeichert werden

**Ohne return:** Funktion macht etwas (Side Effect)
**Mit return:** Funktion berechnet und gibt Ergebnis zurÃ¼ck

````python
# ğŸ” Ãœbungen mit return

# Beispiel 1: Einfache Berechnung mit return
def berechne_schaden(basis_schaden, multiplikator):
    """Berechnet den Gesamtschaden eines Angriffs"""
    gesamtschaden = basis_schaden * multiplikator
    return gesamtschaden

# Ergebnis in Variable speichern
print("=== Schadensberechnung ===")
schaden1 = berechne_schaden(10, 2)
print(f"Schlag 1: {schaden1} Schaden")

schaden2 = berechne_schaden(15, 3)
print(f"Schlag 2: {schaden2} Schaden")

# Ergebnis direkt verwenden
print(f"Kritischer Treffer: {berechne_schaden(20, 5)} Schaden")
````

````python
# Beispiel 2: Funktion mit bedingtem return
def pruefe_level(level):
    """PrÃ¼ft, ob ein Charakter fÃ¼r eine Quest bereit ist"""
    if level >= 10:
        return "Bereit fÃ¼r die Quest!"
    else:
        return f"Noch {10 - level} Level needed"

print("=== Level-PrÃ¼fung ===")
status1 = pruefe_level(5)
print(f"Level 5: {status1}")

status2 = pruefe_level(12)
print(f"Level 12: {status2}")

status3 = pruefe_level(8)
print(f"Level 8: {status3}")
````

````python
# Beispiel 3: Funktion ohne expliziten return (None)
def zeige_inventar():
    """Zeigt das Inventar (gibt nichts zurÃ¼ck)"""
    print("Dein Inventar:")
    print("  - Schwert")
    print("  - Schild")
    print("  - Heiltrank")
    # Kein return - gibt automatisch None zurÃ¼ck

print("=== Inventar anzeigen ===")
ergebnis = zeige_inventar()
print(f"RÃ¼ckgabewert: {ergebnis}")
print("Funktionen ohne return geben immer None zurÃ¼ck!")
````

## ğŸ“¦ Zauberformel 2: Docstrings - Die Magie-BÃ¼cher

**Was es ist:** Docstrings sind Dokumentations-Strings, die erklÃ¤ren, was eine Funktion tut.

**Warum extrem wichtig:**
- ğŸ“– Andere (und du spÃ¤ter) verstehen deine Funktion
- ğŸ” IDEs zeigen dir Hilfe beim Aufruf
- ğŸ¤– Tools kÃ¶nnen automatisch Dokumentation erstellen
- â­ Zeichen von professionellem Code

**Die drei Arten von Docstrings:**
1. **Einzeilig:** `"""Kurze Beschreibung"""`
2. **Mehrzeilig:** `"""Kurze Beschreibung\n\nDetails..."""`
3. **VollstÃ¤ndig:** `"""Beschreibung\n\nArgs:\nReturns:\n"""`

````python
# Beispiel 1: Verschiedene Docstring-Stile
def einfache_funktion():
    """Dies ist ein einzeiliger Docstring."""
    return "Einfach"

def detaillierte_funktion(name, level):
    """Erstellt eine Charakter-Beschreibung.
    
    Args:
        name (str): Der Name des Charakters
        level (int): Das Level des Charakters
    
    Returns:
        str: Eine formatierte Beschreibung
    """
    return f"{name} ist ein Level {level} Charakter."

# Docstring aufrufen (so funktioniert es!)
print("=== Docstring-Beispiele ===")
print(f"Einfacher Docstring: {einfache_funktion.__doc__}")
print()
print(f"Detaillierter Docstring: {detaillierte_funktion.__doc__}")
````

````python
# Beispiel 2: Hilfe-Funktion nutzen
print("=== Hilfe fÃ¼r Funktionen ===")
help(detaillierte_funktion)
````

## âœ¨ Zauberformel 3: Namenskonventionen - Die Regeln der Magie

**Gute Funktionsnamen sind entscheidend!**

**Die Regeln:**
- **Kleinbuchstaben** mit Unterstrichen: `berechne_schaden`
- **Beschreibend**: Was die Funktion **tut** (Verb)
- **Kurz aber klar**: `get_name()` nicht `retrieve_user_data_from_database()`
- **Keine Verben im Namen**: `name_get()` falsch, `get_name()` richtig

**Beispiele:**
- âœ… `berechne_schaden()`
- âœ… `erstelle_charakter()`
- âœ… `pruefe_level()`
- âŒ `BerechneSchaden()` (CamelCase)
- âŒ `berechne_das_schaden()` (zu lang)
- âŒ `schaden_berechnen()` (Substantiv zuerst)

````python
# âœ¨ Beispiele fÃ¼r gute und schlechte Namen

# âœ… Gute Namen (klar und beschreibend)
def berechne_heilung(basis_heilung, multiplikator):
    """Berechnet die Heilungsmenge."""
    return basis_heilung * multiplikator

def finde_max_level(level_liste):
    """Findet das hÃ¶chste Level in einer Liste."""
    return max(level_liste)

def ist_quest_verfuegbar(level, benoetigt_level):
    """PrÃ¼ft, ob eine Quest verfÃ¼gbar ist."""
    return level >= benoetigt_level

# âŒ Schlechte Namen (unklar oder falsche Konvention)
# def berechneHeilung():  # CamelCase
# def bs(b, m):  # Zu kurz, nicht beschreibend
# def healing_calculation():  # Substantiv zuerst

print("=== Gute Funktionsnamen in Aktion ===")
heilung = berechne_heilung(10, 2)
print(f"Heilung: {heilung}")

levels = [5, 8, 12, 3, 15]
max_level = finde_max_level(levels)
print(f"HÃ¶chstes Level: {max_level}")

quest_ok = ist_quest_verfuegbar(8, 10)
print(f"Quest verfÃ¼gbar: {quest_ok}")
````
