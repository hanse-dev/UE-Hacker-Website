## âœ¨ Was sind Stall-Routinen (Funktionen)?

Stell dir eine Funktion wie ein **Rezept in einem Stallhandbuch** vor:

- ğŸ“– **Das Rezept schreiben** (Definition): Du schreibst einmal auf, wie eine Routine funktioniert
- ğŸª„ **Die Routine durchfÃ¼hren** (Aufruf): Jedes Mal, wenn du die Routine brauchst, fÃ¼hrst du sie einfach aus

**Warum ist das so nÃ¼tzlich?**

- ğŸ”„ **Keine Wiederholungen**: Schreibe den Code einmal, nutze ihn tausendmal!
- ğŸ“š **Klare Struktur**: Dein Stallhandbuch ist ordentlich organisiert
- ğŸ§© **GroÃŸe Aufgaben lÃ¶sen**: Zerlege komplexe Routinen in kleine, verstÃ¤ndliche Teile
- ğŸ¯ **Fehler vermeiden**: Ã„ndere die Routine an einer Stelle, sie funktioniert Ã¼berall

**Fun Fact:** Echte PferdehÃ¶fe und Reitschulen nutzen unzÃ¤hlige Routinen, die zusammenarbeiten!

### ğŸ“– Routinen schreiben vs. Routinen durchfÃ¼hren â€“ Der Unterschied!

**Das ist der wichtigste Punkt in der Stall-Akademie:**

- ğŸ“ **Definieren** (`def meine_routine():`): Du schreibst die Anweisung ins Stallhandbuch. Die Routine wird **noch NICHT durchgefÃ¼hrt**! Du legst nur fest, was passieren soll.

- âš¡ **Aufrufen** (`meine_routine()`): Du fÃ¼hrst die Routine aus! **Jetzt** passiert die Arbeit!

**Wichtig:** Eine Routine im Buch (definiert) macht nichts, bis du sie ausfÃ¼hrst (aufrufst)!

### ğŸ”® Die Anatomie einer Stall-Routine

**So baust du eine Routine:**

1. **`def routine_name():`** â€“ Das Stichwort `def` startet die Definition
2. **EinrÃ¼ckung** â€“ Alles, was zur Routine gehÃ¶rt, wird eingerÃ¼ckt (4 Leerzeichen)
3. **Aufruf** â€“ FÃ¼hre den Namen mit `()` aus: `routine_name()`

**Profi-Trick:** Routinen kÃ¶nnen andere Routinen aufrufen! So kombinierst du deine Arbeit!

```python
# ğŸ´ Level 1: Deine erste Stall-Routine!

# DEFINITION der Routine (ins Stallhandbuch schreiben)
def sag_hallo():
    print("âœ¨ Hallo! SchÃ¶n, dass du da bist.")
    print("ğŸª„ Dies ist deine erste Stall-Routine!")

def haupt_routine():
    print("ğŸ”® Ich bin die Hauptroutine und rufe jetzt eine andere Routine auf...")
    sag_hallo()  # Eine Routine ruft eine andere!

# AUFRUF der Routine (Arbeit durchfÃ¼hren!)
print("ğŸ“– Routine wird ins Handbuch geschrieben...")
print("âš¡ Jetzt fÃ¼hren wir die Routine aus!\n")
haupt_routine()

print("\nğŸ‰ +100 XP: Du hast deine erste Routine durchgefÃ¼hrt!")
```

### ğŸ§ª Skill: Routinen-Arbeitsmittel (Parameter)

Manchmal braucht eine Routine **Arbeitsmittel**, um zu funktionieren! Eine FÃ¼tterungs-Routine braucht vielleicht den Pferdenamen, eine Pflege-Routine braucht die Art der Pflege.

Diese "Arbeitsmittel" nennen wir **Parameter**:

- ğŸ“ **Definition**: `def begruesse(name):` â€“ `name` ist ein Parameter (Platzhalter)
- âš¡ **Aufruf**: `begruesse("Luna")` â€“ `"Luna"` ist das Argument (das echte Arbeitsmittel)

**Merke:** Parameter = Platzhalter im Rezept, Argument = echtes Arbeitsmittel bei der AusfÃ¼hrung!

```python
# ğŸ´ Level 2: Routinen mit Arbeitsmitteln!

# Definition mit Parameter 'name' (Platzhalter fÃ¼r ein Arbeitsmittel)
def persoenliche_begruessung(name):
    print(f"ğŸŒŸ Willkommen auf dem Pferdehof, {name}!")
    print(f"âœ¨ MÃ¶ge deine Arbeit mit den Pferden erfolgreich sein, {name}!")

# Aufrufe mit verschiedenen Argumenten (echte Arbeitsmittel)
print("ğŸ§™ Die Stallmeisterin begrÃ¼ÃŸt die neuen Mitarbeiter:innen:\n")
persoenliche_begruessung("Alex")
print()
persoenliche_begruessung("Maria")

print("\nğŸ‰ +150 XP: Du kannst jetzt Routinen mit Arbeitsmitteln erstellen!")
```

### ğŸ’ Power-Move: `return` â€“ Praktische Ergebnisse zurÃ¼ckbekommen

Stell dir vor, du fragst eine Pferdeexpertin: "Wie viel Futter braucht ein Pferd?" 

Du willst nicht, dass sie das Ergebnis nur **ansagt** (`print`), sondern dass sie es **dir gibt** (`return`), damit du damit weiterarbeiten kannst!

**Das macht `return`:**

- ğŸ Gibt ein Ergebnis an die Stelle zurÃ¼ck, wo die Routine durchgefÃ¼hrt wurde
- ğŸ›‘ Beendet die Routine sofort
- ğŸ’¾ ErmÃ¶glicht es, das Ergebnis zu speichern und weiterzuverwenden

**Wann brauchst du `return`?**
- âœ… Wenn du das Ergebnis speichern willst
- âœ… Wenn du damit weiterrechnen willst
- âœ… Wenn du es an eine andere Routine Ã¼bergeben willst

**Wann reicht `print`?**
- âœ… Wenn du nur etwas anzeigen willst (wie eine Nachricht)

```python
# ğŸ´ Level 3: Routinen mit praktischen Ergebnissen!

# Diese Routine GIBT ein Ergebnis zurÃ¼ck (wie ein Geschenk!)
def addiere(a, b):
    ergebnis = a + b
    return ergebnis  # ğŸ Das Ergebnis wird zurÃ¼ckgegeben

# Das Ergebnis wird in 'summe' gespeichert
print("ğŸ§® Die Futterberechnung wird durchgefÃ¼hrt...")
summe = addiere(10, 5)

print(f"ğŸ’ Das praktische Ergebnis: {summe}")
print(f"âœ¨ Wir kÃ¶nnen damit weiterarbeiten: {summe * 2}")
print(f"ğŸ”® Oder es in neuen Berechnungen nutzen: {summe + 100}")

print("\nğŸ‰ +200 XP: Du beherrschst jetzt return!")
```

### ğŸ“š Meister-Tipps: Wie die Profis arbeiten

**Die erfahrene Stallmeisterin verrÃ¤t dir ihre Geheimnisse:**

**1. ğŸ·ï¸ Sprechende Namen (Naming Conventions)**
- âŒ **Schlecht**: `f1()`, `mache_zeug()`, `x()`
- âœ… **Gut**: `berechne_futter()`, `pflege_pferd()`, `saegere_haufen()`

**2. ğŸ¯ Eine Aufgabe pro Routine**

- Nicht: `mache_alles()` âŒ
- Besser: `berechne_futter()`, `zeige_futter()`, `pflege_pferd()` âœ…

**3. ğŸ“ Routinen-Beschreibungen (Docstrings)**
- Schreibe auf, was deine Routine tut!
- Hilft dir und anderen, den Code zu verstehen

```python
# ğŸ´ Level 4: Professionelle Routinen-Dokumentation!

def berechne_pferdejahre(menschenjahre):
    """
    ğŸ Verwandelt Menschenjahre in Pferdejahre.
    
    Eine nÃ¼tzliche Routine, die das Alter zwischen Spezies Ã¼bersetzt!
    
    Parameter:
        menschenjahre (int): Das Alter in Menschenjahren
    
    Returns:
        int: Das Alter in Pferdejahren (x3)
    
    Beispiel:
        >>> berechne_pferdejahre(10)
        30
    """
    return menschenjahre * 3

# Teste die Routine
alter_mensch = 10
alter_pferd = berechne_pferdejahre(alter_mensch)
print(f"ğŸ Ein {alter_mensch}-jÃ¤hriger Mensch ist wie ein {alter_pferd}-jÃ¤hriges Pferd!")

# Zeige die Routinen-Beschreibung
print("\nğŸ“– Routinen-Beschreibung:")
help(berechne_pferdejahre)

print("\nğŸ‰ +250 XP: Du dokumentierst wie eine Meister:in!")
```
