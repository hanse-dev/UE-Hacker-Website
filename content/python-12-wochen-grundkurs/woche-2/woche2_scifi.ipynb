## ğŸ“¦ Systemprotokoll 2: Die vier Quanten-Typen

**Was es ist:** Jede Information auf der Nebula-7 hat einen Quanten-Typ - wie die vier AggregatzustÃ¤nde der Materie.

**Warum nÃ¼tzlich:**
