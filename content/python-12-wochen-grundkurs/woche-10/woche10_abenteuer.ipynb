## üìö Sch√∂pfungs-Konstruktion 1: Die Klasse - Die magische Blaupause

**Was es ist:** Eine Klasse ist eine Vorlage f√ºr Objekte mit Attributen und Methoden.

**Die Anatomie einer Klasse:**
```python
class Held:
    def __init__(self, name, klasse):
        self.name = name
        self.klasse = klasse
    
    def kaempfen(self):
        return f"{self.name} k√§mpft!"
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`class`** - Schl√ºsselwort f√ºr Klassendefinition
2. **`__init__`** - Konstruktor f√ºr Erschaffung
3. **`self`** - Referenz auf die Kreatur selbst
4. **Methoden** - Zauber innerhalb der Klasse

**Warum so n√ºtzlich:**
- üéØ Wiederverwendbare Blaupausen erschaffen
- üîç Eigenschaften und Verhalten b√ºndeln
- ‚ö° Code organisieren und strukturieren
- üé≤ Komplexe Welten modellieren

````python
# Beispiel 1: Erste Klasse erschaffen
class Held:
    def __init__(self, name, klasse, level):
        self.name = name
        self.klasse = klasse
        self.level = level
        self.lebenspunkte = 100
    
    def level_up(self):
        self.level += 1
        self.lebenspunkte += 20
        return f"‚öîÔ∏è {self.name} erreicht Level {self.level}!"
    
    def status(self):
        return f"{self.name} ({self.klasse}): Level {self.level}, LP: {self.lebenspunkte}"

# Helden erschaffen
held1 = Held("Aria", "Magierin", 15)
held2 = Held("Thorin", "Krieger", 18)

print("=== Helden ===")
print(held1.status())
print(held2.status())
print(held1.level_up())
print(held1.status())
````

````python
# Beispiel 2: Attribute und Methoden
class Monster:
    def __init__(self, name, typ, staerke):
        self.name = name
        self.typ = typ
        self.staerke = staerke
        self.beute = []
    
    def beute_fangen(self, item):
        self.beute.append(item)
        return f"{item} zur Beute von {self.name} hinzugef√ºgt"
    
    def info(self):
        beute_liste = ", ".join(self.beute) if self.beute else "keine"
        return f"Monster {self.name} ({self.typ}): St√§rke {self.staerke}, Beute: {beute_liste}"

# Monster erschaffen und jagen
drache = Monster("Feuerdrache", "Feuer", 50)
skelett = Monster("Skelettkrieger", "Untot", 30)

print("=== Monster-System ===")
print(drache.info())
print(skelett.info())
print(drache.beute_fangen("Gold"))
print(skelett.beute_fangen("Knochen"))
print(drache.info())
print(skelett.info())
````

````python
# Beispiel 3: Klassenvariablen und Instanzvariablen
class Gilde:
    # Klassenvariable
    anzahl_gilden = 0
    
    def __init__(self, name, mitglieder):
        # Instanzvariablen
        self.name = name
        self.mitglieder = mitglieder
        self.tresor = 0
        Gilde.anzahl_gilden += 1
    
    def vorstellen(self):
        return f"Gilde {self.name} mit {len(self.mitglieder)} Mitgliedern"
    
    @classmethod
    def get_anzahl_gilden(cls):
        return f"Gilden insgesamt: {cls.anzahl_gilden}"

# Gilden erschaffen
gilde1 = Gilde("Helden von Pyralia", ["Aria", "Thorin"])
gilde2 = Gilde("Schattenw√§chter", ["Luna", "Nova"])
gilde3 = Gilde("Lichtbringer", ["Zara"])

print("=== Gilde-Liste ===")
print(gilde1.vorstellen())
print(gilde2.vorstellen())
print(gilde3.vorstellen())
print(Gilde.get_anzahl_gilden())
````

## üì¶ Sch√∂pfungs-Konstruktion 2: Die Vererbung - Die evolution√§re Blaupause

**Die wichtigsten Vererbungs-Konzepte:**

**Elternklasse (Basisklasse):**
- `class Lebewesen:` - Grundlegende Eigenschaften
- Gemeinsame Attribute und Methoden

**Kindklasse (Abgeleitete Klasse):**
- `class Mensch(Lebewesen):` - Erbt von Lebewesen
- Zus√§tzliche spezialisierte Eigenschaften

**Vorteile:**
- Code-Wiederverwendung
- Hierarchische Struktur
- Polymorphismus

````python
# üîç √úbungen mit Vererbung

# Beispiel 1: Einfache Vererbung
class Lebewesen:
    def __init__(self, name, lebensdauer):
        self.name = name
        self.lebensdauer = lebensdauer
    
    def atmen(self):
        return f"{self.name} atmet"

class Mensch(Lebewesen):
    def __init__(self, name, lebensdauer, beruf):
        super().__init__(name, lebensdauer)
        self.beruf = beruf
    
    def arbeiten(self):
        return f"{self.name} arbeitet als {self.beruf}"

class Elf(Lebewesen):
    def __init__(self, name, lebensdauer, element):
        super().__init__(name, lebensdauer)
        self.element = element
    
    def zaubern(self):
        return f"{self.name} zaubert mit {self.element}-Magie"

# Lebewesen erschaffen
human = Mensch("Alex", 80, "Kommandant")
elf = Elf("Lorien", 500, "Natur")

print("=== Lebensformen ===")
print(human.atmen())
print(human.arbeiten())
print(elf.atmen())
print(elf.zaubern())
````

````python
# Beispiel 2: Methoden √ºberschreiben
# Methoden √ºberschreiben ist ein Kernkonzept der Vererbung.
# Es erlaubt Kind-Klassen, das Verhalten von Eltern-Methoden anzupassen.
#
# Warum Methoden √ºberschreiben?
# 1. Spezialisierung: Kind-Klassen haben speziellere Bed√ºrfnisse
# 2. Verbesserung: Bessere Implementierung f√ºr spezifischen Fall
# 3. Anpassung: Unterschiedliches Verhalten f√ºr unterschiedliche Typen
#
# Wichtige Regeln beim √úberschreiben:
# - Gleicher Methodenname
# - Gleiche Parameter (oder kompatibel)
# - super() kann Eltern-Methode aufrufen

class Waffe:
    def __init__(self, name, schaden):
        self.name = name
        self.schaden = schaden
    
    def angreifen(self):
        # Allgemeine Implementierung f√ºr alle Waffen
        return f"{self.name} verursacht {self.schaden} Schaden"
    
    def status(self):
        return f"{self.name}: {self.schaden} Schaden"

class MagischeWaffe(Waffe):
    def __init__(self, name, schaden, element):
        # super() ruft die __init__ der Elternklasse auf
        # Dadurch werden name und schaden gesetzt
        super().__init__(name, schaden)
        self.element = element  # Zus√§tzliche Eigenschaft
    
    def angreifen(self):
        # √úberschreiben: Magische Waffen verursachen elementaren Schaden
        # Wir nutzen hier die spezifische Element-F√§higkeit statt normalen Schadens
        return f"‚ú® {self.name} verursacht {self.schaden} {self.element}-Schaden"
    
    # Neue Methode, die nur magische Waffen haben
    def zauber_entfesseln(self):
        return f"{self.name} entfesselt {self.element}-Zauber!"
    
    # Status anpassen f√ºr magische Waffe
    def status(self):
        return f"‚ú® {self.name}: {self.schaden} {self.element}-Schaden"

class NormaleWaffe(Waffe):
    def __init__(self, name, schaden, material):
        super().__init__(name, schaden)
        self.material = material
    
    # √úberschreiben mit super() Aufruf
    def status(self):
        # Wir rufen die Eltern-Methode auf und f√ºgen Material hinzu
        basis_status = super().status()
        return f"{basis_status} ({self.material})"
    
    def schaerfen(self):
        return f"{self.name} wird gesch√§rft"

# Waffen testen
schwert = NormaleWaffe("Langschwert", 15, "Stahl")
zauberstab = MagischeWaffe("Feuerstab", 25, "Feuer")
bogen = NormaleWaffe("Langbogen", 12, "Holz")

print("=== Methoden-√úberschreiben-Beispiel ===")
print("\nNormale Waffe:")
normale_waffe = Waffe("Dolch", 8)
print(f"  {normale_waffe.angreifen()}")
print(f"  {normale_waffe.status()}")

print("\nNormale Waffe (erbt und erweitert):")
print(f"  {schwert.angreifen()}")
print(f"  {schwert.status()} - nutzt super() in status()")
print(f"  {schwert.schaerfen()} - neue Methode")

print("\nMagische Waffe (√ºberschreibt angreifen()):")
print(f"  {zauberstab.angreifen()} - komplett andere Implementierung")
print(f"  {zauberstab.status()} - angepasster Status")
print(f"  {zauberstab.zauber_entfesseln()} - spezifische Magie-Methode")

print("\nüéØ Warum ist das n√ºtzlich?")
print("- Polymorphismus: Gleicher Methodenname, unterschiedliches Verhalten")
print("- Spezialisierung: Jede Klasse passt Verhalten an ihre Bed√ºrfnisse an")
print("- Wiederverwendung: Gemeinsamer Code in Basisklasse")
````

````python
# Beispiel 3: Mehrfachvererbung
# Mehrfachvererbung erlaubt es einer Klasse, von mehreren Elternklassen zu erben.
# Dies ist n√ºtzlich, wenn ein Objekt mehrere Rollen oder F√§higkeiten haben soll.
# In der Fantasy-Welt kann ein Held gleichzeitig Krieger, Magier UND Heiler sein.
#
# Vorteile der Mehrfachvererbung:
# - Code-Wiederverwendung aus mehreren Quellen
# - Kombination unterschiedlicher F√§higkeiten
# - Flexible Rollenverteilung
#
# Nachteile/Vorsicht:
# - Diamond-Problem (gleiche Methode in mehreren Eltern)
# - Komplexe Abh√§ngigkeiten
# - Schwieriger zu verstehen
#
# Method Resolution Order (MRO) bestimmt, in welcher Reihenfolge Python nach Methoden sucht.

class Kaempfer:
    def kaempfen(self):
        return "Kann im Nahkampf k√§mpfen"
    
    def verteidigen(self):
        return "Kann sich verteidigen"

class Magier:
    def zaubern(self):
        return "Kann Zauber wirken"
    
    def magie_forschen(self):
        return "Kann neue Magie erforschen"

class Heiler:
    def heilen(self):
        return "Kann Wunden heilen"
    
    def kraeuter_sammeln(self):
        return "Kann heilende Kr√§uter sammeln"

# Der AllroundHeld erbt von ALLEN drei Klassen
class AllroundHeld(Kaempfer, Magier, Heiler):
    def __init__(self, name, titel):
        self.name = name
        self.titel = titel
        self.lebenspunkte = 100
        self.mana = 50
    
    def vorstellen(self):
        faehigkeiten = []
        faehigkeiten.append(self.kaempfen())
        faehigkeiten.append(self.zaubern())
        faehigkeiten.append(self.heilen())
        return f"{self.titel} {self.name}: {' | '.join(faehigkeiten)}"
    
    # Spezielle Methode, die F√§higkeiten kombiniert
    def heilenden_schlag_ausfuehren(self):
        # Nutzt Methoden von unterschiedlichen Eltern
        return f"{self.kaempfen()} und {self.heilen()} gleichzeitig"
    
    def magischen_schild_erzeugen(self):
        # Kombiniert Magie und Verteidigung
        return f"{self.zaubern()} zum {self.verteidigen()}"

# Allround-Held testen
held = AllroundHeld("Alex", "Meisterheld")
print("=== Allround-Held mit Mehrfachvererbung ===")
print(held.vorstellen())
print(f"Spezialf√§higkeit: {held.heilenden_schlag_ausfuehren()}")
print(f"Magische Verteidigung: {held.magischen_schild_erzeugen()}")
print(f"Alle F√§higkeiten: Magieforschung: {held.magie_forschen()}, Kr√§uter: {held.kraeuter_sammeln()}")

# Zeige die Method Resolution Order
print(f"\nMethod Resolution Order: {AllroundHeld.__mro__}")
print("Das zeigt, in welcher Reihenfolge Python nach Methoden sucht!")
````

## ‚ú® Sch√∂pfungs-Konstruktion 3: Der Polymorphismus - Die formwandelnde Blaupause

**Die m√§chtigsten Polymorphismus-Techniken:**

**Methoden-Polymorphismus:**
- Gleiche Methode, unterschiedliche Implementierung
- `objekt.methode()` - Dynamische Bindung

**Magische Methoden:**
- `__str__`, `__add__`, `__len__` etc.
- Benutzerdefiniertes Verhalten

**Duck Typing:**
- "If it walks like a duck..."
- Fokus auf Verhalten, nicht Typ

````python
# Beispiel 1: Polymorphismus mit Methoden
# Polymorphismus bedeutet "viele Formen" - das gleiche Interface, unterschiedliche Implementierungen.
#
# Warum ist Polymorphismus so m√§chtig?
# 1. Flexibilit√§t: Code kann mit verschiedenen Typen arbeiten
# 2. Erweiterbarkeit: Neue Typen k√∂nnen hinzugef√ºgt werden, ohne Code zu √§ndern
# 3. Vereinfachung: Gleiche Methodenaufrufe f√ºr unterschiedliche Objekte
#
# Real-World Beispiel in der Fantasy-Welt:
# - Verschiedene Charaktere (Krieger, Magier, Schurke) haben alle die Methode spezialfaehigkeit()
# - Aber jeder implementiert sie anders je nach seinen F√§higkeiten
# - Eine Funktion kann alle gleich behandeln, ohne Details zu kennen

class Krieger:
    def __init__(self, name, waffe):
        self.name = name
        self.waffe = waffe
        self.rage = 100
    
    def spezialfaehigkeit(self):
        # Krieger nutzen Wut f√ºr m√§chtige Angriffe
        self.rage -= 30
        return f"{self.name} ({self.waffe}) f√ºhrt Wutangriff aus! (Wut: {self.rage}%)"
    
    def verteidigen(self):
        return "Kann Schildblock verwenden"

class Magier:
    def __init__(self, name, element):
        self.name = name
        self.element = element
        self.mana = 100
    
    def spezialfaehigkeit(self):
        # Magier wirken m√§chtige Zauberspr√ºche
        self.mana -= 40
        return f"{self.name} wirkt {self.element}-Feuerball! (Mana: {self.mana}%)"
    
    def studieren(self):
        return "Kann in der Bibliothek forschen"

class Schurke:
    def __init__(self, name, spezialisierung):
        self.name = name
        self.spezialisierung = spezialisierung
        self.energy = 100
    
    def spezialfaehigkeit(self):
        # Schurken nutzen Tarnung und √úberraschung
        self.energy -= 25
        return f"{self.name} ({self.spezialisierung}) f√ºhrt Hinterhalt aus! (Energie: {self.energy}%)"
    
    def stehlen(self):
        return "Kann Items stehlen"

# Polymorphe Funktion - das Herz des Polymorphismus!
# Diese Funktion wei√ü nichts √ºber den genauen Typ des K√§mpfers
# Sie ruft nur die Methode spezialfaehigkeit() auf - polymorph!
def kampf_test(kaempfer):
    print(f"Test: {kaempfer.spezialfaehigkeit()}")
    # Polymorphismus in Aktion: Jeder Typ hat seine eigene Implementierung
    # Aber wir rufen immer die gleiche Methode auf!

# Erweiterte polymorphe Funktion
def arena_kampf(kaempfer_liste):
    print("\n‚öîÔ∏è ARENA-KAMPF BEGINNT:")
    for i, kaempfer in enumerate(kaempfer_liste, 1):
        print(f"{i}. {kaempfer.spezialfaehigkeit()}")
    print("Alle K√§mpfer haben ihre F√§higkeiten gezeigt!")

# Verschiedene K√§mpfer erstellen
kaempfer = [
    Krieger("Thorin", "Axt"),
    Magier("Aria", "Feuer"),
    Schurke("Shadow", "Dolch"),
    Krieger("Borin", "Schwert"),
    Magier("Luna", "Eis")
]

print("=== Polymorphismus-Demo ===")
print("\n1. Einzel-Tests:")
for k in kaempfer:
    kampf_test(k)

arena_kampf(kaempfer)

print("\nüéØ Vorteile des Polymorphismus:")
print("‚úÖ Neue Charakter-Typen k√∂nnen hinzugef√ºgt werden (z.B. Priester, Barde)")
print("‚úÖ Die arena_kampf()-Funktion muss nicht ge√§ndert werden!")
print("‚úÖ Jeder Typ nutzt seine St√§rken (St√§rke, Magie, Geschicklichkeit)")

# Beispiel: Neuer Typ ohne Code-√Ñnderung!
class Priester:
    def __init__(self, name, gottheit):
        self.name = name
        self.gottheit = gottheit
        self.glaube = 100
    
    def spezialfaehigkeit(self):
        self.glaube -= 20
        return f"{self.name} betet zu {self.gottheit} und heilt! (Glaube: {self.glaube}%)"
    
    def segnen(self):
        return "Kann Verb√ºndete segnen"

kaempfer.append(Priester("Marcus", "Licht"))
print("\n2. Mit neuem Priester-Typ (ohne Code-Anpassung!):")
arena_kampf(kaempfer)
````

````python
# Beispiel 2: Magic Methods (Dunder Methods)
class Artefakt:
    def __init__(self, name, macht):
        self.name = name
        self.macht = macht
    
    def __str__(self):
        return f"Artefakt({self.name}, Macht: {self.macht})"
    
    def __add__(self, other):
        neue_macht = self.macht + other.macht
        return Artefakt(f"{self.name}+{other.name}", neue_macht)
    
    def __len__(self):
        return self.macht
    
    def __eq__(self, other):
        return self.name == other.name and self.macht == other.macht

# Artefakte testen
a1 = Artefakt("Feuerstein", 50)
a2 = Artefakt("Wasserkristall", 30)
a3 = Artefakt("Feuerstein", 50)

print("=== Artefakt-Operationen ===")
print(f"a1: {a1}")
print(f"a2: {a2}")
print(f"a1 + a2: {a1 + a2}")
print(f"Macht von a1: {len(a1)}")
print(f"a1 == a3: {a1 == a3}")
````

````python
# Beispiel 3: Duck Typing
class Drache:
    def mach_geraeusch(self):
        return "Br√ºllen!"

class MenschDerWieEinDracheKlingt:
    def mach_geraeusch(self):
        return "Br√ºllen! (aber menschlich)"

class MagischerDrache:
    def mach_geraeusch(self):
        return "*magisches Br√ºllen*"

# Duck Typing Funktion
def was_ist_das(drache):
    return drache.mach_geraeusch()

# Verschiedene "Drachen" testen
drachen = [Drache(), MenschDerWieEinDracheKlingt(), MagischerDrache()]

print("=== Duck Typing ===")
for i, drache in enumerate(drachen, 1):
    print(f"Drache {i}: {was_ist_das(drache)}")
````

**Hast du die Fehler gefunden?** Hier sind die L√∂sungen:

````python
# ‚úÖ L√∂sung Bug #1: Fehlender Doppelpunkt und self

# Problem: Klasse braucht Doppelpunkt, __init__ braucht self
class Held:
    def __init__(self, name, klasse):
        self.name = name
        self.klasse = klasse

held = Held('Aria', 'Magierin')
print(f'Held: {held.name}, Klasse: {held.klasse}')
````

````python
# ‚úÖ L√∂sung Bug #2: Fehlender super() Aufruf

# Problem: Kindklasse nutzt nicht super() f√ºr Initialisierung
class Lebewesen:
    def __init__(self, name):
        self.name = name

class Mensch(Lebewesen):
    def __init__(self, name, beruf):
        super().__init__(name)  # Eltern-Konstruktor aufrufen
        self.beruf = beruf

mensch = Mensch('Alex', 'Kommandant')
print(f'Mensch: {mensch.name}, Beruf: {mensch.beruf}')
````

````python
# ‚úÖ L√∂sung Bug #3: Fehlende __str__ Methode

# Problem: Klasse hat keine String-Repr√§sentation
class Punkt:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f'Punkt({self.x}, {self.y})'

p = Punkt(3, 4)
print(p)
````
