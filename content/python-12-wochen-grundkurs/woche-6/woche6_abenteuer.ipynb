## ğŸ“š Sammlungs-Zauber 1: Die Listen-Erstellung - Der erste Schatz

**Was es ist:** Eine Liste ist eine geordnete Sammlung von Elementen, die du mit eckigen Klammern erstellst.

**Die Anatomie einer Liste:**
```python
# Leere Liste erstellen
leere_liste = []

# Liste mit Elementen erstellen

```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`[]`** - Eckige Klammern definieren eine Liste
2. **`,`** - Kommas trennen die Elemente
3. **`""`** - Strings werden in AnfÃ¼hrungszeichen gesetzt
4. **`[]`** - Mit Index greifst du auf Elemente zu (beginnt bei 0)

**Warum so nÃ¼tzlich:**
- ğŸ¯ Mehrere Werte in einer Variable speichern
- ğŸ” Geordnete Struktur fÃ¼r Daten
- âš¡ Flexibel erweiterbar und verÃ¤nderbar

````python
# Beispiel 1: Einfache Listen erstellen
# Leere Liste
leere_liste = []
print(f"Leere Liste: {leere_liste}")

# Liste mit SchÃ¤tzen
schaetze = ["GoldmÃ¼nzen", "Silberbarren", "Edelsteine"]
print(f"Schatzliste: {schaetze}")

# Liste mit Zahlen
level = [1, 5, 10, 15, 20]
print(f"Level-Liste: {level}")

# Gemischte Liste
mixed = ["Schwert", 25, True, "Schild"]
print(f"Gemischte Liste: {mixed}")
````

## ğŸ“¦ Sammlungs-Zauber 2: Elemente verwalten - Der Schatzmeister

**Die wichtigsten Listen-Methoden:**

**Elemente hinzufÃ¼gen:**
- `append(element)` - FÃ¼gt am Ende hinzu
- `insert(index, element)` - FÃ¼gt an bestimmter Position ein

**Elemente entfernen:**
- `remove(element)` - Entfernt erstes Vorkommen
- `pop(index)` - Entfernt und gibt Element zurÃ¼ck
- `clear()` - Leert die gesamte Liste

````python
# ğŸ” Ãœbungen mit append() und insert()

# Beispiel 1: Elemente hinzufÃ¼gen
beutesack = []
print(f"Leerer Beutesack: {beutesack}")

# Mit append() am Ende hinzufÃ¼gen
beutesack.append("Goldring")
print(f"Nach append: {beutesack}")

beutesack.append("Zauberstab")
beutesack.append("Heiltrank")
print(f"Voll gefÃ¼llt: {beutesack}")

# Mit insert() an bestimmter Position einfÃ¼gen
beutesack.insert(0, "SchlÃ¼ssel")  # Am Anfang
print(f"Nach insert(0): {beutesack}")

beutesack.insert(2, "Amulett")  # An Position 2
print(f"Nach insert(2): {beutesack}")
````

````python
# Beispiel 2: Elemente entfernen
inventar = ["Schwert", "Schild", "Trank", "Bogen", "Pfeile", "Trank"]
print(f"Original-Inventar: {inventar}")

# Mit remove() erstes Vorkommen entfernen
inventar.remove("Trank")
print(f"Nach remove('Trank'): {inventar}")

# Mit pop() letztes Element entfernen und zurÃ¼ckgeben
letztes_item = inventar.pop()
print(f"Nach pop(): {inventar}")
print(f"Entferntes Item: {letztes_item}")

# Mit pop(index) bestimmtes Element entfernen
zweites_item = inventar.pop(1)
print(f"Nach pop(1): {inventar}")
print(f"Entferntes Item: {zweites_item}")

# Mit clear() alles lÃ¶schen
inventar.clear()
print(f"Nach clear(): {inventar}")
````

````python
# Beispiel 3: Listen kombinieren
schaetze = ["Gold", "Silber", "Edelsteine"]
waffen = ["Schwert", "Bogen", "Axt"]
trÃ¤nke = ["Heilung", "Mana", "StÃ¤rke"]

print("=== Listen kombinieren ===")
print(f"SchÃ¤tze: {schaetze}")
print(f"Waffen: {waffen}")
print(f"TrÃ¤nke: {trÃ¤nke}")

# Mit + Operator kombinieren
alles = schaetze + waffen + trÃ¤nke
print(f"Alles kombiniert: {alles}")

# Mit extend() eine Liste zur anderen hinzufÃ¼gen
schaetze.extend(waffen)
print(f"SchÃ¤tze erweitert: {schaetze}")
````

## âœ¨ Sammlungs-Zauber 3: Listen durchsuchen und sortieren - Der Ordnungs-Zauber

**Die mÃ¤chtigsten Such-Methoden:**

**Suchen und prÃ¼fen:**
- `index(element)` - Gibt Index des Elements zurÃ¼ck
- `element in liste` - PrÃ¼ft ob Element vorhanden
- `count(element)` - ZÃ¤hlt Vorkommen

**Sortieren:**
- `sort()` - Sortiert die Liste selbst
- `sorted(liste)` - Gibt sortierte Kopie zurÃ¼ck
- `reverse()` - Kehrt Reihenfolge um

````python
# Beispiel 1: In Listen suchen
inventar = ["Schwert", "Schild", "Trank", "Bogen", "Schwert", "Pfeile"]

print("=== In Listen suchen ===")
print(f"Inventar: {inventar}")

# Mit index() Position finden
position_schild = inventar.index("Schild")
print(f"Schild an Position: {position_schild}")

# Mit in() prÃ¼fen ob vorhanden
hat_bogen = "Bogen" in inventar
hat_zauberstab = "Zauberstab" in inventar
print(f"Bogen vorhanden: {hat_bogen}")
print(f"Zauberstab vorhanden: {hat_zauberstab}")

# Mit count() zÃ¤hlen
anzahl_schwerter = inventar.count("Schwert")
print(f"Anzahl Schwerter: {anzahl_schwerter}")
````

````python
# Beispiel 2: Listen sortieren
zahlen = [5, 2, 8, 1, 9, 3]
namen = ["Zara", "Anna", "Max", "Berta", "Klaus"]

print("=== Listen sortieren ===")
print(f"Original-Zahlen: {zahlen}")
print(f"Original-Namen: {namen}")

# Mit sort() direkt sortieren
zahlen.sort()
print(f"Zahlen nach sort(): {zahlen}")

# Mit sorted() Kopie sortieren
sortierte_namen = sorted(namen)
print(f"Namen Original: {namen}")
print(f"Namen sortiert: {sortierte_namen}")

# Umgekehrt sortieren
zahlen.sort(reverse=True)
print(f"Zahlen absteigend: {zahlen}")

# Mit reverse() Reihenfolge umkehren
namen.reverse()
print(f"Namen umgekehrt: {namen}")
````

````python
# Beispiel 3: Fortgeschrittene Listen-Operationen
monster = ["Goblin", "Ork", "Troll", "Drache", "Goblin", "Ork"]

print("=== Fortgeschrittene Operationen ===")
print(f"Monster-Liste: {monster}")

# Einzigartige Elemente (mit set)
einzigartige_monster = list(set(monster))
print(f"Einzigartige Monster: {einzigartige_monster}")

# Listen mit for-Schleife durchlaufen
print("\nAlle Monster:")
for i, m in enumerate(monster):
    print(f"  Position {i}: {m}")

# List Comprehension (fortgeschritten)
lange_namen = [m for m in monster if len(m) > 4]
print(f"\nMonster mit langen Namen: {lange_namen}")
````
