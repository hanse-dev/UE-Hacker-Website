### ğŸ Tutorial: Deine erste Pferde-Karteikarte erstellen

Die erfahrene Reitlehrerin zeigt dir die Grundformel:

### ğŸ” Skill: Datentypen erkennen

Jede Information hat einen **Typ**. Mit `type()` kannst du herausfinden, welchen Typ eine Variable hat:

- `str` â†’ String (Text)
- `int` â†’ Integer (Ganze Zahl)
- `float` â†’ Float (Kommazahl)
- `bool` â†’ Boolean (True/False)

**Warum ist das wichtig?** Verschiedene Typen haben verschiedene FÃ¤higkeiten! Du kannst Zahlen addieren, aber Text wird anders behandelt.

````python
# ğŸ´ Level 2: Datentypen im Stall prÃ¼fen

name = "Stella"
alter = 5
groesse = 1.58
ist_schimmel = True

print("ğŸ” Datentypen-Check:")
print(f"Name ist vom Typ: {type(name).__name__}")
print(f"Alter ist vom Typ: {type(alter).__name__}")
print(f"GrÃ¶ÃŸe ist vom Typ: {type(groesse).__name__}")
print(f"Ist Schimmel ist vom Typ: {type(ist_schimmel).__name__}")

print(f"\nğŸ‰ +150 Huf-Punkte: Du erkennst jetzt alle Datentypen!")
````

### âš¡ Power-Moves: Mit Zahlen rechnen

Variablen mit Zahlen kÃ¶nnen wie normale Zahlen verwendet werden:

- `+` Addition
- `-` Subtraktion
- `*` Multiplikation
- `/` Division
- `%` Modulo (Rest bei Division)
- `**` Potenz

````python
# ğŸ´ Level 3: Futter-Berechnungen

# Grundrechenarten
heu_pro_pferd = 5  # kg
anzahl_pferde = 8
gesamt_heu = heu_pro_pferd * anzahl_pferde
print(f"ğŸŒ¾ Heu pro Pferd: {heu_pro_pferd} kg")
print(f"ğŸ´ Anzahl Pferde: {anzahl_pferde}")
print(f"ğŸ“¦ Gesamt-Heu: {gesamt_heu} kg")

# Division und Durchschnitt
gesamtkosten = 240
kosten_pro_pferd = gesamtkosten / anzahl_pferde
print(f"\nğŸ’° Gesamtkosten: {gesamtkosten} Euro")
print(f"ğŸ’µ Pro Pferd: {kosten_pro_pferd} Euro")

# Modulo - fÃ¼r Wochentage!
tag = 23
wochentag = tag % 7
print(f"\nğŸ“… Tag: {tag}")
print(f"ğŸ—“ï¸ Wochentag (0-6): {wochentag}")

print(f"\nğŸ‰ +200 Huf-Punkte: Du kannst jetzt Futter berechnen!")
````

### ğŸ“ Text-Magie: Namen kombinieren

Mit Strings (Text) kannst du auch rechnen â€“ aber anders!

- `+` verbindet Texte (Konkatenation)
- `*` wiederholt Texte
- `len()` gibt die LÃ¤nge zurÃ¼ck

````python
# ğŸ´ Level 4: Pferde-Namen und Beschreibungen

vorname = "Prinz"
nachname = "von Sonnental"
voller_name = vorname + " " + nachname
print(f"ğŸ‘‘ Voller Name: {voller_name}")

# Text wiederholen
ruf = "Hier! " * 3
print(f"ğŸ“¢ Lockruf: {ruf}")

# Text-LÃ¤nge
spitzname = "Blitzschnell"
laenge = len(spitzname)
print(f"\nâœ¨ Spitzname: {spitzname}")
print(f"ğŸ“ LÃ¤nge: {laenge} Buchstaben")

print(f"\nğŸ‰ +250 Huf-Punkte: Du beherrschst Text-Operationen!")
````

### ğŸ”„ Die Kunst der Typumwandlung

Manchmal musst du einen Typ in einen anderen verwandeln:

- `int()` â†’ Wandelt in Integer um
- `float()` â†’ Wandelt in Float um
- `str()` â†’ Wandelt in String um
- `bool()` â†’ Wandelt in Boolean um

**Wichtig:** Nicht alle Umwandlungen funktionieren! `int("Blitz")` gibt einen Fehler.

````python
# ğŸ´ Level 5: Daten umwandeln

# String zu Integer
alter_text = "12"
alter_zahl = int(alter_text)
print(f"ğŸ“ Als Text: {alter_text}")
print(f"ğŸ”¢ Als Zahl: {alter_zahl}")
print(f"â• NÃ¤chstes Jahr: {alter_zahl + 1}")

# Integer zu String (fÃ¼r Nachrichten)
aepfel = 15
nachricht = "Du hast " + str(aepfel) + " Ã„pfel!"
print(f"\nğŸ {nachricht}")

# Float zu Integer (Rundung)
gewicht = 487.8
gewicht_gerundet = int(gewicht)
print(f"\nâš–ï¸ Gewicht (genau): {gewicht} kg")
print(f"âš–ï¸ Gewicht (gerundet): {gewicht_gerundet} kg")

# String zu Float
groesse_text = "1.62"
groesse_zahl = float(groesse_text)
print(f"\nğŸ“ GrÃ¶ÃŸe: {groesse_zahl} m")

print(f"\nğŸ‰ +200 Huf-Punkte: Du meisterst Typumwandlungen!")
````

### ğŸŒŸ Erweiterte Variablen-Tricks

Fortgeschrittene Techniken fÃ¼r erfahrene Pferdefreunde:

````python
# ğŸ´ Level 6: Profi-Techniken

# Mehrere Variablen gleichzeitig zuweisen
pferd1, pferd2, pferd3 = "Luna", "Stella", "Blitz"
print(f"ğŸ Pferde: {pferd1}, {pferd2}, {pferd3}")

# Variablen tauschen
box_a = "Luna"
box_b = "Stella"
print(f"\nğŸ”„ Vorher: Box A={box_a}, Box B={box_b}")
box_a, box_b = box_b, box_a
print(f"ğŸ”„ Nachher: Box A={box_a}, Box B={box_b}")

# Verkettete Zuweisungen
heu = hafer = moehren = 10
print(f"\nğŸŒ¾ Heu: {heu} kg, Hafer: {hafer} kg, MÃ¶hren: {moehren} kg")

# Augmented Assignment (Kurzschreibweise)
aepfel = 20
print(f"\nğŸ Start-Ã„pfel: {aepfel}")
aepfel += 10  # Gleich wie: aepfel = aepfel + 10
print(f"ğŸ Nach Einkauf: {aepfel}")
aepfel -= 5   # Pferde haben welche gegessen
print(f"ğŸ Nach FÃ¼tterung: {aepfel}")

print(f"\nğŸ‰ +300 Huf-Punkte: Du bist ein Variablen-Profi!")
````
