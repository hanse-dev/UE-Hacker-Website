## ğŸ“š Lektion 1: Die for-Schleife - Der Dressur-Kreis

**Was es ist:** Eine for-Schleife wiederholt Code fÃ¼r jedes Element in einer Sequenz.

**Die Anatomie einer for-Schleife:**
```python
for variable in sequenz:
    # Code, der fÃ¼r jedes Element ausgefÃ¼hrt wird
```

**Schritt-fÃ¼r-Schritt-ErklÃ¤rung:**
1. **`for`** - Das Kommando, das die Wiederholung startet
2. **`variable`** - Ein Platzhalter, der bei jedem Durchlauf das aktuelle Element enthÃ¤lt
3. **`in`** - Verbindet die Variable mit der Sequenz
4. **`sequenz`** - Die Sammlung, durch die wir iterieren (z.B. Liste, String, range())
5. **`:`** - Beendet die Schleifen-Definition
6. **EingerÃ¼ckter Code** - Wird bei jedem Durchlauf ausgefÃ¼hrt

**Der Iterator:**
- Der Iterator ist wie ein unsichtbarer Reitmeister, der sich merkt, wo wir in der Sequenz sind
- Er startet beim ersten Element
- Nach jedem Durchlauf springt er automatisch zum nÃ¤chsten Element
- Die Schleife endet, wenn kein Element mehr Ã¼brig ist

**Warum so nÃ¼tzlich:**
- ğŸ¯ Automatisiere wiederholende Aufgaben
- ğŸ” Verarbeite jedes Element einzeln
- âš¡ Mache Code effizienter und kÃ¼rzer
- ğŸ² Erstelle Muster und Sequenzen

````python
# Beispiel 1: Einfache for-Schleife mit range()
print("=== Beispiel 1: 5 Runden reiten ===")
# range(5) erzeugt die Zahlen 0, 1, 2, 3, 4
for i in range(5):
    print(f"Runde {i+1}: Das Pferd trabt elegant")

print("\n=== Was passiert hier? ===")
print("1. range(5) erstellt: [0, 1, 2, 3, 4]")
print("2. i wird bei jedem Durchlauf auf das nÃ¤chste Element gesetzt")
print("3. Wir geben i+1 aus, um von 1-5 zu zÃ¤hlen")
print("4. Die Schleife endet nach dem letzten Element")
````

````python
# Beispiel 2: For-Schleife mit Start und Ende
print("=== Beispiel 2: HÃ¼rden von 3 bis 7 ===")
# range(3, 8) erzeugt die Zahlen 3, 4, 5, 6, 7 (8 ist exklusiv!)
for hurde in range(3, 8):
    print(f"HÃ¼rde {hurde}: Sprung!")

print("\n=== Wichtige Beobachtung ===")
print("range(start, ende) enthÃ¤lt start, aber nicht ende!")
print("Wie bei einem Reitplatz: von 3 bis VOR 8")
````

````python
# Beispiel 3: For-Schleife mit Schrittweite
print("=== Beispiel 3: Jede zweite HÃ¼rde ===")
# range(0, 11, 2) erzeugt: 0, 2, 4, 6, 8, 10
for hurde in range(0, 11, 2):
    print(f"HÃ¼rde {hurde}: Gesprungen!")

print("\n=== Die Schrittweite ===")
print("range(start, ende, schritt) Ã¼berspringt HÃ¼rden")
print("Hier: jede zweite HÃ¼rde Ã¼berspringen")
````

````python
# Beispiel 4: For-Schleife Ã¼ber einen String
print("=== Beispiel 4: Buchstaben eines Pferdenamens ===")
name = "Thunder"
print(f"Der Pferdename ist: {name}")
print("Buchstaben fÃ¼r Buchstabe:")

for buchstabe in name:
    print(f"  - {buchstabe}")

print("\n=== Was passiert hier? ===")
print("String wird als Sequenz von Buchstaben behandelt")
print("Jeder Buchstabe wird einzeln durchlaufen")
````

````python
# Beispiel 5: For-Schleife Ã¼ber eine Liste
print("=== Beispiel 5: Pferde im Stall ===")
pferde = ["Thunder", "Storm", "Shadow", "Blaze"]
print("Pferde im Stall:")

for pferd in pferde:
    print(f"  - {pferd}")

print("\n=== Mit Stallnummer und Name ===")
# enumerate gibt uns sowohl Index als auch Wert
for nummer, pferd in enumerate(pferde, start=1):
    print(f"  Stall {nummer}: {pferd}")
````

## ğŸ” Das Geheimnis von range() - Der Zahlenreiter

**range() ist dein treuester Begleiter fÃ¼r Zahlenfolgen!**

### Die drei Formen von range():

1. **`range(n)`** - Zahlen von 0 bis n-1
   ```python
   range(5) â†’ 0, 1, 2, 3, 4
   ```

2. **`range(start, ende)`** - Zahlen von start bis ende-1
   ```python
   range(2, 7) â†’ 2, 3, 4, 5, 6
   ```

3. **`range(start, ende, schritt)`** - Zahlen mit AbstÃ¤nden
   ```python
   range(1, 10, 2) â†’ 1, 3, 5, 7, 9
   ```

**Wichtige Regeln:**
- Die **obere Grenze ist immer exklusiv** (wird nicht mitgezÃ¤hlt)
- **Negative Schritte** sind mÃ¶glich fÃ¼r rÃ¼ckwÃ¤rts zÃ¤hlen
- range() ist **sehr speichereffizient** (erzeugt Zahlen erst bei Bedarf)

````python
# ğŸ” Ãœbungen mit range()

# Ãœbung 1: Countdown zum Start
print("=== Countdown zum Turnierstart ===")
for i in range(5, 0, -1):
    print(f"{i}...")
print("ğŸ‡ LOS!")

# Ãœbung 2: Gerade Bahnnummern
print("\n=== Gerade Bahnnummern ===")
for bahn in range(2, 21, 2):
    print(f"Bahn {bahn}", end=" ")
print()

# Ãœbung 3: Trainingsrunden
print("\n=== Trainingsrunden ===")
for runde in range(1, 6):
    print(f"Runde {runde}: Trab - Galopp - Trab")
````

## ğŸ“¦ Lektion 2: Die while-Schleife - Der Ausdauer-Test

**Was es ist:** Eine while-Schleife wiederholt Code, solange eine Bedingung wahr ist.

**Die Anatomie einer while-Schleife:**
```python
while bedingung:
    # Code, der wiederholt wird, solange bedingung wahr ist
    # WICHTIG: Die Bedingung muss irgendwann falsch werden!
```

**Warum nÃ¼tzlich:**
- â³ Wiederhole, bis etwas eintritt
- ğŸ® Training bis zur ErschÃ¶pfung
- ğŸ” Eingaben prÃ¼fen
- âš¡ Unbekannte Anzahl von Wiederholungen

**âš ï¸ Vorsicht:** Endlosschleifen entstehen, wenn die Bedingung niemals falsch wird!

````python
# Beispiel 1: Einfache while-Schleife
print("=== Beispiel 1: Bis 10 zÃ¤hlen ===")
zaehler = 0
while zaehler <= 10:
    print(f"Hufschlag {zaehler}")
    zaehler += 1  # WICHTIG: HochzÃ¤hlen, sonst Endlosschleife!

print("\n=== Was passiert hier? ===")
print("1. zaehler startet bei 0")
print("2. Solange zaehler <= 10, wird der Code ausgefÃ¼hrt")
print("3. zaehler += 1 erhÃ¶ht den Wert bei jedem Durchlauf")
print("4. Bei zaehler = 11 wird die Bedingung falsch â†’ Schleife endet")
````

````python
# Beispiel 2: Pferd-Training
print("=== Beispiel 2: Training bis zur ErschÃ¶pfung ===")
ausdauer = 100
runde = 1

while ausdauer > 20:
    verbrauch = 15
    ausdauer -= verbrauch
    print(f"Runde {runde}: -{verbrauch} Ausdauer, Ã¼brig: {ausdauer}%")
    runde += 1

print(f"\nğŸ’ª Training beendet nach {runde-1} Runden!")
print("Zeit fÃ¼r eine Pause!")
````

````python
# Beispiel 3: Futter-Abfrage
print("=== Beispiel 3: Futter-Abfrage ===")
futter = ""
versuche = 0
max_versuche = 3

while futter != "Heu" and versuche < max_versuche:
    futter = input("Was fressen Pferde am liebsten? ")
    versuche += 1
    
    if futter == "Heu":
        print("âœ… Richtig!")
    elif versuche >= max_versuche:
        print("âŒ Zu viele Versuche! Die Antwort ist: Heu")
    else:
        print(f"Falsch! Noch {max_versuche - versuche} Versuche")
````

## âœ¨ Lektion 3: break und continue - Die Trainings-Brecher

**break** - Beendet das Training sofort
**continue** - Springt zur nÃ¤chsten Ãœbung

**Warum nÃ¼tzlich:**
- ğŸ›‘ FrÃ¼hzeitiges Beenden bei Problemen
- â­ï¸ Bestimmte Ãœbungen Ã¼berspringen
- ğŸ¯ Effizientes Training
- âš¡ Energie sparen

````python
# Beispiel 1: break bei Problemen
print("=== Beispiel 1: Training bei Schmerz beenden ===")
uebungen = ["AufwÃ¤rmen", "Traben", "Galopp", "Springen", "AbkÃ¼hlen"]

for uebung in uebungen:
    print(f"Ãœbung: {Ã¼bung}")
    if uebung == "Springen":
        print("âš ï¸ Pferd humpelt! Training beendet.")
        break  # Schleife sofort beenden

print("\n=== Ohne break wÃ¼rden alle Ãœbungen gemacht ===")
````

````python
# Beispiel 2: continue fÃ¼r Filter
print("=== Beispiel 2: Nur Dressur-Ãœbungen ===")
uebungen = ["Springen", "Dressur", "Galopp", "Dressur", "Western"]

print("Dressur-Ãœbungen heute:")
for uebung in uebungen:
    # Ãœberspringe, wenn keine Dressur-Ãœbung
    if uebung != "Dressur":
        continue  # Zum nÃ¤chsten Durchlauf springen
    print(f"  - {Ã¼bung}")
````

````python
# Beispiel 3: Komplexes Training
print("=== Beispiel 3: Intensives Training ===")
tag = 1
max_tage = 7

while tag <= max_tage:
    print(f"\nTag {tag}:")
    
    # 3 Trainingseinheiten pro Tag
    for einheit in range(3):
        print(f"  Einheit {einheit+1}: Training")
        
        # Pause bei Ãœberlastung
        if tag == 4 and einheit == 1:
            print("  ğŸ’ª Ãœberlastung! Extra-Pause!")
            continue
    
    tag += 1
    
    # Training abbrechen bei Verletzung
    if tag > 5:
        print("\nğŸ¥ Verletzungsgefahr! Training abgebrochen.")
        break
````
