## üìö Archiv-Struktur 1: Die Stallkarte - Der digitale Pferdepass

**Was es ist:** Ein Dictionary ist eine Sammlung von Schl√ºssel-Wert-Paaren, die du mit geschweiften Klammern erstellst.

**Die Anatomie einer Stallkarte:**
```python
# Leere Stallkarte erstellen
leere_karte = {}

# Stallkarte mit Daten f√ºllen
pferd = {"name": "Thunder", "rasse": "Hannoveraner", "alter": 8}
```

**Schritt-f√ºr-Schritt-Erkl√§rung:**
1. **`{}`** - Geschweifte Klammern definieren eine Stallkarte
2. **`:`** - Doppelpunkt trennt Schl√ºssel vom Wert
3. **`,`** - Kommas trennen die Paare
4. **`[key]`** - Mit Schl√ºssel greifst du auf Daten zu

**Warum so n√ºtzlich:**
- üéØ Daten mit Namen statt Index organisieren
- üîç Schneller Zugriff √ºber Namen
- ‚ö° Flexibel erweiterbar und √§nderbar
- üé≤ Verschiedene Datentypen mischbar

````python
# Beispiel 1: Einfache Stallkarten erstellen
# Leere Stallkarte
leere_karte = {}
print(f"Leere Stallkarte: {leere_karte}")

# Pferde-Stallkarte
pferd = {
    "name": "Thunder",
    "rasse": "Hannoveraner",
    "alter": 8,
    "groesse": 1.72
}
print(f"Pferde-Stallkarte: {pferd}")

# Reiter-Stallkarte
reiter = {
    "name": "Anna",
    "level": "Fortgeschritten",
    "erfahrung": 5,
    "eigene_pferde": True
}
print(f"Reiter-Stallkarte: {reiter}")
````

````python
# Beispiel 2: Auf Stallkarten zugreifen
pferd = {
    "name": "Thunder",
    "rasse": "Hannoveraner",
    "alter": 8,
    "groesse": 1.72
}

print("=== Pferde-Zugriff ===")
print(f"Pferde-Name: {pferd['name']}")
print(f"Rasse: {pferd['rasse']}")
print(f"Alter: {pferd['alter']} Jahre")
print(f"Gr√∂√üe: {pferd['groesse']} m")

# Mit get() Methode (sicherer)
print(f"\nMit get(): {pferd.get('name')}")
print(f"Nicht existierend: {pferd.get('geschlecht', 'Unbekannt')}")
````

````python
# Beispiel 3: Stallkarten √§ndern und erweitern
pferd = {
    "name": "Thunder",
    "rasse": "Hannoveraner",
    "alter": 8
}

print(f"Original: {pferd}")

# Wert √§ndern
pferd["alter"] = 9
print(f"Nach Geburtstag: {pferd}")

# Neues Element hinzuf√ºgen
pferd["besitzer"] = "Anna"
print(f"Nach Besitzer hinzuf√ºgen: {pferd}")

# Element entfernen
entfernt = pferd.pop("rasse")
print(f"Entfernt: {entfernt}")
print(f"Nach entfernen: {pferd}")
````

## üì¶ Archiv-Struktur 2: Die Daten-Kapsel - Die unver√§nderliche Aufzeichnung

**Die wichtigsten Kapsel-Eigenschaften:**

**Erstellung:**
- `kapsel = (1, 2, 3)` - Mit runden Klammern
- `kapsel = 1, 2, 3` - Ohne Klammern
- `kapsel = (1,)` - Ein Element mit Komma

**Unver√§nderlichkeit:**
- Einmal erstellt, k√∂nnen Kapseln nicht ge√§ndert werden
- Perfekt f√ºr konstante Daten
- Gesch√ºtzt vor unbeabsichtigten √Ñnderungen

````python
# üîç √úbungen mit Daten-Kapseln

# Beispiel 1: Kapseln erstellen
koordinaten = (100, 200)
print(f"Koordinaten: {koordinaten}")

# Ohne Klammern
position = 10, 20
print(f"Position: {position}")

# Ein Element
einzel = (42,)
print(f"Einzelne Kapsel: {einzel}")

# Zugriff auf Elemente
print(f"\nX-Koordinate: {koordinaten[0]}")
print(f"Y-Koordinate: {koordinaten[1]}")
````

````python
# Beispiel 2: Kapseln sind unver√§nderlich
trainings_status = ("Aktiv", "Stabil", "Gesund")
print(f"Trainings-Status: {trainings_status}")

# Zugriff funktioniert
print(f"Status: {trainings_status[0]}")

# √Ñndern funktioniert NICHT!
try:
    trainings_status[0] = "Inaktiv"
except TypeError as e:
    print(f"Fehler: {e}")

# Aber Kapseln in Listen k√∂nnen ge√§ndert werden
training = [
    ("Dressur", "Aktiv"),
    ("Springen", "Stabil"),
    ("Western", "Gesund")
]
training[1] = ("Springen", "Inaktiv")
print(f"\nTraining nach √Ñnderung: {training}")
````

````python
# Beispiel 3: Kapsel unpacking und Methoden
pferd_daten = ("Thunder", "Hannoveraner", 8, 1.72)

# Kapsel unpacking
name, rasse, alter, groesse = pferd_daten
print("=== Kapsel Unpacking ===")
print(f"Name: {name}")
print(f"Rasse: {rasse}")
print(f"Alter: {alter}")
print(f"Gr√∂√üe: {groesse} m")

# Kapsel-Methoden
print(f"\nAnzahl Elemente: {len(pferd_daten)}")
print(f"Index von 'Hannoveraner': {pferd_daten.index('Hannoveraner')}")
print(f"Anzahl 'Hannoveraner': {pferd_daten.count('Hannoveraner')}")
````

## ‚ú® Archiv-Struktur 3: Verschachtelte Strukturen - Das Daten-Netzwerk

**Die m√§chtigsten Kombinationen:**

**Stallkarte in Stallkarte:**
- `pferd = {"name": "Thunder", "gesundheit": {"zustand": "Gut", "impfung": "Aktuell"}}`

**Liste von Stallkarten:**
- `pferde = [{"name": "T", "rasse": "H"}, {"name": "L", "rasse": "I"}]`

**Kapsel als Stallkarte-Schl√ºssel:**
- `positionen = {(10, 20): "Box A", (30, 40): "Box B"}`

````python
# Beispiel 1: Stallkarte in Stallkarte
stall = {
    "name": "Sonnental",
    "location": {
        "region": "S√ºddeutschland",
        "koordinaten": (100, 200),
        "bundesland": "Bayern"
    },
    "einrichtungen": {
        "reithalle": "20x40m",
        "reitplatz": "30x60m",
        "paddocks": "5 St√ºck"
    }
}

print("=== Verschachtelte Stallkarte ===")
print(f"Stall: {stall['name']}")
print(f"Region: {stall['location']['region']}")
print(f"Koordinaten: {stall['location']['koordinaten']}")
print(f"Reithalle: {stall['einrichtungen']['reithalle']}")
````

````python
# Beispiel 2: Liste von Stallkarten
pferde = [
    {
        "name": "Thunder",
        "rasse": "Hannoveraner",
        "alter": 8,
        "trainingslevel": "M"
    },
    {
        "name": "Luna",
        "rasse": "Isl√§nder",
        "alter": 6,
        "trainingslevel": "A"
    },
    {
        "name": "Stormy",
        "rasse": "Quarter Horse",
        "alter": 10,
        "trainingslevel": "L"
    }
]

print("=== Pferde-Liste ===")
for pferd in pferde:
    print(f"{pferd['name']} - {pferd['rasse']} ({pferd['alter']} Jahre)")

# Nach Alter filtern
junge = [p for p in pferde if p['alter'] < 8]
print(f"\nJunge Pferde: {[p['name'] for p in junge]}")
````

````python
# Beispiel 3: Komplexe Archiv-Strukturen
# Kapsel als Stallkarte-Schl√ºssel
box_karte = {
    (1, 1): "Thunder",
    (1, 2): "Luna",
    (2, 1): "Stormy"
}

print("=== Box-Karte ===")
for position, pferd in box_karte.items():
    print(f"Box {position}: {pferd}")

# Komplexes Trainingssystem
training = {
    "aktiv": [
        {
            "pferd": "Thunder",
            "disziplin": "Dressur",
            "zeitplan": ("Mo", "Mi", "Fr"),
            "status": "In Bearbeitung"
        }
    ],
    "abgeschlossen": [
        {
            "pferd": "Luna",
            "disziplin": "Western",
            "zeitplan": ("Di", "Do"),
            "status": "Erfolg"
        }
    ]
}

print(f"\nAktive Trainings: {len(training['aktiv'])}")
print(f"Abgeschlossene Trainings: {len(training['abgeschlossen'])}")
````
